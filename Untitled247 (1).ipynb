```python
# ============================================================
# EXPANSION-HISTORY ACTIVATED MATTER COUPLING
# Canonical diagnostics cell (self-contained)
# No external data required
# ============================================================

import numpy as np
from scipy.integrate import solve_ivp

# ------------------------------------------------------------
# 1. Core model definitions
# ------------------------------------------------------------

def u_eq(Om):
    return (1.0 / 3.0) * np.log((1.0 - Om) / Om)

def x_of_z(z, Om, alpha):
    return np.log(1.0 + z) / (alpha * u_eq(Om))

def M_of_z(z, Om, alpha, nu):
    x = x_of_z(z, Om, alpha)
    return np.where(x > 0.0, x**nu, 0.0)

def S_of_z(z, Om, alpha, nu):
    # inactive fraction / survival fraction
    return np.exp(-M_of_z(z, Om, alpha, nu))

def g_of_z(z, Om, alpha, nu):
    # activated fraction
    return 1.0 - S_of_z(z, Om, alpha, nu)

def z_s_exact(Om, alpha):
    # x(z_s) = 1
    return np.exp(alpha * u_eq(Om)) - 1.0

def z_at_activation_fraction(p, Om, alpha, nu):
    # solve g(z_p) = p exactly
    # p must be in (0,1)
    if not (0.0 < p < 1.0):
        raise ValueError("p must lie in (0,1)")
    x = (-np.log(1.0 - p))**(1.0 / nu)
    return np.exp(alpha * u_eq(Om) * x) - 1.0

# ------------------------------------------------------------
# 2. Background expansion
# ------------------------------------------------------------

def E2_bg(z, H0, Om, alpha, nu):
    g = g_of_z(z, Om, alpha, nu)
    return Om * (1.0 + z)**3 * g + (1.0 - Om)

def E_bg(z, H0, Om, alpha, nu):
    return np.sqrt(E2_bg(z, H0, Om, alpha, nu))

def dg_dz(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    x = x_of_z(z, Om, alpha)
    out = np.zeros_like(z)

    mask = x > 0.0
    if np.any(mask):
        xm = x[mask]
        gm = g_of_z(z[mask], Om, alpha, nu)
        out[mask] = (
            nu * xm**(nu - 1.0) * (1.0 - gm)
            / ((1.0 + z[mask]) * alpha * u_eq(Om))
        )
    return out

def dE2_dz(z, H0, Om, alpha, nu):
    g = g_of_z(z, Om, alpha, nu)
    dg = dg_dz(z, Om, alpha, nu)
    return Om * (3.0 * (1.0 + z)**2 * g + (1.0 + z)**3 * dg)

def q_of_z(z, H0, Om, alpha, nu):
    # deceleration parameter
    E2 = E2_bg(z, H0, Om, alpha, nu)
    dE2 = dE2_dz(z, H0, Om, alpha, nu)
    return -1.0 + (1.0 + z) * dE2 / (2.0 * E2)

def rho_de_eff(z, Om, alpha, nu):
    g = g_of_z(z, Om, alpha, nu)
    return (1.0 - Om) - Om * (1.0 + z)**3 * (1.0 - g)

def drho_de_eff_dz(z, Om, alpha, nu):
    g = g_of_z(z, Om, alpha, nu)
    dg = dg_dz(z, Om, alpha, nu)
    return -Om * (3.0 * (1.0 + z)**2 * (1.0 - g) - (1.0 + z)**3 * dg)

def w_de_eff(z, Om, alpha, nu):
    rho = rho_de_eff(z, Om, alpha, nu)
    drho = drho_de_eff_dz(z, Om, alpha, nu)
    return -1.0 + (1.0 + z) * drho / (3.0 * rho)

def Omega_m_bg(z, Om, alpha, nu):
    return Om * (1.0 + z)**3 * g_of_z(z, Om, alpha, nu) / E2_bg(z, 1.0, Om, alpha, nu)

# ------------------------------------------------------------
# 3. Perturbation-sector closures
# ------------------------------------------------------------

def g_pert_additive(z, Om, alpha, nu, eps):
    # OLD rough closure: can exceed 1 at high z
    return g_of_z(z, Om, alpha, nu) + eps

def g_pert_survival_floor(z, Om, alpha, nu, eps):
    # NEW canonical bounded closure:
    # g_pert = 1 - (1-eps) exp(-x^nu)
    #        = eps + (1-eps) g
    #        = g + eps * (1-g)
    return 1.0 - (1.0 - eps) * S_of_z(z, Om, alpha, nu)

def Omega_m_pert_minimal(z, Om, alpha, nu):
    return Om * (1.0 + z)**3 * g_of_z(z, Om, alpha, nu) / E2_bg(z, 1.0, Om, alpha, nu)

def Omega_m_pert_survival_floor(z, Om, alpha, nu, eps):
    return Om * (1.0 + z)**3 * g_pert_survival_floor(z, Om, alpha, nu, eps) / E2_bg(z, 1.0, Om, alpha, nu)

# ------------------------------------------------------------
# 4. Linear growth solver in N = ln(a)
# ------------------------------------------------------------

def z_of_a(a):
    return 1.0 / a - 1.0

def dlnH_dN(a, H0, Om, alpha, nu):
    # N = ln a, z = a^{-1} - 1
    z = z_of_a(a)
    E2 = E2_bg(z, H0, Om, alpha, nu)
    dE2dz = dE2_dz(z, H0, Om, alpha, nu)
    dzdN = -(1.0 + z)
    dE2dN = dE2dz * dzdN
    return 0.5 * dE2dN / E2

def growth_rhs_minimal(N, y, H0, Om, alpha, nu):
    D, Dn = y
    a = np.exp(N)
    src = Omega_m_pert_minimal(z_of_a(a), Om, alpha, nu)
    friction = 2.0 + dlnH_dN(a, H0, Om, alpha, nu)
    Dnn = -(friction * Dn) + 1.5 * src * D
    return [Dn, Dnn]

def growth_rhs_survival_floor(N, y, H0, Om, alpha, nu, eps):
    D, Dn = y
    a = np.exp(N)
    src = Omega_m_pert_survival_floor(z_of_a(a), Om, alpha, nu, eps)
    friction = 2.0 + dlnH_dN(a, H0, Om, alpha, nu)
    Dnn = -(friction * Dn) + 1.5 * src * D
    return [Dn, Dnn]

def solve_growth(H0, Om, alpha, nu, eps=None, a_ini=1e-3, a_fin=1.0, n_eval=500):
    N_ini = np.log(a_ini)
    N_fin = np.log(a_fin)
    N_eval = np.linspace(N_ini, N_fin, n_eval)

    # growing-mode IC in matter era: D ~ a, so D_N = D
    y0 = [a_ini, a_ini]

    if eps is None:
        sol = solve_ivp(
            growth_rhs_minimal, [N_ini, N_fin], y0,
            t_eval=N_eval, args=(H0, Om, alpha, nu),
            rtol=1e-8, atol=1e-10
        )
    else:
        sol = solve_ivp(
            growth_rhs_survival_floor, [N_ini, N_fin], y0,
            t_eval=N_eval, args=(H0, Om, alpha, nu, eps),
            rtol=1e-8, atol=1e-10
        )

    N = sol.t
    a = np.exp(N)
    z = z_of_a(a)
    D = sol.y[0]
    Dn = sol.y[1]

    # normalize to D(z=0)=1
    D /= D[-1]
    Dn /= sol.y[0][-1]

    f = Dn / D
    return {
        "N": N,
        "a": a,
        "z": z,
        "D": D,
        "f": f,
        "fs8_over_sigma8_0": f * D
    }

# ------------------------------------------------------------
# 5. Fiducial parameter set from the draft
# ------------------------------------------------------------

H0 = 74.0
Om = 0.231
alpha = 0.483
nu = 0.886
eps = 0.08

# ------------------------------------------------------------
# 6. Diagnostics
# ------------------------------------------------------------

print("=== Canonical equality-tied activation diagnostics ===")
print(f"Omega_m       = {Om:.6f}")
print(f"alpha         = {alpha:.6f}")
print(f"nu            = {nu:.6f}")
print(f"u_eq          = {u_eq(Om):.6f}")
print(f"u_s           = {alpha * u_eq(Om):.6f}")
print(f"z_s (x=1)     = {z_s_exact(Om, alpha):.6f}")
print(f"z_50 (g=0.5)  = {z_at_activation_fraction(0.5, Om, alpha, nu):.6f}")
print(f"z_90 (g=0.9)  = {z_at_activation_fraction(0.9, Om, alpha, nu):.6f}")
print()

# exact threshold slope diagnostics
slope_x1 = nu / np.e
slope_ln1pz_x1 = nu / (np.e * alpha * u_eq(Om))
print("=== Transition sharpness ===")
print(f"(dg/dx)_(x=1)                = {slope_x1:.6f}")
print(f"(dg/dln(1+z))_(x=1)          = {slope_ln1pz_x1:.6f}")
print()

# asymptotic tests
z_test = np.array([0.0, 0.1, 0.21, 0.5, 1.0, 3.0, 10.0, 1100.0])
g_bg_vals = g_of_z(z_test, Om, alpha, nu)
g_add_vals = g_pert_additive(z_test, Om, alpha, nu, eps)
g_surv_vals = g_pert_survival_floor(z_test, Om, alpha, nu, eps)

print("=== Perturbation closure comparison ===")
print("   z        g_bg        g+eps      g_survival_floor")
for zi, gb, ga, gs in zip(z_test, g_bg_vals, g_add_vals, g_surv_vals):
    print(f"{zi:7.2f}   {gb:9.6f}   {ga:9.6f}   {gs:16.6f}")
print()

print("Note: g+eps exceeds 1 at high z; survival-floor closure remains bounded by 1.")
print()

# background and growth test
bg = solve_growth(H0, Om, alpha, nu, eps=None)
bg_eps = solve_growth(H0, Om, alpha, nu, eps=eps)

print("=== Growth-sector proxy (relative only; sigma8_0 not inserted) ===")
print(f"Minimal closure:       fD(z=0) = {bg['fs8_over_sigma8_0'][-1]:.6f}")
print(f"With survival floor:   fD(z=0) = {bg_eps['fs8_over_sigma8_0'][-1]:.6f}")
print()

# small table at selected redshifts
sample_z = [0.0, 0.5, 1.0]
print("=== Selected background diagnostics ===")
print("   z       g(z)      E(z)      q(z)      w_DE_eff(z)")
for zi in sample_z:
    print(
        f"{zi:6.2f}   "
        f"{g_of_z(zi, Om, alpha, nu):8.5f}   "
        f"{E_bg(zi, H0, Om, alpha, nu):8.5f}   "
        f"{q_of_z(zi, H0, Om, alpha, nu):8.5f}   "
        f"{w_de_eff(zi, Om, alpha, nu):12.6f}"
    )
```

```python
# ============================================================
# EXPANSION-HISTORY ACTIVATED MATTER COUPLING
# Renormalised equality-tied model diagnostics (self-contained)
# ============================================================

import numpy as np
from scipy.integrate import solve_ivp

# ------------------------------------------------------------
# 1. Core control variables
# ------------------------------------------------------------

def u_eq(Om):
    return (1.0 / 3.0) * np.log((1.0 - Om) / Om)

def x_of_z(z, Om, alpha):
    return np.log(1.0 + z) / (alpha * u_eq(Om))

def M_of_z(z, Om, alpha, nu):
    x = x_of_z(z, Om, alpha)
    return np.where(x > 0.0, x**nu, 0.0)

def S_of_z(z, Om, alpha, nu):
    return np.exp(-M_of_z(z, Om, alpha, nu))

def g_of_z(z, Om, alpha, nu):
    return 1.0 - S_of_z(z, Om, alpha, nu)

def z_s_exact(Om, alpha):
    return np.exp(alpha * u_eq(Om)) - 1.0

def z_at_activation_fraction(p, Om, alpha, nu):
    if not (0.0 < p < 1.0):
        raise ValueError("p must lie in (0,1)")
    x = (-np.log(1.0 - p))**(1.0 / nu)
    return np.exp(alpha * u_eq(Om) * x) - 1.0

# ------------------------------------------------------------
# 2. Renormalised background
#    E^2(z) = 1 + Om[(1+z)^3 g(z) - 1]
# ------------------------------------------------------------

def E2_bg(z, H0, Om, alpha, nu):
    g = g_of_z(z, Om, alpha, nu)
    return 1.0 + Om * ((1.0 + z)**3 * g - 1.0)

def E_bg(z, H0, Om, alpha, nu):
    return np.sqrt(E2_bg(z, H0, Om, alpha, nu))

def dg_dz(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    x = x_of_z(z, Om, alpha)
    out = np.zeros_like(z)

    mask = x > 0.0
    if np.any(mask):
        xm = x[mask]
        gm = g_of_z(z[mask], Om, alpha, nu)
        out[mask] = (
            nu * xm**(nu - 1.0) * (1.0 - gm)
            / ((1.0 + z[mask]) * alpha * u_eq(Om))
        )
    return out

def dE2_dz(z, H0, Om, alpha, nu):
    g = g_of_z(z, Om, alpha, nu)
    dg = dg_dz(z, Om, alpha, nu)
    return Om * (3.0 * (1.0 + z)**2 * g + (1.0 + z)**3 * dg)

def q_of_z(z, H0, Om, alpha, nu):
    E2 = E2_bg(z, H0, Om, alpha, nu)
    dE2 = dE2_dz(z, H0, Om, alpha, nu)
    return -1.0 + (1.0 + z) * dE2 / (2.0 * E2)

# effective DE defined by E^2 = Om(1+z)^3 + rho_DE_eff
def rho_de_eff(z, Om, alpha, nu):
    g = g_of_z(z, Om, alpha, nu)
    return 1.0 - Om + Om * (1.0 + z)**3 * (g - 1.0)

def drho_de_eff_dz(z, Om, alpha, nu):
    g = g_of_z(z, Om, alpha, nu)
    dg = dg_dz(z, Om, alpha, nu)
    return Om * (3.0 * (1.0 + z)**2 * (g - 1.0) + (1.0 + z)**3 * dg)

def w_de_eff(z, Om, alpha, nu):
    rho = rho_de_eff(z, Om, alpha, nu)
    drho = drho_de_eff_dz(z, Om, alpha, nu)
    return -1.0 + (1.0 + z) * drho / (3.0 * rho)

def Omega_m_bg(z, Om, alpha, nu):
    return Om * (1.0 + z)**3 * g_of_z(z, Om, alpha, nu) / E2_bg(z, 1.0, Om, alpha, nu)

# ------------------------------------------------------------
# 3. Perturbation-sector closure
#    bounded survival-floor version
# ------------------------------------------------------------

def g_pert_survival_floor(z, Om, alpha, nu, eps):
    return 1.0 - (1.0 - eps) * S_of_z(z, Om, alpha, nu)

def Omega_m_pert_survival_floor(z, Om, alpha, nu, eps):
    return Om * (1.0 + z)**3 * g_pert_survival_floor(z, Om, alpha, nu, eps) / E2_bg(z, 1.0, Om, alpha, nu)

def Omega_m_pert_minimal(z, Om, alpha, nu):
    return Om * (1.0 + z)**3 * g_of_z(z, Om, alpha, nu) / E2_bg(z, 1.0, Om, alpha, nu)

# ------------------------------------------------------------
# 4. Growth solver in N = ln a
# ------------------------------------------------------------

def z_of_a(a):
    return 1.0 / a - 1.0

def dlnH_dN(a, H0, Om, alpha, nu):
    z = z_of_a(a)
    E2 = E2_bg(z, H0, Om, alpha, nu)
    dE2dz = dE2_dz(z, H0, Om, alpha, nu)
    dzdN = -(1.0 + z)
    dE2dN = dE2dz * dzdN
    return 0.5 * dE2dN / E2

def growth_rhs_minimal(N, y, H0, Om, alpha, nu):
    D, Dn = y
    a = np.exp(N)
    src = Omega_m_pert_minimal(z_of_a(a), Om, alpha, nu)
    friction = 2.0 + dlnH_dN(a, H0, Om, alpha, nu)
    Dnn = -(friction * Dn) + 1.5 * src * D
    return [Dn, Dnn]

def growth_rhs_survival_floor(N, y, H0, Om, alpha, nu, eps):
    D, Dn = y
    a = np.exp(N)
    src = Omega_m_pert_survival_floor(z_of_a(a), Om, alpha, nu, eps)
    friction = 2.0 + dlnH_dN(a, H0, Om, alpha, nu)
    Dnn = -(friction * Dn) + 1.5 * src * D
    return [Dn, Dnn]

def solve_growth(H0, Om, alpha, nu, eps=None, a_ini=1e-3, a_fin=1.0, n_eval=500):
    N_ini = np.log(a_ini)
    N_fin = np.log(a_fin)
    N_eval = np.linspace(N_ini, N_fin, n_eval)

    y0 = [a_ini, a_ini]

    if eps is None:
        sol = solve_ivp(
            growth_rhs_minimal, [N_ini, N_fin], y0,
            t_eval=N_eval, args=(H0, Om, alpha, nu),
            rtol=1e-8, atol=1e-10
        )
    else:
        sol = solve_ivp(
            growth_rhs_survival_floor, [N_ini, N_fin], y0,
            t_eval=N_eval, args=(H0, Om, alpha, nu, eps),
            rtol=1e-8, atol=1e-10
        )

    N = sol.t
    a = np.exp(N)
    z = z_of_a(a)
    D_raw = sol.y[0].copy()
    Dn_raw = sol.y[1].copy()

    D = D_raw / D_raw[-1]
    Dn = Dn_raw / D_raw[-1]
    f = Dn / D

    return {
        "N": N,
        "a": a,
        "z": z,
        "D": D,
        "f": f,
        "fs8_over_sigma8_0": f * D
    }

# ------------------------------------------------------------
# 5. Fiducial parameters
# ------------------------------------------------------------

H0 = 74.0
Om = 0.231
alpha = 0.483
nu = 0.886
eps = 0.08

# ------------------------------------------------------------
# 6. Diagnostics
# ------------------------------------------------------------

print("=== Renormalised equality-tied activation diagnostics ===")
print(f"Omega_m       = {Om:.6f}")
print(f"alpha         = {alpha:.6f}")
print(f"nu            = {nu:.6f}")
print(f"u_eq          = {u_eq(Om):.6f}")
print(f"u_s           = {alpha * u_eq(Om):.6f}")
print(f"z_s (x=1)     = {z_s_exact(Om, alpha):.6f}")
print(f"z_50 (g=0.5)  = {z_at_activation_fraction(0.5, Om, alpha, nu):.6f}")
print(f"z_90 (g=0.9)  = {z_at_activation_fraction(0.9, Om, alpha, nu):.6f}")
print()

print("=== Normalisation check ===")
print(f"E(0)          = {E_bg(0.0, H0, Om, alpha, nu):.6f}")
print(f"E^2(0)        = {E2_bg(0.0, H0, Om, alpha, nu):.6f}")
print()

print("=== Perturbation closure (bounded) ===")
for zi in [0.0, 0.1, 0.21, 0.5, 1.0, 3.0, 10.0]:
    gb = g_of_z(zi, Om, alpha, nu)
    gp = g_pert_survival_floor(zi, Om, alpha, nu, eps)
    print(f"z={zi:5.2f}   g_bg={gb:9.6f}   g_pert={gp:9.6f}")
print()

bg = solve_growth(H0, Om, alpha, nu, eps=None)
bg_eps = solve_growth(H0, Om, alpha, nu, eps=eps)

print("=== Growth-sector proxy (relative only) ===")
print(f"Minimal closure:       fD(z=0) = {bg['fs8_over_sigma8_0'][-1]:.6f}")
print(f"With survival floor:   fD(z=0) = {bg_eps['fs8_over_sigma8_0'][-1]:.6f}")
print()

print("=== Selected background diagnostics ===")
print("   z       g(z)      E(z)      q(z)      w_DE_eff(z)")
for zi in [0.0, 0.5, 1.0]:
    print(
        f"{zi:6.2f}   "
        f"{g_of_z(zi, Om, alpha, nu):8.5f}   "
        f"{E_bg(zi, H0, Om, alpha, nu):8.5f}   "
        f"{q_of_z(zi, H0, Om, alpha, nu):8.5f}   "
        f"{w_de_eff(zi, Om, alpha, nu):12.6f}"
    )
```

```python
# ============================================================
# EXPANSION-HISTORY ACTIVATED MATTER COUPLING
# Corrected self-normalised equality-tied model diagnostics
# ============================================================

import numpy as np
from scipy.integrate import solve_ivp

# ------------------------------------------------------------
# 1. Canonical control variables
# ------------------------------------------------------------

def u_eq(Om):
    return (1.0 / 3.0) * np.log((1.0 - Om) / Om)

def u_s(Om, alpha):
    return alpha * u_eq(Om)

def x_of_z(z, Om, alpha):
    return np.log(1.0 + z) / u_s(Om, alpha)

def M_of_z(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    x = x_of_z(z, Om, alpha)
    out = np.zeros_like(x)
    mask = x > 0.0
    out[mask] = x[mask] ** nu
    return out

def S_of_z(z, Om, alpha, nu):
    return np.exp(-M_of_z(z, Om, alpha, nu))

def g_of_z(z, Om, alpha, nu):
    return 1.0 - S_of_z(z, Om, alpha, nu)

def z_s_exact(Om, alpha):
    # x(z_s)=1
    return np.exp(u_s(Om, alpha)) - 1.0

def z_at_activation_fraction(p, Om, alpha, nu):
    if not (0.0 < p < 1.0):
        raise ValueError("p must lie in (0,1)")
    x = (-np.log(1.0 - p)) ** (1.0 / nu)
    return np.exp(u_s(Om, alpha) * x) - 1.0

# ------------------------------------------------------------
# 2. Corrected self-normalised background
#
# E^2(z) = 1 + Om * g(z) * [ (1+z)^3 - 1 ]
#
# This satisfies:
#   E(0)=1
#   E^2 -> Om(1+z)^3 + (1-Om) as g->1
# ------------------------------------------------------------

def E2_bg(z, H0, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    g = g_of_z(z, Om, alpha, nu)
    return 1.0 + Om * g * ((1.0 + z)**3 - 1.0)

def E_bg(z, H0, Om, alpha, nu):
    return np.sqrt(E2_bg(z, H0, Om, alpha, nu))

def dg_dz_safe(z, Om, alpha, nu):
    """
    Safe dg/dz away from z=0.
    Note: for nu<1, dg/dz diverges as z->0, so do not use this
    directly at z=0 for physics. Use dE2_dz_safe instead.
    """
    z = np.asarray(z, dtype=float)
    x = x_of_z(z, Om, alpha)
    g = g_of_z(z, Om, alpha, nu)
    out = np.zeros_like(z)

    mask = z > 0.0
    if np.any(mask):
        xm = x[mask]
        gm = g[mask]
        out[mask] = (
            nu * xm**(nu - 1.0) * (1.0 - gm)
            / ((1.0 + z[mask]) * u_s(Om, alpha))
        )
    return out

def dE2_dz_safe(z, H0, Om, alpha, nu):
    """
    Stable derivative of E^2.
    At z=0, the exact limit is 0 because E^2 = 1 + O(z^(nu+1)).
    """
    z = np.asarray(z, dtype=float)
    g = g_of_z(z, Om, alpha, nu)
    dg = dg_dz_safe(z, Om, alpha, nu)

    out = Om * (
        dg * ((1.0 + z)**3 - 1.0)
        + g * 3.0 * (1.0 + z)**2
    )

    # exact limit at z=0
    out = np.where(z == 0.0, 0.0, out)
    return out

def q_of_z(z, H0, Om, alpha, nu):
    E2 = E2_bg(z, H0, Om, alpha, nu)
    dE2 = dE2_dz_safe(z, H0, Om, alpha, nu)
    return -1.0 + (1.0 + z) * dE2 / (2.0 * E2)

# ------------------------------------------------------------
# 3. Effective dark-energy interpretation
#
# E^2 = Om(1+z)^3 + rho_DE_eff(z)
# ------------------------------------------------------------

def rho_de_eff(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    return E2_bg(z, 1.0, Om, alpha, nu) - Om * (1.0 + z)**3

def drho_de_eff_dz(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    return dE2_dz_safe(z, 1.0, Om, alpha, nu) - 3.0 * Om * (1.0 + z)**2

def w_de_eff(z, Om, alpha, nu):
    rho = rho_de_eff(z, Om, alpha, nu)
    drho = drho_de_eff_dz(z, Om, alpha, nu)
    return -1.0 + (1.0 + z) * drho / (3.0 * rho)

# ------------------------------------------------------------
# 4. Perturbation-sector closure
#
# Preferred bounded version:
# g_pert = 1 - (1-eps) * exp(-x^nu)
#        = eps + (1-eps) g
# ------------------------------------------------------------

def g_pert_survival_floor(z, Om, alpha, nu, eps):
    return 1.0 - (1.0 - eps) * S_of_z(z, Om, alpha, nu)

def Omega_m_pert_minimal(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    return Om * (1.0 + z)**3 * g_of_z(z, Om, alpha, nu) / E2_bg(z, 1.0, Om, alpha, nu)

def Omega_m_pert_survival_floor(z, Om, alpha, nu, eps):
    z = np.asarray(z, dtype=float)
    return Om * (1.0 + z)**3 * g_pert_survival_floor(z, Om, alpha, nu, eps) / E2_bg(z, 1.0, Om, alpha, nu)

# ------------------------------------------------------------
# 5. Growth solver in N = ln(a)
# ------------------------------------------------------------

def z_of_a(a):
    return 1.0 / a - 1.0

def dlnH_dN(a, H0, Om, alpha, nu):
    z = z_of_a(a)
    E2 = E2_bg(z, H0, Om, alpha, nu)
    dE2dz = dE2_dz_safe(z, H0, Om, alpha, nu)
    dzdN = -(1.0 + z)
    dE2dN = dE2dz * dzdN
    return 0.5 * dE2dN / E2

def growth_rhs_minimal(N, y, H0, Om, alpha, nu):
    D, Dn = y
    a = np.exp(N)
    src = Omega_m_pert_minimal(z_of_a(a), Om, alpha, nu)
    friction = 2.0 + dlnH_dN(a, H0, Om, alpha, nu)
    Dnn = -(friction * Dn) + 1.5 * src * D
    return [Dn, Dnn]

def growth_rhs_survival_floor(N, y, H0, Om, alpha, nu, eps):
    D, Dn = y
    a = np.exp(N)
    src = Omega_m_pert_survival_floor(z_of_a(a), Om, alpha, nu, eps)
    friction = 2.0 + dlnH_dN(a, H0, Om, alpha, nu)
    Dnn = -(friction * Dn) + 1.5 * src * D
    return [Dn, Dnn]

def solve_growth(H0, Om, alpha, nu, eps=None, a_ini=1e-3, a_fin=1.0, n_eval=500):
    N_ini = np.log(a_ini)
    N_fin = np.log(a_fin)
    N_eval = np.linspace(N_ini, N_fin, n_eval)

    # growing-mode IC: D ~ a in matter era
    y0 = [a_ini, a_ini]

    if eps is None:
        sol = solve_ivp(
            growth_rhs_minimal, [N_ini, N_fin], y0,
            t_eval=N_eval, args=(H0, Om, alpha, nu),
            rtol=1e-8, atol=1e-10
        )
    else:
        sol = solve_ivp(
            growth_rhs_survival_floor, [N_ini, N_fin], y0,
            t_eval=N_eval, args=(H0, Om, alpha, nu, eps),
            rtol=1e-8, atol=1e-10
        )

    N = sol.t
    a = np.exp(N)
    z = z_of_a(a)

    D_raw = sol.y[0].copy()
    Dn_raw = sol.y[1].copy()

    D = D_raw / D_raw[-1]
    Dn = Dn_raw / D_raw[-1]
    f = Dn / D

    return {
        "N": N,
        "a": a,
        "z": z,
        "D": D,
        "f": f,
        "fs8_over_sigma8_0": f * D
    }

# ------------------------------------------------------------
# 6. Fiducial parameters
# ------------------------------------------------------------

H0 = 74.0
Om = 0.231
alpha = 0.483
nu = 0.886
eps = 0.08

# ------------------------------------------------------------
# 7. Diagnostics
# ------------------------------------------------------------

print("=== Corrected self-normalised equality-tied diagnostics ===")
print(f"Omega_m       = {Om:.6f}")
print(f"alpha         = {alpha:.6f}")
print(f"nu            = {nu:.6f}")
print(f"u_eq          = {u_eq(Om):.6f}")
print(f"u_s           = {u_s(Om, alpha):.6f}")
print(f"z_s (x=1)     = {z_s_exact(Om, alpha):.6f}")
print(f"z_50 (g=0.5)  = {z_at_activation_fraction(0.5, Om, alpha, nu):.6f}")
print(f"z_90 (g=0.9)  = {z_at_activation_fraction(0.9, Om, alpha, nu):.6f}")
print()

print("=== Normalisation check ===")
print(f"E(0)          = {E_bg(0.0, H0, Om, alpha, nu):.6f}")
print(f"E^2(0)        = {E2_bg(0.0, H0, Om, alpha, nu):.6f}")
print(f"rho_DE_eff(0) = {rho_de_eff(0.0, Om, alpha, nu):.6f}")
print()

print("=== Perturbation closure (bounded) ===")
for zi in [0.0, 0.1, 0.21, 0.5, 1.0, 3.0, 10.0]:
    gb = g_of_z(zi, Om, alpha, nu)
    gp = g_pert_survival_floor(zi, Om, alpha, nu, eps)
    print(f"z={zi:5.2f}   g_bg={gb:9.6f}   g_pert={gp:9.6f}")
print()

bg = solve_growth(H0, Om, alpha, nu, eps=None)
bg_eps = solve_growth(H0, Om, alpha, nu, eps=eps)

print("=== Growth-sector proxy (relative only) ===")
print(f"Minimal closure:       fD(z=0) = {bg['fs8_over_sigma8_0'][-1]:.6f}")
print(f"With survival floor:   fD(z=0) = {bg_eps['fs8_over_sigma8_0'][-1]:.6f}")
print()

print("=== Selected background diagnostics ===")
print("   z       g(z)      E(z)      q(z)      w_DE_eff(z)")
for zi in [0.0, 0.5, 1.0]:
    print(
        f"{zi:6.2f}   "
        f"{g_of_z(zi, Om, alpha, nu):8.5f}   "
        f"{E_bg(zi, H0, Om, alpha, nu):8.5f}   "
        f"{q_of_z(zi, H0, Om, alpha, nu):8.5f}   "
        f"{w_de_eff(zi, Om, alpha, nu):12.6f}"
    )
```

```python
# ============================================================
# EXPANSION-HISTORY ACTIVATED MATTER COUPLING
# Structural kinematics scan for the corrected self-normalised model
# Self-contained
# ============================================================

import numpy as np
from scipy.integrate import solve_ivp
from scipy.optimize import brentq

# ------------------------------------------------------------
# 1. Canonical control variables
# ------------------------------------------------------------

def u_eq(Om):
    return (1.0 / 3.0) * np.log((1.0 - Om) / Om)

def u_s(Om, alpha):
    return alpha * u_eq(Om)

def x_of_z(z, Om, alpha):
    return np.log(1.0 + z) / u_s(Om, alpha)

def M_of_z(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    x = x_of_z(z, Om, alpha)
    out = np.zeros_like(x)
    mask = x > 0.0
    out[mask] = x[mask] ** nu
    return out

def S_of_z(z, Om, alpha, nu):
    return np.exp(-M_of_z(z, Om, alpha, nu))

def g_of_z(z, Om, alpha, nu):
    return 1.0 - S_of_z(z, Om, alpha, nu)

def z_s_exact(Om, alpha):
    return np.exp(u_s(Om, alpha)) - 1.0

def z_at_activation_fraction(p, Om, alpha, nu):
    if not (0.0 < p < 1.0):
        raise ValueError("p must lie in (0,1)")
    x = (-np.log(1.0 - p)) ** (1.0 / nu)
    return np.exp(u_s(Om, alpha) * x) - 1.0

# ------------------------------------------------------------
# 2. Corrected self-normalised background
# ------------------------------------------------------------

def E2_bg(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    g = g_of_z(z, Om, alpha, nu)
    return 1.0 + Om * g * ((1.0 + z)**3 - 1.0)

def E_bg(z, Om, alpha, nu):
    return np.sqrt(E2_bg(z, Om, alpha, nu))

def dg_dz_safe(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    x = x_of_z(z, Om, alpha)
    g = g_of_z(z, Om, alpha, nu)
    out = np.zeros_like(z)

    mask = z > 0.0
    if np.any(mask):
        xm = x[mask]
        gm = g[mask]
        out[mask] = (
            nu * xm**(nu - 1.0) * (1.0 - gm)
            / ((1.0 + z[mask]) * u_s(Om, alpha))
        )
    return out

def dE2_dz_safe(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    g = g_of_z(z, Om, alpha, nu)
    dg = dg_dz_safe(z, Om, alpha, nu)

    out = Om * (
        dg * ((1.0 + z)**3 - 1.0)
        + g * 3.0 * (1.0 + z)**2
    )
    out = np.where(z == 0.0, 0.0, out)
    return out

def q_of_z(z, Om, alpha, nu):
    E2 = E2_bg(z, Om, alpha, nu)
    dE2 = dE2_dz_safe(z, Om, alpha, nu)
    return -1.0 + (1.0 + z) * dE2 / (2.0 * E2)

def rho_de_eff(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    return E2_bg(z, Om, alpha, nu) - Om * (1.0 + z)**3

def drho_de_eff_dz(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    return dE2_dz_safe(z, Om, alpha, nu) - 3.0 * Om * (1.0 + z)**2

def w_de_eff(z, Om, alpha, nu):
    rho = rho_de_eff(z, Om, alpha, nu)
    drho = drho_de_eff_dz(z, Om, alpha, nu)
    return -1.0 + (1.0 + z) * drho / (3.0 * rho)

# ------------------------------------------------------------
# 3. Perturbation-sector closure
# ------------------------------------------------------------

def g_pert_survival_floor(z, Om, alpha, nu, eps):
    return 1.0 - (1.0 - eps) * S_of_z(z, Om, alpha, nu)

# ------------------------------------------------------------
# 4. Growth solver
# ------------------------------------------------------------

def z_of_a(a):
    return 1.0 / a - 1.0

def Omega_m_pert_minimal(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    return Om * (1.0 + z)**3 * g_of_z(z, Om, alpha, nu) / E2_bg(z, Om, alpha, nu)

def Omega_m_pert_survival_floor(z, Om, alpha, nu, eps):
    z = np.asarray(z, dtype=float)
    return Om * (1.0 + z)**3 * g_pert_survival_floor(z, Om, alpha, nu, eps) / E2_bg(z, Om, alpha, nu)

def dlnH_dN(a, Om, alpha, nu):
    z = z_of_a(a)
    E2 = E2_bg(z, Om, alpha, nu)
    dE2dz = dE2_dz_safe(z, Om, alpha, nu)
    dzdN = -(1.0 + z)
    dE2dN = dE2dz * dzdN
    return 0.5 * dE2dN / E2

def growth_rhs_minimal(N, y, Om, alpha, nu):
    D, Dn = y
    a = np.exp(N)
    src = Omega_m_pert_minimal(z_of_a(a), Om, alpha, nu)
    friction = 2.0 + dlnH_dN(a, Om, alpha, nu)
    Dnn = -(friction * Dn) + 1.5 * src * D
    return [Dn, Dnn]

def growth_rhs_survival_floor(N, y, Om, alpha, nu, eps):
    D, Dn = y
    a = np.exp(N)
    src = Omega_m_pert_survival_floor(z_of_a(a), Om, alpha, nu, eps)
    friction = 2.0 + dlnH_dN(a, Om, alpha, nu)
    Dnn = -(friction * Dn) + 1.5 * src * D
    return [Dn, Dnn]

def solve_growth(Om, alpha, nu, eps=None, a_ini=1e-3, a_fin=1.0, n_eval=600):
    N_ini = np.log(a_ini)
    N_fin = np.log(a_fin)
    N_eval = np.linspace(N_ini, N_fin, n_eval)
    y0 = [a_ini, a_ini]

    if eps is None:
        sol = solve_ivp(
            growth_rhs_minimal, [N_ini, N_fin], y0,
            t_eval=N_eval, args=(Om, alpha, nu),
            rtol=1e-8, atol=1e-10
        )
    else:
        sol = solve_ivp(
            growth_rhs_survival_floor, [N_ini, N_fin], y0,
            t_eval=N_eval, args=(Om, alpha, nu, eps),
            rtol=1e-8, atol=1e-10
        )

    N = sol.t
    a = np.exp(N)
    z = z_of_a(a)

    D_raw = sol.y[0].copy()
    Dn_raw = sol.y[1].copy()

    D = D_raw / D_raw[-1]
    Dn = Dn_raw / D_raw[-1]
    f = Dn / D

    return {"z": z, "D": D, "f": f, "fD": f * D}

# ------------------------------------------------------------
# 5. Structural markers
# ------------------------------------------------------------

def find_root_scan(func, zmin, zmax, n=2000):
    zz = np.linspace(zmin, zmax, n)
    yy = np.array([func(z) for z in zz], dtype=float)
    for i in range(len(zz) - 1):
        y0, y1 = yy[i], yy[i + 1]
        if np.isfinite(y0) and np.isfinite(y1) and y0 == 0.0:
            return zz[i]
        if np.isfinite(y0) and np.isfinite(y1) and y0 * y1 < 0.0:
            return brentq(func, zz[i], zz[i + 1])
    return None

def finite_slope_g0(Om, alpha, nu, dz=1e-6):
    # safe finite-difference slope of g near z=0
    return (g_of_z(dz, Om, alpha, nu) - g_of_z(0.0, Om, alpha, nu)) / dz

# ------------------------------------------------------------
# 6. Fiducial parameters
# ------------------------------------------------------------

Om = 0.231
alpha = 0.483
nu = 0.886
eps = 0.08

# ------------------------------------------------------------
# 7. Derived markers
# ------------------------------------------------------------

z_s = z_s_exact(Om, alpha)
z_50 = z_at_activation_fraction(0.5, Om, alpha, nu)
z_90 = z_at_activation_fraction(0.9, Om, alpha, nu)

z_acc = find_root_scan(lambda z: q_of_z(z, Om, alpha, nu), 0.0, 3.0)
z_wm1 = find_root_scan(lambda z: w_de_eff(z, Om, alpha, nu) + 1.0, 0.0, 3.0)

print("=== Derived structural markers ===")
print(f"z_s    (x=1)      = {z_s:.6f}")
print(f"z_50   (g=0.5)    = {z_50:.6f}")
print(f"z_90   (g=0.9)    = {z_90:.6f}")
print(f"z_acc  (q=0)      = {z_acc if z_acc is not None else 'None'}")
print(f"z_w=-1 crossing   = {z_wm1 if z_wm1 is not None else 'None'}")
print()

print("=== Transition ordering ===")
markers = [("z_50", z_50), ("z_s", z_s), ("z_acc", z_acc), ("z_90", z_90)]
markers = [(k, v) for k, v in markers if v is not None]
markers.sort(key=lambda kv: kv[1])
for k, v in markers:
    print(f"{k:8s} = {v:.6f}")
print()

print("=== Activation finite-difference near z=0 ===")
for dz in [1e-2, 1e-4, 1e-6, 1e-8]:
    print(f"dz={dz:10.1e}   [g(dz)-g(0)]/dz = {finite_slope_g0(Om, alpha, nu, dz):.6f}")
print()

print("=== Background diagnostics table ===")
print("   z        g(z)      E(z)      q(z)      w_DE_eff(z)")
for zi in [0.0, 0.1, 0.2, 0.5, 1.0, 2.0]:
    print(
        f"{zi:6.2f}   "
        f"{g_of_z(zi, Om, alpha, nu):8.5f}   "
        f"{E_bg(zi, Om, alpha, nu):8.5f}   "
        f"{q_of_z(zi, Om, alpha, nu):8.5f}   "
        f"{w_de_eff(zi, Om, alpha, nu):12.6f}"
    )
print()

# Growth comparison at a few redshifts
gmin = solve_growth(Om, alpha, nu, eps=None)
gflr = solve_growth(Om, alpha, nu, eps=eps)

# helper interpolation onto requested z points
def interp_desc(z_target, z_arr, y_arr):
    order = np.argsort(z_arr)
    return np.interp(z_target, z_arr[order], y_arr[order])

print("=== Growth comparison (relative proxy fD = f * D) ===")
print("   z      minimal      floor      delta")
for zi in [0.0, 0.3, 0.5, 1.0, 2.0]:
    y0 = interp_desc(zi, gmin["z"], gmin["fD"])
    y1 = interp_desc(zi, gflr["z"], gflr["fD"])
    print(f"{zi:6.2f}   {y0:9.6f}   {y1:9.6f}   {y1-y0:+9.6f}")
```

```python
# ============================================================
# Compare loading choices:
#   1) log loading: u = ln(1+z)/u_s
#   2) linear loading: u = z/u_s
#
# Same corrected self-normalised background:
#   E^2 = 1 + Om * g(z) * [ (1+z)^3 - 1 ]
# ============================================================

import numpy as np
from scipy.integrate import solve_ivp
from scipy.optimize import brentq

# ------------------------------------------------------------
# 1. Shared utilities
# ------------------------------------------------------------

def u_eq(Om):
    return (1.0 / 3.0) * np.log((1.0 - Om) / Om)

def u_s(Om, alpha):
    return alpha * u_eq(Om)

def find_root_scan(func, zmin, zmax, n=2000):
    zz = np.linspace(zmin, zmax, n)
    yy = np.array([func(z) for z in zz], dtype=float)
    for i in range(len(zz) - 1):
        y0, y1 = yy[i], yy[i + 1]
        if np.isfinite(y0) and np.isfinite(y1) and y0 == 0.0:
            return zz[i]
        if np.isfinite(y0) and np.isfinite(y1) and y0 * y1 < 0.0:
            return brentq(func, zz[i], zz[i + 1])
    return None

def finite_slope_g0(gfunc, Om, alpha, nu, dz=1e-6):
    return (gfunc(dz, Om, alpha, nu) - gfunc(0.0, Om, alpha, nu)) / dz

# ------------------------------------------------------------
# 2. Two loading models
# ------------------------------------------------------------

def x_log(z, Om, alpha):
    return np.log(1.0 + z) / u_s(Om, alpha)

def x_lin(z, Om, alpha):
    return z / u_s(Om, alpha)

def g_from_x(x, nu):
    x = np.asarray(x, dtype=float)
    out = np.zeros_like(x)
    mask = x > 0.0
    out[mask] = 1.0 - np.exp(-(x[mask] ** nu))
    return out

def g_log(z, Om, alpha, nu):
    return g_from_x(x_log(z, Om, alpha), nu)

def g_lin(z, Om, alpha, nu):
    return g_from_x(x_lin(z, Om, alpha), nu)

def dg_dz_log_safe(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    x = x_log(z, Om, alpha)
    g = g_log(z, Om, alpha, nu)
    out = np.zeros_like(z)
    mask = z > 0.0
    if np.any(mask):
        out[mask] = (
            nu * x[mask]**(nu - 1.0) * (1.0 - g[mask])
            / ((1.0 + z[mask]) * u_s(Om, alpha))
        )
    return out

def dg_dz_lin_safe(z, Om, alpha, nu):
    z = np.asarray(z, dtype=float)
    x = x_lin(z, Om, alpha)
    g = g_lin(z, Om, alpha, nu)
    out = np.zeros_like(z)
    mask = z > 0.0
    if np.any(mask):
        out[mask] = (
            nu * x[mask]**(nu - 1.0) * (1.0 - g[mask])
            / u_s(Om, alpha)
        )
    return out

# ------------------------------------------------------------
# 3. Background and diagnostics builder
# ------------------------------------------------------------

def make_model(gfunc, dgfunc, label):
    def E2_bg(z, Om, alpha, nu):
        z = np.asarray(z, dtype=float)
        g = gfunc(z, Om, alpha, nu)
        return 1.0 + Om * g * ((1.0 + z)**3 - 1.0)

    def E_bg(z, Om, alpha, nu):
        return np.sqrt(E2_bg(z, Om, alpha, nu))

    def dE2_dz_safe(z, Om, alpha, nu):
        z = np.asarray(z, dtype=float)
        g = gfunc(z, Om, alpha, nu)
        dg = dgfunc(z, Om, alpha, nu)
        out = Om * (dg * ((1.0 + z)**3 - 1.0) + g * 3.0 * (1.0 + z)**2)
        out = np.where(z == 0.0, 0.0, out)
        return out

    def q_of_z(z, Om, alpha, nu):
        E2 = E2_bg(z, Om, alpha, nu)
        dE2 = dE2_dz_safe(z, Om, alpha, nu)
        return -1.0 + (1.0 + z) * dE2 / (2.0 * E2)

    def rho_de_eff(z, Om, alpha, nu):
        z = np.asarray(z, dtype=float)
        return E2_bg(z, Om, alpha, nu) - Om * (1.0 + z)**3

    def drho_de_eff_dz(z, Om, alpha, nu):
        z = np.asarray(z, dtype=float)
        return dE2_dz_safe(z, Om, alpha, nu) - 3.0 * Om * (1.0 + z)**2

    def w_de_eff(z, Om, alpha, nu):
        rho = rho_de_eff(z, Om, alpha, nu)
        drho = drho_de_eff_dz(z, Om, alpha, nu)
        return -1.0 + (1.0 + z) * drho / (3.0 * rho)

    def z_s_exact(Om, alpha):
        # x(z_s)=1
        if label == "log":
            return np.exp(u_s(Om, alpha)) - 1.0
        elif label == "linear":
            return u_s(Om, alpha)
        else:
            raise ValueError("unknown label")

    def z_at_activation_fraction(p, Om, alpha, nu):
        x = (-np.log(1.0 - p)) ** (1.0 / nu)
        if label == "log":
            return np.exp(u_s(Om, alpha) * x) - 1.0
        elif label == "linear":
            return u_s(Om, alpha) * x
        else:
            raise ValueError("unknown label")

    def z_of_a(a):
        return 1.0 / a - 1.0

    def Omega_m_pert(z, Om, alpha, nu, eps=None):
        z = np.asarray(z, dtype=float)
        g = gfunc(z, Om, alpha, nu)
        if eps is not None:
            g = eps + (1.0 - eps) * g
        return Om * (1.0 + z)**3 * g / E2_bg(z, Om, alpha, nu)

    def dlnH_dN(a, Om, alpha, nu):
        z = z_of_a(a)
        E2 = E2_bg(z, Om, alpha, nu)
        dE2dz = dE2_dz_safe(z, Om, alpha, nu)
        dE2dN = dE2dz * (-(1.0 + z))
        return 0.5 * dE2dN / E2

    def growth_rhs(N, y, Om, alpha, nu, eps):
        D, Dn = y
        a = np.exp(N)
        src = Omega_m_pert(z_of_a(a), Om, alpha, nu, eps)
        friction = 2.0 + dlnH_dN(a, Om, alpha, nu)
        Dnn = -(friction * Dn) + 1.5 * src * D
        return [Dn, Dnn]

    def solve_growth(Om, alpha, nu, eps=None, a_ini=1e-3, a_fin=1.0, n_eval=600):
        N_ini = np.log(a_ini)
        N_fin = np.log(a_fin)
        N_eval = np.linspace(N_ini, N_fin, n_eval)
        y0 = [a_ini, a_ini]
        sol = solve_ivp(
            growth_rhs, [N_ini, N_fin], y0,
            t_eval=N_eval, args=(Om, alpha, nu, eps),
            rtol=1e-8, atol=1e-10
        )
        N = sol.t
        a = np.exp(N)
        z = z_of_a(a)
        D_raw = sol.y[0].copy()
        Dn_raw = sol.y[1].copy()
        D = D_raw / D_raw[-1]
        Dn = Dn_raw / D_raw[-1]
        f = Dn / D
        return {"z": z, "fD": f * D}

    return {
        "label": label,
        "g": gfunc,
        "E_bg": E_bg,
        "q": q_of_z,
        "w": w_de_eff,
        "z_s": z_s_exact,
        "z_p": z_at_activation_fraction,
        "solve_growth": solve_growth,
    }

# ------------------------------------------------------------
# 4. Build models
# ------------------------------------------------------------

model_log = make_model(g_log, dg_dz_log_safe, "log")
model_lin = make_model(g_lin, dg_dz_lin_safe, "linear")

# ------------------------------------------------------------
# 5. Fiducial parameters
# ------------------------------------------------------------

Om = 0.231
alpha = 0.483
nu = 0.886
eps = 0.08

# ------------------------------------------------------------
# 6. Report function
# ------------------------------------------------------------

def report_model(model):
    g = model["g"]
    E = model["E_bg"]
    q = model["q"]
    w = model["w"]
    z_s = model["z_s"](Om, alpha)
    z_50 = model["z_p"](0.5, Om, alpha, nu)
    z_90 = model["z_p"](0.9, Om, alpha, nu)
    z_acc = find_root_scan(lambda z: q(z, Om, alpha, nu), 0.0, 3.0)
    z_wm1 = find_root_scan(lambda z: w(z, Om, alpha, nu) + 1.0, 0.0, 3.0)

    print(f"=== {model['label'].upper()} loading ===")
    print(f"z_s    = {z_s:.6f}")
    print(f"z_50   = {z_50:.6f}")
    print(f"z_90   = {z_90:.6f}")
    print(f"z_acc  = {z_acc if z_acc is not None else 'None'}")
    print(f"z_w=-1 = {z_wm1 if z_wm1 is not None else 'None'}")
    print()

    print("finite-difference [g(dz)-g(0)]/dz")
    for dz in [1e-2, 1e-4, 1e-6, 1e-8]:
        print(f"dz={dz:10.1e}   {finite_slope_g0(g, Om, alpha, nu, dz):.6f}")
    print()

    print("   z        g(z)      E(z)      q(z)      w_DE_eff(z)")
    for zi in [0.0, 0.1, 0.2, 0.5, 1.0, 2.0]:
        print(
            f"{zi:6.2f}   "
            f"{g(zi, Om, alpha, nu):8.5f}   "
            f"{E(zi, Om, alpha, nu):8.5f}   "
            f"{q(zi, Om, alpha, nu):8.5f}   "
            f"{w(zi, Om, alpha, nu):12.6f}"
        )
    print()

    gmin = model["solve_growth"](Om, alpha, nu, eps=None)
    gflr = model["solve_growth"](Om, alpha, nu, eps=eps)

    def interp_desc(z_target, z_arr, y_arr):
        order = np.argsort(z_arr)
        return np.interp(z_target, z_arr[order], y_arr[order])

    print("growth comparison fD")
    print("   z      minimal      floor      delta")
    for zi in [0.0, 0.3, 0.5, 1.0, 2.0]:
        y0 = interp_desc(zi, gmin["z"], gmin["fD"])
        y1 = interp_desc(zi, gflr["z"], gflr["fD"])
        print(f"{zi:6.2f}   {y0:9.6f}   {y1:9.6f}   {y1-y0:+9.6f}")
    print()

# ------------------------------------------------------------
# 7. Run both
# ------------------------------------------------------------

report_model(model_log)
report_model(model_lin)
```

```python
# ============================================================
# Compare pure vs regularised Weibull activation kernel
# Background: corrected self-normalised form
# Loading: equality-tied log loading
# Perturbations: bounded survival-floor closure
# ============================================================

import numpy as np
from scipy.integrate import solve_ivp
from scipy.optimize import brentq

# ------------------------------------------------------------
# 1. Shared cosmology utilities
# ------------------------------------------------------------

def u_eq(Om):
    return (1.0 / 3.0) * np.log((1.0 - Om) / Om)

def u_s(Om, alpha):
    return alpha * u_eq(Om)

def u_of_z(z, Om, alpha):
    return np.log(1.0 + z) / u_s(Om, alpha)

def find_root_scan(func, zmin, zmax, n=3000):
    zz = np.linspace(zmin, zmax, n)
    yy = np.array([func(z) for z in zz], dtype=float)
    for i in range(len(zz) - 1):
        y0, y1 = yy[i], yy[i + 1]
        if np.isfinite(y0) and np.isfinite(y1) and y0 == 0.0:
            return zz[i]
        if np.isfinite(y0) and np.isfinite(y1) and y0 * y1 < 0.0:
            return brentq(func, zz[i], zz[i + 1])
    return None

# ------------------------------------------------------------
# 2. Kernel definitions
# ------------------------------------------------------------

def M_pure(u, nu):
    u = np.asarray(u, dtype=float)
    out = np.zeros_like(u)
    mask = u > 0.0
    out[mask] = u[mask] ** nu
    return out

def M_reg(u, nu, u_c):
    u = np.asarray(u, dtype=float)
    return (u*u + u_c*u_c) ** (0.5 * nu) - (u_c ** nu)

def g_from_M(M):
    return 1.0 - np.exp(-M)

def g_pure_u(u, nu):
    return g_from_M(M_pure(u, nu))

def g_reg_u(u, nu, u_c):
    return g_from_M(M_reg(u, nu, u_c))

# ------------------------------------------------------------
# 3. Build model from chosen kernel
# ------------------------------------------------------------

def make_model(label, nu, u_c=None):
    def g_of_z(z, Om, alpha):
        u = u_of_z(np.asarray(z, dtype=float), Om, alpha)
        if label == "pure":
            return g_pure_u(u, nu)
        elif label == "reg":
            return g_reg_u(u, nu, u_c)
        else:
            raise ValueError("unknown label")

    def dM_du(u):
        u = np.asarray(u, dtype=float)
        out = np.zeros_like(u)
        if label == "pure":
            mask = u > 0.0
            out[mask] = nu * u[mask] ** (nu - 1.0)
        elif label == "reg":
            out = nu * u * (u*u + u_c*u_c) ** (0.5 * nu - 1.0)
        else:
            raise ValueError("unknown label")
        return out

    def dg_dz_safe(z, Om, alpha):
        z = np.asarray(z, dtype=float)
        u = u_of_z(z, Om, alpha)
        g = g_of_z(z, Om, alpha)
        du_dz = 1.0 / ((1.0 + z) * u_s(Om, alpha))
        dM = dM_du(u)
        out = (1.0 - g) * dM * du_dz
        # exact safe handling at z=0
        out = np.where(z == 0.0, 0.0 if label == "reg" else out, out)
        return out

    def E2_bg(z, Om, alpha):
        z = np.asarray(z, dtype=float)
        g = g_of_z(z, Om, alpha)
        return 1.0 + Om * g * ((1.0 + z)**3 - 1.0)

    def E_bg(z, Om, alpha):
        return np.sqrt(E2_bg(z, Om, alpha))

    def dE2_dz_safe(z, Om, alpha):
        z = np.asarray(z, dtype=float)
        g = g_of_z(z, Om, alpha)
        dg = dg_dz_safe(z, Om, alpha)
        out = Om * (dg * ((1.0 + z)**3 - 1.0) + g * 3.0 * (1.0 + z)**2)
        out = np.where(z == 0.0, 0.0, out)
        return out

    def q_of_z(z, Om, alpha):
        E2 = E2_bg(z, Om, alpha)
        dE2 = dE2_dz_safe(z, Om, alpha)
        return -1.0 + (1.0 + z) * dE2 / (2.0 * E2)

    def rho_de_eff(z, Om, alpha):
        z = np.asarray(z, dtype=float)
        return E2_bg(z, Om, alpha) - Om * (1.0 + z)**3

    def drho_de_eff_dz(z, Om, alpha):
        z = np.asarray(z, dtype=float)
        return dE2_dz_safe(z, Om, alpha) - 3.0 * Om * (1.0 + z)**2

    def w_de_eff(z, Om, alpha):
        rho = rho_de_eff(z, Om, alpha)
        drho = drho_de_eff_dz(z, Om, alpha)
        return -1.0 + (1.0 + z) * drho / (3.0 * rho)

    def z_at_x_eq_1(Om, alpha):
        return np.exp(u_s(Om, alpha)) - 1.0

    def z_at_activation_fraction(p, Om, alpha):
        if not (0.0 < p < 1.0):
            raise ValueError("p must lie in (0,1)")

        def f(z):
            return g_of_z(z, Om, alpha) - p

        return find_root_scan(f, 0.0, 5.0)

    def z_of_a(a):
        return 1.0 / a - 1.0

    def g_pert_of_z(z, Om, alpha, eps):
        g = g_of_z(z, Om, alpha)
        return eps + (1.0 - eps) * g

    def Omega_m_pert(z, Om, alpha, eps=None):
        z = np.asarray(z, dtype=float)
        g = g_of_z(z, Om, alpha) if eps is None else g_pert_of_z(z, Om, alpha, eps)
        return Om * (1.0 + z)**3 * g / E2_bg(z, Om, alpha)

    def dlnH_dN(a, Om, alpha):
        z = z_of_a(a)
        E2 = E2_bg(z, Om, alpha)
        dE2dz = dE2_dz_safe(z, Om, alpha)
        dE2dN = dE2dz * (-(1.0 + z))
        return 0.5 * dE2dN / E2

    def growth_rhs(N, y, Om, alpha, eps):
        D, Dn = y
        a = np.exp(N)
        src = Omega_m_pert(z_of_a(a), Om, alpha, eps)
        friction = 2.0 + dlnH_dN(a, Om, alpha)
        Dnn = -(friction * Dn) + 1.5 * src * D
        return [Dn, Dnn]

    def solve_growth(Om, alpha, eps=None, a_ini=1e-3, a_fin=1.0, n_eval=700):
        N_ini = np.log(a_ini)
        N_fin = np.log(a_fin)
        N_eval = np.linspace(N_ini, N_fin, n_eval)
        y0 = [a_ini, a_ini]
        sol = solve_ivp(
            growth_rhs, [N_ini, N_fin], y0,
            t_eval=N_eval, args=(Om, alpha, eps),
            rtol=1e-8, atol=1e-10
        )
        N = sol.t
        a = np.exp(N)
        z = z_of_a(a)
        D_raw = sol.y[0].copy()
        Dn_raw = sol.y[1].copy()
        D = D_raw / D_raw[-1]
        Dn = Dn_raw / D_raw[-1]
        f = Dn / D
        return {"z": z, "fD": f * D}

    return {
        "label": label if label == "pure" else f"reg(u_c={u_c})",
        "g": g_of_z,
        "E": E_bg,
        "q": q_of_z,
        "w": w_de_eff,
        "z_s": z_at_x_eq_1,
        "z_p": z_at_activation_fraction,
        "solve_growth": solve_growth,
    }

# ------------------------------------------------------------
# 4. Report block
# ------------------------------------------------------------

def finite_slope_g0(gfunc, Om, alpha, dz):
    return (gfunc(dz, Om, alpha) - gfunc(0.0, Om, alpha)) / dz

def report_model(model, Om, alpha, eps):
    g = model["g"]
    E = model["E"]
    q = model["q"]
    w = model["w"]

    z_s = model["z_s"](Om, alpha)
    z_50 = model["z_p"](0.5, Om, alpha)
    z_90 = model["z_p"](0.9, Om, alpha)
    z_acc = find_root_scan(lambda z: q(z, Om, alpha), 0.0, 3.0)
    z_wm1 = find_root_scan(lambda z: w(z, Om, alpha) + 1.0, 0.0, 3.0)

    print(f"=== {model['label']} ===")
    print(f"z_s    = {z_s:.6f}")
    print(f"z_50   = {z_50:.6f}")
    print(f"z_90   = {z_90:.6f}")
    print(f"z_acc  = {z_acc if z_acc is not None else 'None'}")
    print(f"z_w=-1 = {z_wm1 if z_wm1 is not None else 'None'}")
    print()

    print("finite-difference [g(dz)-g(0)]/dz")
    for dz in [1e-2, 1e-4, 1e-6, 1e-8]:
        print(f"dz={dz:10.1e}   {finite_slope_g0(g, Om, alpha, dz):.6f}")
    print()

    print("   z        g(z)      E(z)      q(z)      w_DE_eff(z)")
    for zi in [0.0, 0.1, 0.2, 0.5, 1.0, 2.0]:
        print(
            f"{zi:6.2f}   "
            f"{g(zi, Om, alpha):8.5f}   "
            f"{E(zi, Om, alpha):8.5f}   "
            f"{q(zi, Om, alpha):8.5f}   "
            f"{w(zi, Om, alpha):12.6f}"
        )
    print()

    gmin = model["solve_growth"](Om, alpha, eps=None)
    gflr = model["solve_growth"](Om, alpha, eps=eps)

    def interp_desc(z_target, z_arr, y_arr):
        order = np.argsort(z_arr)
        return np.interp(z_target, z_arr[order], y_arr[order])

    print("growth comparison fD")
    print("   z      minimal      floor      delta")
    for zi in [0.0, 0.3, 0.5, 1.0, 2.0]:
        y0 = interp_desc(zi, gmin["z"], gmin["fD"])
        y1 = interp_desc(zi, gflr["z"], gflr["fD"])
        print(f"{zi:6.2f}   {y0:9.6f}   {y1:9.6f}   {y1-y0:+9.6f}")
    print()

# ------------------------------------------------------------
# 5. Fiducial parameters
# ------------------------------------------------------------

Om = 0.231
alpha = 0.483
nu = 0.886
eps = 0.08

# ------------------------------------------------------------
# 6. Run pure and regularised cases
# ------------------------------------------------------------

models = [
    make_model("pure", nu),
    make_model("reg", nu, u_c=0.02),
    make_model("reg", nu, u_c=0.05),
    make_model("reg", nu, u_c=0.10),
]

for model in models:
    report_model(model, Om, alpha, eps)
```

```python
"""
t5000_weak_cluster_diagnostic.py
================================

Purpose
-------
Target the small weak-geometry cluster found at T=5000.

Previous result:
  strong_geometry     : 139 / 147
  borderline_geometry :   3 / 147
  weak_geometry       :   5 / 147

All weak cases were localized near T=5000:
  - n=20, shift=-3
  - n=20, shift=-2
  - n=20, shift=-1
  - n=20, shift=+3
  - n=16, shift=-3

This script asks whether those weak cases are:

  (A) pool-size artifacts
  (B) edge-window artifacts
  (C) basis-construction artifacts
  (D) local spacing-structure effects
  (E) genuine local softening of the reduced-flow geometry

Unified convention
------------------
Reduced repulsive flow:

    dt_i/ds = +2 * sum_{j!=i} 1/(t_i - t_j),   s >= 0

This is the same convention used in the previous scripts.

Main idea
---------
For each target window, compare two diagnostics:

  1. INTERNAL basis:
     Hadamard/density basis constructed only from the window itself.

  2. CONTEXT basis:
     Hadamard/density basis constructed from an expanded local context,
     then restricted back to the target window.

If weak geometry improves strongly under context basis, the weakness is
probably an edge/basis artifact. If not, it is more likely intrinsic to
that local zero configuration.
"""

import math
import mpmath
import numpy as np
from scipy.linalg import svd

mpmath.mp.dps = 25

# =========================================================
# User controls
# =========================================================

T_TARGET = 5000

# pool sizes to test
POOL_SIZES = [56, 80, 120]

# weak cases identified previously
TARGET_WINDOWS = [
    (16, -3),
    (20, -3),
    (20, -2),
    (20, -1),
    (20, +3),
]

# include neighbouring shifts for comparison
NEIGHBOUR_SHIFTS = [-4, -3, -2, -1, 0, 1, 2, 3, 4]

WINDOW_SIZES = [12, 16, 20, 24]

# context padding on each side for context-basis construction
CONTEXT_PAD = 8

S_MAX = 0.04
N_S = 100

# geometry thresholds
COS1_STRONG = 0.998
R2_STRONG   = 0.997
S1S2_STRONG = 10.0
S2S3_STRONG = 8.0

COS1_BORDER = 0.995
R2_BORDER   = 0.985
S1S2_BORDER = 8.0
S2S3_BORDER = 6.0


# =========================================================
# Zero finding
# =========================================================

def find_zeros_pool(T, n, span=None):
    if span is None:
        span = max(40.0, n * 2.0 * math.pi / math.log(T / (2.0 * math.pi)) * 1.7)

    ts = np.linspace(T - span, T + span, max(16000, int(span * 70)))
    zv = np.array([float(mpmath.siegelz(t)) for t in ts])

    roots = []
    for i in range(len(zv) - 1):
        if zv[i] * zv[i + 1] < 0:
            r = float(mpmath.findroot(mpmath.siegelz, (ts[i], ts[i+1]), solver="illinois"))
            roots.append(r)

    roots = sorted(roots)
    if len(roots) < n:
        raise RuntimeError(f"Only found {len(roots)} zeros near T={T}; needed {n}. Increase span.")

    mid = len(roots) // 2
    start = max(0, mid - n // 2)
    end = start + n
    if end > len(roots):
        end = len(roots)
        start = end - n

    return np.array(roots[start:end], dtype=float)


# =========================================================
# Reduced flow
# =========================================================

def rk4_repulsive(z, ds):
    def vel(x):
        n = len(x)
        v = np.zeros(n, dtype=float)
        for i in range(n):
            for j in range(n):
                if i != j:
                    d = x[i] - x[j]
                    if abs(d) > 1e-12:
                        v[i] += 2.0 / d
        return v

    k1 = vel(z)
    k2 = vel(z + 0.5 * ds * k1)
    k3 = vel(z + 0.5 * ds * k2)
    k4 = vel(z + ds * k3)
    return z + (ds / 6.0) * (k1 + 2*k2 + 2*k3 + k4)

def build_velocity_matrix(z0, s_vals):
    s_full = np.concatenate([[0.0], s_vals])
    hist = np.zeros((len(s_full), len(z0)), dtype=float)
    hist[0] = z0.copy()

    z = z0.copy()
    for k in range(1, len(s_full)):
        z = rk4_repulsive(z, s_full[k] - s_full[k - 1])
        hist[k] = z

    V = np.zeros((len(z0), len(s_vals)), dtype=float)
    for k, s in enumerate(s_vals):
        col = (hist[k+1] - z0) / s
        col -= col.mean()
        V[:, k] = col

    return V


# =========================================================
# Basis construction
# =========================================================

def hadamard_direction(z):
    n = len(z)
    vh = np.zeros(n, dtype=float)
    for i in range(n):
        for j in range(n):
            if i != j:
                vh[i] += 2.0 / (z[i] - z[j])
    vh -= vh.mean()
    nrm = np.linalg.norm(vh)
    if nrm < 1e-15:
        return np.zeros_like(vh)
    return vh / nrm

def density_direction(z, e1):
    n = len(z)
    g = np.diff(z)
    vd = np.zeros(n, dtype=float)

    for i in range(1, n - 1):
        di = 0.5 * (z[i+1] - z[i-1])
        if di > 1e-12:
            vd[i] = -(g[i] - g[i-1]) / (2.0 * di**2)

    if len(g) > 1:
        vd[0]  = -(g[1]  - g[0])  / (2.0 * g[0]**2)
        vd[-1] = -(g[-1] - g[-2]) / (2.0 * g[-1]**2)

    vd -= vd.mean()
    vd -= np.dot(vd, e1) * e1

    nrm = np.linalg.norm(vd)
    if nrm < 1e-15:
        return np.zeros_like(vd)
    return vd / nrm

def internal_basis(z0):
    e1 = hadamard_direction(z0)
    e2 = density_direction(z0, e1)
    return e1, e2

def context_basis(pool, start, end, pad):
    """
    Build e1/e2 on a larger context window, then restrict to target.
    """
    cstart = max(0, start - pad)
    cend = min(len(pool), end + pad)

    zc = pool[cstart:cend]
    e1c = hadamard_direction(zc)
    e2c = density_direction(zc, e1c)

    # restriction indices inside context
    rs = start - cstart
    re = end - cstart

    e1 = e1c[rs:re].copy()
    e2 = e2c[rs:re].copy()

    # re-centre and re-orthonormalise on target window
    e1 -= e1.mean()
    n1 = np.linalg.norm(e1)
    if n1 > 1e-15:
        e1 /= n1

    e2 -= e2.mean()
    e2 -= np.dot(e2, e1) * e1
    n2 = np.linalg.norm(e2)
    if n2 > 1e-15:
        e2 /= n2

    return e1, e2


# =========================================================
# Geometry diagnostics
# =========================================================

def spacing_stats(z0):
    g = np.diff(z0)
    D = g.mean()
    dd = np.diff(g)

    return {
        "D": float(D),
        "gminD": float(g.min() / D),
        "gmaxD": float(g.max() / D),
        "gapCV": float(g.std() / D),
        "rangeD": float((g.max() - g.min()) / D),
        "ddRMSD": float(np.sqrt(np.mean(dd**2)) / D) if len(dd) else 0.0,
        "left_gapD": float(g[0] / D),
        "right_gapD": float(g[-1] / D),
    }

def classify_geom(cos1, R2, S1S2, S2S3):
    strong = (
        cos1 >= COS1_STRONG and
        R2 >= R2_STRONG and
        S1S2 >= S1S2_STRONG and
        S2S3 >= S2S3_STRONG
    )

    borderline_ok = (
        cos1 >= COS1_BORDER and
        R2 >= R2_BORDER and
        S1S2 >= S1S2_BORDER and
        S2S3 >= S2S3_BORDER
    )

    if strong:
        return "strong"
    if borderline_ok:
        return "borderline"
    return "weak"

def geometry_with_basis(z0, e1, e2):
    s_vals = np.exp(np.linspace(np.log(1e-4), np.log(S_MAX), N_S))
    V = build_velocity_matrix(z0, s_vals)

    U, S, _ = svd(V, full_matrices=False)

    cos1 = abs(np.dot(U[:, 0], e1))

    r2s = []
    for k in range(V.shape[1]):
        vk = V[:, k]
        nrm = np.linalg.norm(vk)
        if nrm > 1e-15:
            proj = np.dot(vk, e1) * e1 + np.dot(vk, e2) * e2
            r2s.append(1.0 - (np.linalg.norm(vk - proj) / nrm)**2)

    R2 = float(np.mean(r2s))
    S1S2 = float(S[0] / S[1])
    S2S3 = float(S[1] / S[2])

    return {
        "cos1": float(cos1),
        "R2": R2,
        "S1S2": S1S2,
        "S2S3": S2S3,
        "tier": classify_geom(float(cos1), R2, S1S2, S2S3),
    }


# =========================================================
# Window extraction
# =========================================================

def window_indices(pool_len, n, shift):
    center = pool_len // 2 + shift
    start = center - n // 2
    end = start + n
    if start < 0 or end > pool_len:
        return None, None
    return start, end


# =========================================================
# Reporting
# =========================================================

def print_section(title):
    print()
    print("=" * 120)
    print(title)
    print("-" * 120)

def print_header():
    hdr = (
        f"{'pool':>5} {'n':>4} {'shift':>6} "
        f"{'basis':>9} {'tier':>11} "
        f"{'cos1':>9} {'R²':>9} {'S1/S2':>9} {'S2/S3':>9} "
        f"{'gminD':>8} {'gmaxD':>8} {'gapCV':>8} {'ddRMSD':>8} {'leftD':>8} {'rightD':>8}"
    )
    print(hdr)
    print("-" * len(hdr))

def print_row(pool_size, n, shift, basis_name, geom, sp):
    print(
        f"{pool_size:5d} {n:4d} {shift:6d} "
        f"{basis_name:>9s} {geom['tier']:>11s} "
        f"{geom['cos1']:9.6f} {geom['R2']:9.6f} {geom['S1S2']:9.3f} {geom['S2S3']:9.3f} "
        f"{sp['gminD']:8.4f} {sp['gmaxD']:8.4f} {sp['gapCV']:8.4f} {sp['ddRMSD']:8.4f} "
        f"{sp['left_gapD']:8.4f} {sp['right_gapD']:8.4f}"
    )


# =========================================================
# Main
# =========================================================

if __name__ == "__main__":
    print("t5000_weak_cluster_diagnostic.py")
    print("=" * 120)
    print(f"Target height T = {T_TARGET}")
    print("Reduced flow: dt_i/ds = +2 sum_{j!=i} 1/(t_i-t_j)")

    # -----------------------------------------------------
    # Part 1: test original weak windows across pool sizes
    # -----------------------------------------------------
    print_section("PART 1 — original weak windows across different zero-pool sizes")
    print_header()

    part1_rows = []

    for pool_size in POOL_SIZES:
        pool = find_zeros_pool(T_TARGET, pool_size)

        for n, shift in TARGET_WINDOWS:
            start, end = window_indices(len(pool), n, shift)
            if start is None:
                continue

            z0 = pool[start:end]
            sp = spacing_stats(z0)

            e1i, e2i = internal_basis(z0)
            gi = geometry_with_basis(z0, e1i, e2i)

            e1c, e2c = context_basis(pool, start, end, CONTEXT_PAD)
            gc = geometry_with_basis(z0, e1c, e2c)

            print_row(pool_size, n, shift, "internal", gi, sp)
            print_row(pool_size, n, shift, "context", gc, sp)

            part1_rows.append((pool_size, n, shift, gi, gc, sp))

    # -----------------------------------------------------
    # Part 2: neighbouring shifts and sizes
    # -----------------------------------------------------
    print_section("PART 2 — neighbourhood scan around T=5000")
    print_header()

    part2_rows = []

    pool = find_zeros_pool(T_TARGET, max(POOL_SIZES))

    for n in WINDOW_SIZES:
        for shift in NEIGHBOUR_SHIFTS:
            start, end = window_indices(len(pool), n, shift)
            if start is None:
                continue

            z0 = pool[start:end]
            sp = spacing_stats(z0)

            e1i, e2i = internal_basis(z0)
            gi = geometry_with_basis(z0, e1i, e2i)

            e1c, e2c = context_basis(pool, start, end, CONTEXT_PAD)
            gc = geometry_with_basis(z0, e1c, e2c)

            print_row(len(pool), n, shift, "internal", gi, sp)
            print_row(len(pool), n, shift, "context", gc, sp)

            part2_rows.append((n, shift, gi, gc, sp))

    # -----------------------------------------------------
    # Part 3: compact verdict tables
    # -----------------------------------------------------
    print_section("PART 3 — compact verdict: does context rescue weak geometry?")

    print(f"{'n':>4} {'shift':>6} {'pool':>5} {'internal':>11} {'context':>11} {'Δtier':>10}")
    print("-" * 60)

    tier_rank = {"weak": 0, "borderline": 1, "strong": 2}

    for pool_size, n, shift, gi, gc, sp in part1_rows:
        delta = tier_rank[gc["tier"]] - tier_rank[gi["tier"]]
        if delta > 0:
            dlabel = "rescued"
        elif delta == 0:
            dlabel = "same"
        else:
            dlabel = "worse"

        print(f"{n:4d} {shift:6d} {pool_size:5d} {gi['tier']:>11s} {gc['tier']:>11s} {dlabel:>10s}")

    print()
    print("Interpretation guide")
    print("-" * 80)
    print("If weak internal windows become strong/borderline under context basis:")
    print("  -> likely edge/basis artifact.")
    print("If weak internal windows remain weak under context basis and larger pools:")
    print("  -> likely genuine local softening.")
    print("If changing pool size changes the result:")
    print("  -> pool extraction alignment matters and the weak cluster is not intrinsic.")
```

```python
"""
small_gap_geometry_stress_test.py
=================================

Purpose
-------
Test whether weak reduced-flow geometry is predicted by unusually small
local gaps.

Hypothesis
----------
The T=5000 weak-geometry cluster is not random. It is associated with
windows containing a very small normalized gap:

    g_min / Δ̄ ≈ 0.11 - 0.13

This script scans all tested heights/windows and asks:

  1. Do weak-geometry windows have systematically smaller g_min / Δ̄?
  2. Is there a clean threshold below which geometry tends to weaken?
  3. Are there false positives: tiny gaps but still strong geometry?
  4. Are there false negatives: weak geometry without tiny gaps?

Unified convention
------------------
Reduced repulsive flow:

    dt_i/ds = +2 * sum_{j!=i} 1/(t_i - t_j),   s >= 0

This is the same reduced-flow convention used in previous diagnostics.
"""

import math
import mpmath
import numpy as np
from scipy.linalg import svd

mpmath.mp.dps = 25

# =========================================================
# User controls
# =========================================================

T_VALUES = [50, 100, 200, 500, 1000, 5000, 10000]
N_POOL = 56
WINDOW_SIZES = [12, 16, 20, 24]
SHIFTS = [-4, -3, -2, -1, 0, 1, 2, 3, 4]

S_MAX = 0.04
N_S = 100

# geometry tier thresholds
COS1_STRONG = 0.998
R2_STRONG   = 0.997
S1S2_STRONG = 10.0
S2S3_STRONG = 8.0

COS1_BORDER = 0.995
R2_BORDER   = 0.985
S1S2_BORDER = 8.0
S2S3_BORDER = 6.0

# candidate small-gap thresholds
GMIN_THRESHOLDS = [0.10, 0.12, 0.14, 0.16, 0.18, 0.20, 0.25, 0.30]


# =========================================================
# Zero finding
# =========================================================

def find_zeros_pool(T, n, span=None):
    if span is None:
        span = max(40.0, n * 2.0 * math.pi / math.log(T / (2.0 * math.pi)) * 1.7)

    ts = np.linspace(T - span, T + span, max(16000, int(span * 70)))
    zv = np.array([float(mpmath.siegelz(t)) for t in ts])

    roots = []
    for i in range(len(zv) - 1):
        if zv[i] * zv[i + 1] < 0:
            r = float(mpmath.findroot(mpmath.siegelz, (ts[i], ts[i + 1]), solver="illinois"))
            roots.append(r)

    roots = sorted(roots)

    if len(roots) < n:
        raise RuntimeError(f"Only found {len(roots)} zeros near T={T}; needed {n}. Increase span.")

    mid = len(roots) // 2
    start = max(0, mid - n // 2)
    end = start + n
    if end > len(roots):
        end = len(roots)
        start = end - n

    return np.array(roots[start:end], dtype=float)


# =========================================================
# Reduced flow
# =========================================================

def rk4_repulsive(z, ds):
    def vel(x):
        n = len(x)
        v = np.zeros(n, dtype=float)
        for i in range(n):
            for j in range(n):
                if i != j:
                    d = x[i] - x[j]
                    if abs(d) > 1e-12:
                        v[i] += 2.0 / d
        return v

    k1 = vel(z)
    k2 = vel(z + 0.5 * ds * k1)
    k3 = vel(z + 0.5 * ds * k2)
    k4 = vel(z + ds * k3)
    return z + (ds / 6.0) * (k1 + 2*k2 + 2*k3 + k4)

def build_velocity_matrix(z0, s_vals):
    s_full = np.concatenate([[0.0], s_vals])
    hist = np.zeros((len(s_full), len(z0)), dtype=float)
    hist[0] = z0.copy()

    z = z0.copy()
    for k in range(1, len(s_full)):
        z = rk4_repulsive(z, s_full[k] - s_full[k - 1])
        hist[k] = z

    V = np.zeros((len(z0), len(s_vals)), dtype=float)
    for k, s in enumerate(s_vals):
        col = (hist[k+1] - z0) / s
        col -= col.mean()
        V[:, k] = col

    return V


# =========================================================
# Basis and geometry
# =========================================================

def hadamard_direction(z0):
    n = len(z0)
    vh = np.zeros(n, dtype=float)

    for i in range(n):
        for j in range(n):
            if i != j:
                vh[i] += 2.0 / (z0[i] - z0[j])

    vh -= vh.mean()
    nrm = np.linalg.norm(vh)
    if nrm < 1e-15:
        return np.zeros_like(vh)
    return vh / nrm

def density_direction(z0, e1):
    n = len(z0)
    g = np.diff(z0)
    vd = np.zeros(n, dtype=float)

    for i in range(1, n - 1):
        di = 0.5 * (z0[i+1] - z0[i-1])
        if di > 1e-12:
            vd[i] = -(g[i] - g[i-1]) / (2.0 * di**2)

    if len(g) > 1:
        vd[0]  = -(g[1]  - g[0])  / (2.0 * g[0]**2)
        vd[-1] = -(g[-1] - g[-2]) / (2.0 * g[-1]**2)

    vd -= vd.mean()
    vd -= np.dot(vd, e1) * e1

    nrm = np.linalg.norm(vd)
    if nrm < 1e-15:
        return np.zeros_like(vd)
    return vd / nrm

def spacing_stats(z0):
    g = np.diff(z0)
    D = g.mean()
    dd = np.diff(g)

    return {
        "D": float(D),
        "gminD": float(g.min() / D),
        "gmaxD": float(g.max() / D),
        "gapCV": float(g.std() / D),
        "rangeD": float((g.max() - g.min()) / D),
        "ddRMSD": float(np.sqrt(np.mean(dd**2)) / D) if len(dd) else 0.0,
        "leftD": float(g[0] / D),
        "rightD": float(g[-1] / D),
        "min_gap_index": int(np.argmin(g)),
    }

def geometry_diagnostics(z0):
    s_vals = np.exp(np.linspace(np.log(1e-4), np.log(S_MAX), N_S))
    V = build_velocity_matrix(z0, s_vals)

    e1 = hadamard_direction(z0)
    e2 = density_direction(z0, e1)

    U, S, _ = svd(V, full_matrices=False)
    cos1 = abs(np.dot(U[:, 0], e1))

    r2s = []
    for k in range(V.shape[1]):
        vk = V[:, k]
        nrm = np.linalg.norm(vk)
        if nrm > 1e-15:
            proj = np.dot(vk, e1) * e1 + np.dot(vk, e2) * e2
            r2s.append(1.0 - (np.linalg.norm(vk - proj) / nrm)**2)

    R2 = float(np.mean(r2s))
    S1S2 = float(S[0] / S[1])
    S2S3 = float(S[1] / S[2])

    return {
        "cos1": float(cos1),
        "R2": R2,
        "S1S2": S1S2,
        "S2S3": S2S3,
    }

def classify_geometry(out):
    strong = (
        out["cos1"] >= COS1_STRONG and
        out["R2"] >= R2_STRONG and
        out["S1S2"] >= S1S2_STRONG and
        out["S2S3"] >= S2S3_STRONG
    )

    borderline = (
        out["cos1"] >= COS1_BORDER and
        out["R2"] >= R2_BORDER and
        out["S1S2"] >= S1S2_BORDER and
        out["S2S3"] >= S2S3_BORDER
    )

    if strong:
        return "strong"
    if borderline:
        return "borderline"
    return "weak"


# =========================================================
# Window extraction
# =========================================================

def extract_window(pool, n, shift):
    center = len(pool) // 2 + shift
    start = center - n // 2
    end = start + n
    if start < 0 or end > len(pool):
        return None
    return pool[start:end].copy()


# =========================================================
# Analysis helpers
# =========================================================

def summarize_by_tier(rows):
    tiers = ["strong", "borderline", "weak"]

    print()
    print("=" * 100)
    print("SUMMARY BY GEOMETRY TIER")
    print("-" * 100)
    print(f"{'tier':>12} {'N':>5} {'gminD_med':>12} {'gminD_mean':>12} {'gminD_min':>12} {'gminD_max':>12} {'gapCV_med':>12} {'ddRMSD_med':>12}")
    print("-" * 100)

    for tier in tiers:
        sub = [r for r in rows if r["tier"] == tier]
        if not sub:
            continue
        gmin = np.array([r["gminD"] for r in sub])
        gapcv = np.array([r["gapCV"] for r in sub])
        ddr = np.array([r["ddRMSD"] for r in sub])
        print(
            f"{tier:>12s} {len(sub):5d} "
            f"{np.median(gmin):12.4f} {np.mean(gmin):12.4f} "
            f"{gmin.min():12.4f} {gmin.max():12.4f} "
            f"{np.median(gapcv):12.4f} {np.median(ddr):12.4f}"
        )

def threshold_table(rows):
    print()
    print("=" * 100)
    print("SMALL-GAP THRESHOLD TABLE")
    print("-" * 100)
    print("Treat 'weak' as the positive class.")
    print()
    print(f"{'gminD<thr':>10} {'TP':>4} {'FP':>4} {'FN':>4} {'TN':>4} {'precision':>10} {'recall':>10} {'F1':>10}")
    print("-" * 100)

    for thr in GMIN_THRESHOLDS:
        tp = fp = fn = tn = 0
        for r in rows:
            pred = r["gminD"] < thr
            actual = r["tier"] == "weak"

            if pred and actual:
                tp += 1
            elif pred and not actual:
                fp += 1
            elif (not pred) and actual:
                fn += 1
            else:
                tn += 1

        prec = tp / (tp + fp) if (tp + fp) else 0.0
        rec = tp / (tp + fn) if (tp + fn) else 0.0
        f1 = 2 * prec * rec / (prec + rec) if (prec + rec) else 0.0

        print(f"{thr:10.2f} {tp:4d} {fp:4d} {fn:4d} {tn:4d} {prec:10.3f} {rec:10.3f} {f1:10.3f}")

def print_extreme_windows(rows):
    print()
    print("=" * 100)
    print("WINDOWS SORTED BY SMALLEST gminD")
    print("-" * 100)
    print(f"{'T':>6} {'n':>4} {'shift':>6} {'tier':>11} {'gminD':>8} {'gapCV':>8} {'ddRMSD':>8} {'cos1':>9} {'R²':>9} {'S1/S2':>9} {'minidx':>7}")
    print("-" * 100)

    for r in sorted(rows, key=lambda x: x["gminD"])[:40]:
        print(
            f"{r['T']:6d} {r['n']:4d} {r['shift']:6d} {r['tier']:>11s} "
            f"{r['gminD']:8.4f} {r['gapCV']:8.4f} {r['ddRMSD']:8.4f} "
            f"{r['cos1']:9.6f} {r['R2']:9.6f} {r['S1S2']:9.3f} {r['min_gap_index']:7d}"
        )

def correlation_report(rows):
    print()
    print("=" * 100)
    print("CORRELATIONS WITH GEOMETRY QUALITY")
    print("-" * 100)

    # geometry quality score: higher is better
    # use conservative components, normalised loosely
    quality = np.array([
        min(r["cos1"], 1.0)
        + min(r["R2"], 1.0)
        + min(r["S1S2"] / 20.0, 1.0)
        + min(r["S2S3"] / 20.0, 1.0)
        for r in rows
    ])

    keys = ["gminD", "gmaxD", "gapCV", "rangeD", "ddRMSD", "leftD", "rightD"]

    for k in keys:
        x = np.array([r[k] for r in rows])
        if np.std(x) < 1e-15:
            continue
        corr = np.corrcoef(x, quality)[0, 1]
        print(f"corr({k:>8s}, geometry_quality) = {corr:+.4f}")


# =========================================================
# Main
# =========================================================

if __name__ == "__main__":
    print("small_gap_geometry_stress_test.py")
    print("=" * 100)
    print("Reduced flow: dt_i/ds = +2 sum_{j!=i} 1/(t_i-t_j)")
    print()

    rows = []

    for T in T_VALUES:
        pool = find_zeros_pool(T, N_POOL)

        for n in WINDOW_SIZES:
            for shift in SHIFTS:
                z0 = extract_window(pool, n, shift)
                if z0 is None:
                    continue

                sp = spacing_stats(z0)
                gd = geometry_diagnostics(z0)
                tier = classify_geometry(gd)

                row = {
                    "T": T,
                    "n": n,
                    "shift": shift,
                    "tier": tier,
                    **sp,
                    **gd,
                }
                rows.append(row)

    print(f"Total windows scanned: {len(rows)}")

    summarize_by_tier(rows)
    threshold_table(rows)
    print_extreme_windows(rows)
    correlation_report(rows)

    print()
    print("=" * 100)
    print("INTERPRETATION GUIDE")
    print("-" * 100)
    print("If weak windows cluster below gminD≈0.14 with few false positives:")
    print("  -> small-gap compression is a strong diagnostic of geometry softening.")
    print("If many strong windows also have gminD below the same threshold:")
    print("  -> small gaps are not sufficient; another structural variable is needed.")
    print("If weak windows occur at large gminD:")
    print("  -> small-gap hypothesis is incomplete.")
```

```python
"""
local_force_decomposition.py
============================

Purpose
-------
Test the mechanism behind weak reduced-flow geometry.

Previous result:
  Weak geometry was perfectly classified by:

      g_min / Δ̄ < 0.14

This script asks WHY.

Hypothesis
----------
When a window contains an unusually small local gap, the reduced-flow
velocity is dominated by the adjacent-pair singular interaction around
that gap. This introduces a local direction that competes with the
global two-mode geometry.

Reduced flow convention
-----------------------
We use:

    dt_i/ds = +2 * sum_{j!=i} 1/(t_i - t_j),   s >= 0

For a very small gap g between t_k and t_{k+1}, the adjacent-pair
contribution scales like:

    4/g

in the gap equation.

What this script computes
-------------------------
For each window:
  1. geometry tier
  2. minimum gap index k
  3. force decomposition at s=0:
       - adjacent-pair contribution
       - local-neighbour contribution
       - background contribution
  4. norm fractions:
       - adjacent_pair_fraction
       - local_4body_fraction
       - background_fraction
  5. gap-equation decomposition:
       dg_min/ds = adjacent + local_tail + background_tail

Outputs compare:
  - strong windows
  - borderline windows
  - weak windows

The key question:
  Do weak windows have much larger adjacent/local dominance than
  healthy windows?
"""

import math
import mpmath
import numpy as np
from scipy.linalg import svd

mpmath.mp.dps = 25

# =========================================================
# User controls
# =========================================================

T_VALUES = [50, 100, 200, 500, 1000, 5000, 10000]
N_POOL = 56
WINDOW_SIZES = [12, 16, 20, 24]
SHIFTS = [-4, -3, -2, -1, 0, 1, 2, 3, 4]

S_MAX = 0.04
N_S = 100

# Geometry tier thresholds
COS1_STRONG = 0.998
R2_STRONG   = 0.997
S1S2_STRONG = 10.0
S2S3_STRONG = 8.0

COS1_BORDER = 0.995
R2_BORDER   = 0.985
S1S2_BORDER = 8.0
S2S3_BORDER = 6.0


# =========================================================
# Zero finding
# =========================================================

def find_zeros_pool(T, n, span=None):
    if span is None:
        span = max(40.0, n * 2.0 * math.pi / math.log(T / (2.0 * math.pi)) * 1.7)

    ts = np.linspace(T - span, T + span, max(16000, int(span * 70)))
    zv = np.array([float(mpmath.siegelz(t)) for t in ts])

    roots = []
    for i in range(len(zv) - 1):
        if zv[i] * zv[i + 1] < 0:
            r = float(mpmath.findroot(mpmath.siegelz, (ts[i], ts[i + 1]), solver="illinois"))
            roots.append(r)

    roots = sorted(roots)

    if len(roots) < n:
        raise RuntimeError(f"Only found {len(roots)} zeros near T={T}; needed {n}. Increase span.")

    mid = len(roots) // 2
    start = max(0, mid - n // 2)
    end = start + n
    if end > len(roots):
        end = len(roots)
        start = end - n

    return np.array(roots[start:end], dtype=float)


# =========================================================
# Reduced flow
# =========================================================

def reduced_velocity(z):
    """
    Full reduced repulsive velocity:

        v_i = 2 sum_{j!=i} 1/(z_i - z_j)
    """
    n = len(z)
    v = np.zeros(n, dtype=float)

    for i in range(n):
        for j in range(n):
            if i != j:
                d = z[i] - z[j]
                if abs(d) > 1e-12:
                    v[i] += 2.0 / d

    return v

def rk4_repulsive(z, ds):
    def vel(x):
        return reduced_velocity(x)

    k1 = vel(z)
    k2 = vel(z + 0.5 * ds * k1)
    k3 = vel(z + 0.5 * ds * k2)
    k4 = vel(z + ds * k3)
    return z + (ds / 6.0) * (k1 + 2*k2 + 2*k3 + k4)

def build_velocity_matrix(z0, s_vals):
    s_full = np.concatenate([[0.0], s_vals])
    hist = np.zeros((len(s_full), len(z0)), dtype=float)
    hist[0] = z0.copy()

    z = z0.copy()
    for k in range(1, len(s_full)):
        z = rk4_repulsive(z, s_full[k] - s_full[k - 1])
        hist[k] = z

    V = np.zeros((len(z0), len(s_vals)), dtype=float)
    for k, s in enumerate(s_vals):
        col = (hist[k+1] - z0) / s
        col -= col.mean()
        V[:, k] = col

    return V


# =========================================================
# Geometry
# =========================================================

def hadamard_direction(z0):
    vh = reduced_velocity(z0)
    vh -= vh.mean()
    nrm = np.linalg.norm(vh)
    if nrm < 1e-15:
        return np.zeros_like(vh)
    return vh / nrm

def density_direction(z0, e1):
    n = len(z0)
    g = np.diff(z0)
    vd = np.zeros(n, dtype=float)

    for i in range(1, n - 1):
        di = 0.5 * (z0[i+1] - z0[i-1])
        if di > 1e-12:
            vd[i] = -(g[i] - g[i-1]) / (2.0 * di**2)

    if len(g) > 1:
        vd[0]  = -(g[1]  - g[0])  / (2.0 * g[0]**2)
        vd[-1] = -(g[-1] - g[-2]) / (2.0 * g[-1]**2)

    vd -= vd.mean()
    vd -= np.dot(vd, e1) * e1

    nrm = np.linalg.norm(vd)
    if nrm < 1e-15:
        return np.zeros_like(vd)
    return vd / nrm

def geometry_diagnostics(z0):
    s_vals = np.exp(np.linspace(np.log(1e-4), np.log(S_MAX), N_S))
    V = build_velocity_matrix(z0, s_vals)

    e1 = hadamard_direction(z0)
    e2 = density_direction(z0, e1)

    U, S, _ = svd(V, full_matrices=False)

    cos1 = abs(np.dot(U[:, 0], e1))

    r2s = []
    for k in range(V.shape[1]):
        vk = V[:, k]
        nrm = np.linalg.norm(vk)
        if nrm > 1e-15:
            proj = np.dot(vk, e1) * e1 + np.dot(vk, e2) * e2
            r2s.append(1.0 - (np.linalg.norm(vk - proj) / nrm)**2)

    R2 = float(np.mean(r2s))
    S1S2 = float(S[0] / S[1])
    S2S3 = float(S[1] / S[2])

    return {
        "cos1": float(cos1),
        "R2": R2,
        "S1S2": S1S2,
        "S2S3": S2S3,
    }

def classify_geometry(out):
    strong = (
        out["cos1"] >= COS1_STRONG and
        out["R2"] >= R2_STRONG and
        out["S1S2"] >= S1S2_STRONG and
        out["S2S3"] >= S2S3_STRONG
    )

    borderline = (
        out["cos1"] >= COS1_BORDER and
        out["R2"] >= R2_BORDER and
        out["S1S2"] >= S1S2_BORDER and
        out["S2S3"] >= S2S3_BORDER
    )

    if strong:
        return "strong"
    if borderline:
        return "borderline"
    return "weak"


# =========================================================
# Spacing stats
# =========================================================

def spacing_stats(z):
    g = np.diff(z)
    D = g.mean()
    k = int(np.argmin(g))

    return {
        "D": float(D),
        "gmin": float(g[k]),
        "gminD": float(g[k] / D),
        "gmaxD": float(g.max() / D),
        "gapCV": float(g.std() / D),
        "min_gap_index": k,
    }


# =========================================================
# Force decomposition
# =========================================================

def contribution_from_set(z, target_indices, source_indices):
    """
    Contribution to velocity on target_indices from source_indices.

    Returns an n-vector with nonzero entries only at target indices.
    """
    n = len(z)
    v = np.zeros(n, dtype=float)

    for i in target_indices:
        for j in source_indices:
            if i == j:
                continue
            d = z[i] - z[j]
            if abs(d) > 1e-12:
                v[i] += 2.0 / d

    return v

def full_contribution_from_sources(z, source_indices):
    """
    Contribution to every particle from a subset of source particles.
    """
    n = len(z)
    v = np.zeros(n, dtype=float)

    for i in range(n):
        for j in source_indices:
            if i == j:
                continue
            d = z[i] - z[j]
            if abs(d) > 1e-12:
                v[i] += 2.0 / d

    v -= v.mean()
    return v

def force_decomposition(z):
    """
    Decompose velocity around the minimum gap.

    Let min gap be between k and k+1.

    Adjacent-pair vector:
      contribution only between k and k+1.

    Local-4 source set:
      {k-1, k, k+1, k+2} intersect window.

    Background:
      all remaining sources.

    We compute:
      - norm fractions of full velocity
      - gap derivative decomposition
    """
    n = len(z)
    g = np.diff(z)
    D = g.mean()
    k = int(np.argmin(g))

    i_left = k
    i_right = k + 1

    pair_set = [i_left, i_right]

    local_set = sorted(set([x for x in [k-1, k, k+1, k+2] if 0 <= x < n]))
    background_set = [j for j in range(n) if j not in local_set]

    full_v = reduced_velocity(z)
    full_v -= full_v.mean()
    full_norm = np.linalg.norm(full_v) + 1e-30

    # Adjacent pair contribution as a source set restricted to the pair
    pair_v = np.zeros(n, dtype=float)
    pair_v[i_left]  = 2.0 / (z[i_left]  - z[i_right])
    pair_v[i_right] = 2.0 / (z[i_right] - z[i_left])
    pair_v -= pair_v.mean()

    local_v = full_contribution_from_sources(z, local_set)
    background_v = full_contribution_from_sources(z, background_set)

    # Orthogonal residual check, not expected to be zero because source
    # partitions include induced velocities on all particles.
    pair_frac = np.linalg.norm(pair_v) / full_norm
    local_frac = np.linalg.norm(local_v) / full_norm
    background_frac = np.linalg.norm(background_v) / full_norm

    # Gap derivative decomposition
    # g = z[k+1] - z[k]
    # dg/ds = v[k+1] - v[k]
    full_gap_deriv = full_v[i_right] - full_v[i_left]

    pair_gap_deriv = (
        2.0 / (z[i_right] - z[i_left])
        -
        2.0 / (z[i_left] - z[i_right])
    )

    # contributions to gap derivative from nearby non-pair neighbours
    local_tail_sources = [j for j in local_set if j not in pair_set]
    bg_sources = [j for j in range(n) if j not in local_set]

    def gap_contribution_from_sources(source_indices):
        vr = 0.0
        vl = 0.0
        for j in source_indices:
            if j != i_right:
                vr += 2.0 / (z[i_right] - z[j])
            if j != i_left:
                vl += 2.0 / (z[i_left] - z[j])
        return vr - vl

    local_tail_gap = gap_contribution_from_sources(local_tail_sources)
    background_gap = gap_contribution_from_sources(bg_sources)

    # theoretical compression coefficient:
    # dg/ds = 4/g - C*g
    gmin = g[k]
    C_tail = (pair_gap_deriv - full_gap_deriv) / (gmin + 1e-30)

    return {
        "min_gap_index": k,
        "gminD": float(gmin / D),
        "pair_frac": float(pair_frac),
        "local4_frac": float(local_frac),
        "background_frac": float(background_frac),
        "full_gap_deriv": float(full_gap_deriv),
        "pair_gap_deriv": float(pair_gap_deriv),
        "local_tail_gap": float(local_tail_gap),
        "background_gap": float(background_gap),
        "tail_total_gap": float(local_tail_gap + background_gap),
        "tail_over_pair": float(abs(local_tail_gap + background_gap) / (abs(pair_gap_deriv) + 1e-30)),
        "C_tail": float(C_tail),
    }


# =========================================================
# Window extraction
# =========================================================

def extract_window(pool, n, shift):
    center = len(pool) // 2 + shift
    start = center - n // 2
    end = start + n
    if start < 0 or end > len(pool):
        return None
    return pool[start:end].copy()


# =========================================================
# Reporting
# =========================================================

def summarize_by_tier(rows):
    print()
    print("=" * 120)
    print("SUMMARY BY GEOMETRY TIER")
    print("-" * 120)

    hdr = (
        f"{'tier':>12} {'N':>5} "
        f"{'gminD_med':>10} {'pairFrac_med':>13} {'local4_med':>11} "
        f"{'bg_med':>9} {'tail/pair_med':>14} {'Ctail_med':>11}"
    )
    print(hdr)
    print("-" * len(hdr))

    for tier in ["strong", "borderline", "weak"]:
        sub = [r for r in rows if r["tier"] == tier]
        if not sub:
            continue

        def med(key):
            return float(np.median([r[key] for r in sub]))

        print(
            f"{tier:>12s} {len(sub):5d} "
            f"{med('gminD'):10.4f} {med('pair_frac'):13.4f} {med('local4_frac'):11.4f} "
            f"{med('background_frac'):9.4f} {med('tail_over_pair'):14.4f} {med('C_tail'):11.4f}"
        )

def threshold_report(rows):
    print()
    print("=" * 120)
    print("PAIR / LOCAL DOMINANCE THRESHOLD CHECK")
    print("-" * 120)

    weak = [r for r in rows if r["tier"] == "weak"]
    nonweak = [r for r in rows if r["tier"] != "weak"]

    for key in ["pair_frac", "local4_frac", "tail_over_pair", "C_tail"]:
        weak_vals = np.array([r[key] for r in weak])
        non_vals = np.array([r[key] for r in nonweak])

        print()
        print(f"{key}:")
        print(f"  weak    median={np.median(weak_vals):.4f}, min={weak_vals.min():.4f}, max={weak_vals.max():.4f}")
        print(f"  nonweak median={np.median(non_vals):.4f}, min={non_vals.min():.4f}, max={non_vals.max():.4f}")

def print_weak_windows(rows):
    print()
    print("=" * 120)
    print("WEAK WINDOWS — LOCAL FORCE DETAILS")
    print("-" * 120)

    hdr = (
        f"{'T':>6} {'n':>4} {'shift':>6} {'k':>4} "
        f"{'gminD':>8} {'cos1':>9} {'R²':>9} {'S1/S2':>9} "
        f"{'pairFrac':>10} {'local4':>9} {'bg':>9} "
        f"{'pairGap':>10} {'tailGap':>10} {'tail/pair':>10} {'Ctail':>10}"
    )
    print(hdr)
    print("-" * len(hdr))

    for r in sorted([x for x in rows if x["tier"] == "weak"], key=lambda x: x["gminD"]):
        print(
            f"{r['T']:6d} {r['n']:4d} {r['shift']:6d} {r['min_gap_index']:4d} "
            f"{r['gminD']:8.4f} {r['cos1']:9.6f} {r['R2']:9.6f} {r['S1S2']:9.3f} "
            f"{r['pair_frac']:10.4f} {r['local4_frac']:9.4f} {r['background_frac']:9.4f} "
            f"{r['pair_gap_deriv']:10.4f} {r['tail_total_gap']:10.4f} "
            f"{r['tail_over_pair']:10.4f} {r['C_tail']:10.4f}"
        )

def correlation_report(rows):
    print()
    print("=" * 120)
    print("CORRELATIONS WITH GEOMETRY QUALITY")
    print("-" * 120)

    quality = np.array([
        min(r["cos1"], 1.0)
        + min(r["R2"], 1.0)
        + min(r["S1S2"] / 20.0, 1.0)
        + min(r["S2S3"] / 20.0, 1.0)
        for r in rows
    ])

    keys = ["gminD", "pair_frac", "local4_frac", "background_frac", "tail_over_pair", "C_tail"]

    for key in keys:
        x = np.array([r[key] for r in rows])
        if np.std(x) < 1e-15:
            continue
        corr = np.corrcoef(x, quality)[0, 1]
        print(f"corr({key:>15s}, geometry_quality) = {corr:+.4f}")


# =========================================================
# Main
# =========================================================

if __name__ == "__main__":
    print("local_force_decomposition.py")
    print("=" * 120)
    print("Reduced flow: dt_i/ds = +2 sum_{j!=i} 1/(t_i-t_j)")
    print()

    rows = []

    for T in T_VALUES:
        pool = find_zeros_pool(T, N_POOL)

        for n in WINDOW_SIZES:
            for shift in SHIFTS:
                z0 = extract_window(pool, n, shift)
                if z0 is None:
                    continue

                sp = spacing_stats(z0)
                gd = geometry_diagnostics(z0)
                tier = classify_geometry(gd)
                fd = force_decomposition(z0)

                row = {
                    "T": T,
                    "n": n,
                    "shift": shift,
                    "tier": tier,
                    **sp,
                    **gd,
                    **fd,
                }
                rows.append(row)

    print(f"Total windows scanned: {len(rows)}")

    summarize_by_tier(rows)
    threshold_report(rows)
    print_weak_windows(rows)
    correlation_report(rows)

    print()
    print("=" * 120)
    print("INTERPRETATION GUIDE")
    print("-" * 120)
    print("If weak windows have much larger pair_frac/local4_frac:")
    print("  -> weak geometry is driven by local singular interaction dominance.")
    print("If weak windows mainly have larger tail_over_pair or C_tail:")
    print("  -> weak geometry comes from tail compression overwhelming the pair mode.")
    print("If force fractions overlap strongly between weak and nonweak:")
    print("  -> small gap is a classifier, but force decomposition needs a better basis.")
```

```python
"""
local_rescaled_geometry.py
==========================

Purpose
-------
Test whether weak reduced-flow geometry recovers under local minimum-gap
rescaling.

Previous established results:
  1. Weak geometry occurs exactly when:

         g_min / Δ̄ < 0.14

  2. Weak windows are dynamically local-pair dominated:

         pairFrac ≈ 0.80

  3. Tail contribution is tiny:

         tail/pair ≈ 0.01

Question
--------
Is weak geometry merely a global-scaling artifact?

Test
----
For each window, compare geometry metrics under two coordinate choices:

  A. raw coordinates:
       t_i

  B. local minimum-gap coordinates:
       ttilde_i = (t_i - t_k) / g_min

where g_min = t_{k+1} - t_k.

Important note
--------------
The reduced ODE

    dt_i/ds = 2 Σ 1/(t_i - t_j)

has a scaling symmetry. If all coordinates are scaled by a factor a,
then the natural flow-time scale changes like a^2.

Therefore this script tests both:
  - fixed s-range
  - locally scaled s-range

to avoid mistaking time-scale mismatch for geometry failure.

Reduced flow convention
-----------------------
    dt_i/ds = +2 * sum_{j!=i} 1/(t_i - t_j),   s >= 0
"""

import math
import mpmath
import numpy as np
from scipy.linalg import svd

mpmath.mp.dps = 25

# =========================================================
# User controls
# =========================================================

T_VALUES = [50, 100, 200, 500, 1000, 5000, 10000]
N_POOL = 56
WINDOW_SIZES = [12, 16, 20, 24]
SHIFTS = [-4, -3, -2, -1, 0, 1, 2, 3, 4]

S_MAX = 0.04
N_S = 100

# geometry thresholds
COS1_STRONG = 0.998
R2_STRONG   = 0.997
S1S2_STRONG = 10.0
S2S3_STRONG = 8.0

COS1_BORDER = 0.995
R2_BORDER   = 0.985
S1S2_BORDER = 8.0
S2S3_BORDER = 6.0


# =========================================================
# Zero finding
# =========================================================

def find_zeros_pool(T, n, span=None):
    if span is None:
        span = max(40.0, n * 2.0 * math.pi / math.log(T / (2.0 * math.pi)) * 1.7)

    ts = np.linspace(T - span, T + span, max(16000, int(span * 70)))
    zv = np.array([float(mpmath.siegelz(t)) for t in ts])

    roots = []
    for i in range(len(zv) - 1):
        if zv[i] * zv[i + 1] < 0:
            r = float(mpmath.findroot(mpmath.siegelz, (ts[i], ts[i + 1]), solver="illinois"))
            roots.append(r)

    roots = sorted(roots)

    if len(roots) < n:
        raise RuntimeError(f"Only found {len(roots)} zeros near T={T}; needed {n}. Increase span.")

    mid = len(roots) // 2
    start = max(0, mid - n // 2)
    end = start + n

    if end > len(roots):
        end = len(roots)
        start = end - n

    return np.array(roots[start:end], dtype=float)


# =========================================================
# Reduced flow
# =========================================================

def reduced_velocity(z):
    n = len(z)
    v = np.zeros(n, dtype=float)

    for i in range(n):
        for j in range(n):
            if i != j:
                d = z[i] - z[j]
                if abs(d) > 1e-12:
                    v[i] += 2.0 / d

    return v

def rk4_repulsive(z, ds):
    def vel(x):
        return reduced_velocity(x)

    k1 = vel(z)
    k2 = vel(z + 0.5 * ds * k1)
    k3 = vel(z + 0.5 * ds * k2)
    k4 = vel(z + ds * k3)

    return z + (ds / 6.0) * (k1 + 2*k2 + 2*k3 + k4)

def build_velocity_matrix(z0, s_vals):
    s_full = np.concatenate([[0.0], s_vals])
    hist = np.zeros((len(s_full), len(z0)), dtype=float)
    hist[0] = z0.copy()

    z = z0.copy()
    for k in range(1, len(s_full)):
        z = rk4_repulsive(z, s_full[k] - s_full[k - 1])
        hist[k] = z

    V = np.zeros((len(z0), len(s_vals)), dtype=float)

    for k, s in enumerate(s_vals):
        col = (hist[k + 1] - z0) / s
        col -= col.mean()
        V[:, k] = col

    return V


# =========================================================
# Geometry basis
# =========================================================

def hadamard_direction(z0):
    vh = reduced_velocity(z0)
    vh -= vh.mean()
    nrm = np.linalg.norm(vh)
    if nrm < 1e-15:
        return np.zeros_like(vh)
    return vh / nrm

def density_direction(z0, e1):
    n = len(z0)
    g = np.diff(z0)
    vd = np.zeros(n, dtype=float)

    for i in range(1, n - 1):
        di = 0.5 * (z0[i + 1] - z0[i - 1])
        if di > 1e-12:
            vd[i] = -(g[i] - g[i - 1]) / (2.0 * di**2)

    if len(g) > 1:
        vd[0]  = -(g[1]  - g[0])  / (2.0 * g[0]**2)
        vd[-1] = -(g[-1] - g[-2]) / (2.0 * g[-1]**2)

    vd -= vd.mean()
    vd -= np.dot(vd, e1) * e1

    nrm = np.linalg.norm(vd)
    if nrm < 1e-15:
        return np.zeros_like(vd)

    return vd / nrm


# =========================================================
# Geometry diagnostics
# =========================================================

def classify_geometry(out):
    strong = (
        out["cos1"] >= COS1_STRONG and
        out["R2"] >= R2_STRONG and
        out["S1S2"] >= S1S2_STRONG and
        out["S2S3"] >= S2S3_STRONG
    )

    borderline = (
        out["cos1"] >= COS1_BORDER and
        out["R2"] >= R2_BORDER and
        out["S1S2"] >= S1S2_BORDER and
        out["S2S3"] >= S2S3_BORDER
    )

    if strong:
        return "strong"
    if borderline:
        return "borderline"
    return "weak"

def geometry_diagnostics(z0, s_max):
    s_vals = np.exp(np.linspace(np.log(1e-4), np.log(s_max), N_S))
    V = build_velocity_matrix(z0, s_vals)

    e1 = hadamard_direction(z0)
    e2 = density_direction(z0, e1)

    U, S, _ = svd(V, full_matrices=False)
    cos1 = abs(np.dot(U[:, 0], e1))

    r2s = []

    for k in range(V.shape[1]):
        vk = V[:, k]
        nrm = np.linalg.norm(vk)

        if nrm > 1e-15:
            proj = np.dot(vk, e1) * e1 + np.dot(vk, e2) * e2
            r2s.append(1.0 - (np.linalg.norm(vk - proj) / nrm)**2)

    out = {
        "cos1": float(cos1),
        "R2": float(np.mean(r2s)),
        "S1S2": float(S[0] / S[1]),
        "S2S3": float(S[1] / S[2]),
    }

    out["tier"] = classify_geometry(out)
    return out


# =========================================================
# Spacing and scaling
# =========================================================

def spacing_stats(z):
    g = np.diff(z)
    D = g.mean()
    k = int(np.argmin(g))

    return {
        "D": float(D),
        "gmin": float(g[k]),
        "gminD": float(g[k] / D),
        "min_gap_index": k,
    }

def local_rescale(z):
    """
    Rescale around the minimum gap.

    ttilde_i = (t_i - t_k) / g_min
    """
    g = np.diff(z)
    k = int(np.argmin(g))
    gmin = g[k]

    return (z - z[k]) / gmin, k, gmin


# =========================================================
# Force pair fraction
# =========================================================

def pair_fraction(z):
    g = np.diff(z)
    k = int(np.argmin(g))
    n = len(z)

    full_v = reduced_velocity(z)
    full_v -= full_v.mean()
    full_norm = np.linalg.norm(full_v) + 1e-30

    pair_v = np.zeros(n, dtype=float)
    pair_v[k]     = 2.0 / (z[k]     - z[k + 1])
    pair_v[k + 1] = 2.0 / (z[k + 1] - z[k])

    pair_v -= pair_v.mean()

    return float(np.linalg.norm(pair_v) / full_norm)


# =========================================================
# Window extraction
# =========================================================

def extract_window(pool, n, shift):
    center = len(pool) // 2 + shift
    start = center - n // 2
    end = start + n

    if start < 0 or end > len(pool):
        return None

    return pool[start:end].copy()


# =========================================================
# Reporting helpers
# =========================================================

def print_header():
    hdr = (
        f"{'T':>6} {'n':>4} {'shift':>6} "
        f"{'gminD':>8} {'pairF':>8} "
        f"{'raw':>10} {'loc_fixed':>10} {'loc_scaled':>11} "
        f"{'raw_cos':>9} {'raw_R²':>9} {'raw_S12':>9} "
        f"{'scaled_cos':>10} {'scaled_R²':>10} {'scaled_S12':>10}"
    )
    print(hdr)
    print("-" * len(hdr))

def print_row(r):
    print(
        f"{r['T']:6d} {r['n']:4d} {r['shift']:6d} "
        f"{r['gminD']:8.4f} {r['pair_frac']:8.4f} "
        f"{r['raw_tier']:>10s} {r['local_fixed_tier']:>10s} {r['local_scaled_tier']:>11s} "
        f"{r['raw_cos1']:9.6f} {r['raw_R2']:9.6f} {r['raw_S1S2']:9.3f} "
        f"{r['scaled_cos1']:10.6f} {r['scaled_R2']:10.6f} {r['scaled_S1S2']:10.3f}"
    )

def summarize(rows):
    print()
    print("=" * 120)
    print("SUMMARY")
    print("-" * 120)

    labels = ["strong", "borderline", "weak"]

    for key in ["raw_tier", "local_fixed_tier", "local_scaled_tier"]:
        print()
        print(key)
        counts = {lab: 0 for lab in labels}
        for r in rows:
            counts[r[key]] += 1
        for lab in labels:
            print(f"  {lab:>10s}: {counts[lab]}")

    print()
    print("Transition summary: raw -> local_scaled")
    trans = {}
    for r in rows:
        t = (r["raw_tier"], r["local_scaled_tier"])
        trans[t] = trans.get(t, 0) + 1

    for (a, b), c in sorted(trans.items()):
        print(f"  {a:>10s} -> {b:<10s}: {c}")

    print()
    weak_rows = [r for r in rows if r["raw_tier"] == "weak"]
    if weak_rows:
        rec = sum(1 for r in weak_rows if r["local_scaled_tier"] != "weak")
        print(f"Weak windows rescued by local_scaled geometry: {rec}/{len(weak_rows)}")

        med_raw_s12 = np.median([r["raw_S1S2"] for r in weak_rows])
        med_scaled_s12 = np.median([r["scaled_S1S2"] for r in weak_rows])
        med_raw_r2 = np.median([r["raw_R2"] for r in weak_rows])
        med_scaled_r2 = np.median([r["scaled_R2"] for r in weak_rows])

        print(f"Weak median S1/S2 raw -> scaled: {med_raw_s12:.4f} -> {med_scaled_s12:.4f}")
        print(f"Weak median R² raw -> scaled:    {med_raw_r2:.4f} -> {med_scaled_r2:.4f}")


# =========================================================
# Main
# =========================================================

if __name__ == "__main__":
    print("local_rescaled_geometry.py")
    print("=" * 120)
    print("Reduced flow: dt_i/ds = +2 sum_{j!=i} 1/(t_i-t_j)")
    print()

    rows = []

    for T in T_VALUES:
        pool = find_zeros_pool(T, N_POOL)

        for n in WINDOW_SIZES:
            for shift in SHIFTS:
                z0 = extract_window(pool, n, shift)
                if z0 is None:
                    continue

                sp = spacing_stats(z0)
                pf = pair_fraction(z0)

                raw = geometry_diagnostics(z0, S_MAX)

                zloc, k, gmin = local_rescale(z0)

                # Test 1: same absolute s range after local rescaling
                local_fixed = geometry_diagnostics(zloc, S_MAX)

                # Test 2: scale s range by coordinate scaling symmetry.
                # zloc = z / gmin, so lengths are scaled by 1/gmin.
                # Natural time scale scales as length^2, so s should scale by 1/gmin^2.
                local_scaled_smax = S_MAX / (gmin**2)
                local_scaled = geometry_diagnostics(zloc, local_scaled_smax)

                row = {
                    "T": T,
                    "n": n,
                    "shift": shift,
                    "gminD": sp["gminD"],
                    "pair_frac": pf,

                    "raw_tier": raw["tier"],
                    "raw_cos1": raw["cos1"],
                    "raw_R2": raw["R2"],
                    "raw_S1S2": raw["S1S2"],
                    "raw_S2S3": raw["S2S3"],

                    "local_fixed_tier": local_fixed["tier"],
                    "fixed_cos1": local_fixed["cos1"],
                    "fixed_R2": local_fixed["R2"],
                    "fixed_S1S2": local_fixed["S1S2"],
                    "fixed_S2S3": local_fixed["S2S3"],

                    "local_scaled_tier": local_scaled["tier"],
                    "scaled_cos1": local_scaled["cos1"],
                    "scaled_R2": local_scaled["R2"],
                    "scaled_S1S2": local_scaled["S1S2"],
                    "scaled_S2S3": local_scaled["S2S3"],
                }

                rows.append(row)

    # sort raw weak first, then by gmin
    tier_rank = {"weak": 0, "borderline": 1, "strong": 2}
    rows = sorted(rows, key=lambda r: (tier_rank[r["raw_tier"]], r["gminD"]))

    print_header()
    for r in rows:
        # keep full printout; user can inspect all transitions
        print_row(r)

    summarize(rows)

    print()
    print("=" * 120)
    print("INTERPRETATION GUIDE")
    print("-" * 120)
    print("If weak raw windows become strong/borderline under local_scaled:")
    print("  -> weak geometry was mainly a scaling mismatch.")
    print("If weak raw windows remain weak under local_scaled:")
    print("  -> weak geometry is a true local-pair dominated regime.")
    print("If local_fixed and local_scaled differ strongly:")
    print("  -> time-scale rescaling matters under coordinate scaling.")
```

```python
"""
local_minimum_barrier_check.py
==============================

Purpose
-------
Directly test the corrected finite-model small-gap barrier.

Finite reduced model:
    dt_i/ds = +2 * sum_{j!=i} 1/(t_i - t_j)

For a neighbouring gap:
    g_i = t_{i+1} - t_i

Exact finite-model identity:
    g_i' = 4/g_i - C_i g_i

where the nonlocal tail is compressive.

Corrected infinite-tail sufficient condition
--------------------------------------------
Let:
    g_min = η Δbar
    g_second = a Δbar

If all other gaps are at least a Δbar and:

    η < a

then the minimum gap is in the reopening regime:
    g_min' > 0

This script checks, for every window:
  - g_min / Δbar
  - g_second / Δbar
  - rho = g_min / g_second
  - actual g_min'
  - pair term 4/g_min
  - tail term
  - whether η < a predicts reopening
  - whether actual g_min' confirms reopening

Key question
------------
Are the weak-geometry windows actually protected by a local-minimum
barrier rather than dangerous collapse windows?
"""

import math
import mpmath
import numpy as np
from scipy.linalg import svd

mpmath.mp.dps = 25

# =========================================================
# User controls
# =========================================================

T_VALUES = [50, 100, 200, 500, 1000, 5000, 10000]
N_POOL = 56
WINDOW_SIZES = [12, 16, 20, 24]
SHIFTS = [-4, -3, -2, -1, 0, 1, 2, 3, 4]

S_MAX = 0.04
N_S = 100

# Geometry tier thresholds
COS1_STRONG = 0.998
R2_STRONG   = 0.997
S1S2_STRONG = 10.0
S2S3_STRONG = 8.0

COS1_BORDER = 0.995
R2_BORDER   = 0.985
S1S2_BORDER = 8.0
S2S3_BORDER = 6.0


# =========================================================
# Zero finding
# =========================================================

def find_zeros_pool(T, n, span=None):
    if span is None:
        span = max(40.0, n * 2.0 * math.pi / math.log(T / (2.0 * math.pi)) * 1.7)

    ts = np.linspace(T - span, T + span, max(16000, int(span * 70)))
    zv = np.array([float(mpmath.siegelz(t)) for t in ts])

    roots = []
    for i in range(len(zv) - 1):
        if zv[i] * zv[i + 1] < 0:
            r = float(mpmath.findroot(mpmath.siegelz, (ts[i], ts[i + 1]), solver="illinois"))
            roots.append(r)

    roots = sorted(roots)

    if len(roots) < n:
        raise RuntimeError(f"Only found {len(roots)} zeros near T={T}; needed {n}. Increase span.")

    mid = len(roots) // 2
    start = max(0, mid - n // 2)
    end = start + n

    if end > len(roots):
        end = len(roots)
        start = end - n

    return np.array(roots[start:end], dtype=float)


# =========================================================
# Reduced flow
# =========================================================

def reduced_velocity(z):
    n = len(z)
    v = np.zeros(n, dtype=float)

    for i in range(n):
        for j in range(n):
            if i != j:
                d = z[i] - z[j]
                if abs(d) > 1e-12:
                    v[i] += 2.0 / d

    return v

def rk4_repulsive(z, ds):
    def vel(x):
        return reduced_velocity(x)

    k1 = vel(z)
    k2 = vel(z + 0.5 * ds * k1)
    k3 = vel(z + 0.5 * ds * k2)
    k4 = vel(z + ds * k3)

    return z + (ds / 6.0) * (k1 + 2*k2 + 2*k3 + k4)

def build_velocity_matrix(z0, s_vals):
    s_full = np.concatenate([[0.0], s_vals])
    hist = np.zeros((len(s_full), len(z0)), dtype=float)
    hist[0] = z0.copy()

    z = z0.copy()
    for k in range(1, len(s_full)):
        z = rk4_repulsive(z, s_full[k] - s_full[k-1])
        hist[k] = z

    V = np.zeros((len(z0), len(s_vals)), dtype=float)
    for k, s in enumerate(s_vals):
        col = (hist[k+1] - z0) / s
        col -= col.mean()
        V[:, k] = col

    return V


# =========================================================
# Geometry
# =========================================================

def hadamard_direction(z0):
    vh = reduced_velocity(z0)
    vh -= vh.mean()
    nrm = np.linalg.norm(vh)
    if nrm < 1e-15:
        return np.zeros_like(vh)
    return vh / nrm

def density_direction(z0, e1):
    n = len(z0)
    g = np.diff(z0)
    vd = np.zeros(n, dtype=float)

    for i in range(1, n - 1):
        di = 0.5 * (z0[i+1] - z0[i-1])
        if di > 1e-12:
            vd[i] = -(g[i] - g[i-1]) / (2.0 * di**2)

    if len(g) > 1:
        vd[0]  = -(g[1]  - g[0])  / (2.0 * g[0]**2)
        vd[-1] = -(g[-1] - g[-2]) / (2.0 * g[-1]**2)

    vd -= vd.mean()
    vd -= np.dot(vd, e1) * e1

    nrm = np.linalg.norm(vd)
    if nrm < 1e-15:
        return np.zeros_like(vd)

    return vd / nrm

def geometry_diagnostics(z0):
    s_vals = np.exp(np.linspace(np.log(1e-4), np.log(S_MAX), N_S))
    V = build_velocity_matrix(z0, s_vals)

    e1 = hadamard_direction(z0)
    e2 = density_direction(z0, e1)

    U, S, _ = svd(V, full_matrices=False)
    cos1 = abs(np.dot(U[:, 0], e1))

    r2s = []
    for k in range(V.shape[1]):
        vk = V[:, k]
        nrm = np.linalg.norm(vk)
        if nrm > 1e-15:
            proj = np.dot(vk, e1) * e1 + np.dot(vk, e2) * e2
            r2s.append(1.0 - (np.linalg.norm(vk - proj) / nrm)**2)

    return {
        "cos1": float(cos1),
        "R2": float(np.mean(r2s)),
        "S1S2": float(S[0] / S[1]),
        "S2S3": float(S[1] / S[2]),
    }

def classify_geometry(out):
    strong = (
        out["cos1"] >= COS1_STRONG and
        out["R2"] >= R2_STRONG and
        out["S1S2"] >= S1S2_STRONG and
        out["S2S3"] >= S2S3_STRONG
    )

    borderline = (
        out["cos1"] >= COS1_BORDER and
        out["R2"] >= R2_BORDER and
        out["S1S2"] >= S1S2_BORDER and
        out["S2S3"] >= S2S3_BORDER
    )

    if strong:
        return "strong"
    if borderline:
        return "borderline"
    return "weak"


# =========================================================
# Gap barrier diagnostics
# =========================================================

def gap_barrier_diagnostics(z):
    g = np.diff(z)
    D = float(g.mean())

    order = np.argsort(g)
    k_min = int(order[0])
    k_second = int(order[1]) if len(order) > 1 else None

    gmin = float(g[k_min])
    gsecond = float(g[k_second]) if k_second is not None else np.inf

    eta = gmin / D
    a = gsecond / D
    rho = gmin / gsecond if np.isfinite(gsecond) else 0.0

    v = reduced_velocity(z)

    # actual minimum-gap derivative
    gprime = float(v[k_min + 1] - v[k_min])

    # pair term in exact gap equation
    pair_term = float(4.0 / gmin)

    # tail term is actual minus pair contribution
    tail_term = float(gprime - pair_term)

    # C_tail from g' = 4/g - C g
    C_tail = float((pair_term - gprime) / (gmin + 1e-30))

    # finite-model sufficient condition from corrected lemma
    predicts_reopen = bool(eta < a)

    actual_reopen = bool(gprime > 0.0)

    # margin in the lemma
    margin = float(a - eta)

    # direct barrier margin: 4/g - Cg
    # same as gprime, but useful as named output
    barrier_margin = gprime

    return {
        "D": D,
        "k_min": k_min,
        "k_second": k_second,
        "gminD": eta,
        "gsecondD": a,
        "rho": rho,
        "margin_eta_a": margin,
        "gprime": gprime,
        "pair_term": pair_term,
        "tail_term": tail_term,
        "tail_over_pair": abs(tail_term) / (abs(pair_term) + 1e-30),
        "C_tail": C_tail,
        "predicts_reopen": predicts_reopen,
        "actual_reopen": actual_reopen,
        "barrier_margin": barrier_margin,
    }


# =========================================================
# Window extraction
# =========================================================

def extract_window(pool, n, shift):
    center = len(pool) // 2 + shift
    start = center - n // 2
    end = start + n

    if start < 0 or end > len(pool):
        return None

    return pool[start:end].copy()


# =========================================================
# Reporting
# =========================================================

def summarize(rows):
    print()
    print("=" * 120)
    print("SUMMARY BY GEOMETRY TIER")
    print("-" * 120)

    hdr = (
        f"{'tier':>12} {'N':>5} "
        f"{'gminD_med':>10} {'g2D_med':>10} {'rho_med':>10} "
        f"{'gprime_med':>12} {'tail/pair_med':>14} "
        f"{'pred_ok':>8} {'actual_ok':>10}"
    )
    print(hdr)
    print("-" * len(hdr))

    for tier in ["strong", "borderline", "weak"]:
        sub = [r for r in rows if r["tier"] == tier]
        if not sub:
            continue

        def med(key):
            return float(np.median([r[key] for r in sub]))

        pred_ok = sum(1 for r in sub if r["predicts_reopen"])
        actual_ok = sum(1 for r in sub if r["actual_reopen"])

        print(
            f"{tier:>12s} {len(sub):5d} "
            f"{med('gminD'):10.4f} {med('gsecondD'):10.4f} {med('rho'):10.4f} "
            f"{med('gprime'):12.4f} {med('tail_over_pair'):14.4f} "
            f"{pred_ok:8d} {actual_ok:10d}"
        )

def print_weak(rows):
    print()
    print("=" * 120)
    print("WEAK WINDOWS — LOCAL-MINIMUM BARRIER DETAILS")
    print("-" * 120)

    hdr = (
        f"{'T':>6} {'n':>4} {'shift':>6} {'k':>4} "
        f"{'gminD':>8} {'g2D':>8} {'rho':>8} {'η<a':>5} "
        f"{'gprime':>10} {'pair':>10} {'tail':>10} {'tail/pair':>10} {'Ctail':>10}"
    )
    print(hdr)
    print("-" * len(hdr))

    for r in sorted([x for x in rows if x["tier"] == "weak"], key=lambda x: x["gminD"]):
        print(
            f"{r['T']:6d} {r['n']:4d} {r['shift']:6d} {r['k_min']:4d} "
            f"{r['gminD']:8.4f} {r['gsecondD']:8.4f} {r['rho']:8.4f} "
            f"{str(r['predicts_reopen']):>5s} "
            f"{r['gprime']:10.4f} {r['pair_term']:10.4f} {r['tail_term']:10.4f} "
            f"{r['tail_over_pair']:10.4f} {r['C_tail']:10.4f}"
        )

def prediction_confusion(rows):
    print()
    print("=" * 120)
    print("PREDICTION CHECK: η < a  VS  ACTUAL g_min' > 0")
    print("-" * 120)

    tp = fp = fn = tn = 0

    for r in rows:
        pred = r["predicts_reopen"]
        act = r["actual_reopen"]

        if pred and act:
            tp += 1
        elif pred and not act:
            fp += 1
        elif (not pred) and act:
            fn += 1
        else:
            tn += 1

    print(f"TP pred reopen / actual reopen     : {tp}")
    print(f"FP pred reopen / actual not reopen : {fp}")
    print(f"FN pred no / actual reopen         : {fn}")
    print(f"TN pred no / actual no             : {tn}")

    total = len(rows)
    acc = (tp + tn) / total if total else 0.0
    print(f"accuracy: {acc:.4f}")

def print_counterexamples(rows):
    bad = [r for r in rows if r["predicts_reopen"] and not r["actual_reopen"]]

    print()
    print("=" * 120)
    print("COUNTEREXAMPLES TO SUFFICIENT CONDITION")
    print("-" * 120)

    if not bad:
        print("None found.")
        return

    for r in bad:
        print(
            f"T={r['T']}, n={r['n']}, shift={r['shift']}, tier={r['tier']}, "
            f"gminD={r['gminD']:.4f}, g2D={r['gsecondD']:.4f}, "
            f"rho={r['rho']:.4f}, gprime={r['gprime']:.4f}"
        )


# =========================================================
# Main
# =========================================================

if __name__ == "__main__":
    print("local_minimum_barrier_check.py")
    print("=" * 120)
    print("Reduced flow: dt_i/ds = +2 sum_{j!=i} 1/(t_i-t_j)")
    print("Testing corrected finite-model condition: η < a")
    print()

    rows = []

    for T in T_VALUES:
        pool = find_zeros_pool(T, N_POOL)

        for n in WINDOW_SIZES:
            for shift in SHIFTS:
                z0 = extract_window(pool, n, shift)
                if z0 is None:
                    continue

                gd = geometry_diagnostics(z0)
                tier = classify_geometry(gd)
                bd = gap_barrier_diagnostics(z0)

                row = {
                    "T": T,
                    "n": n,
                    "shift": shift,
                    "tier": tier,
                    **gd,
                    **bd,
                }

                rows.append(row)

    print(f"Total windows scanned: {len(rows)}")

    summarize(rows)
    print_weak(rows)
    prediction_confusion(rows)
    print_counterexamples(rows)

    print()
    print("=" * 120)
    print("INTERPRETATION GUIDE")
    print("-" * 120)
    print("If every weak window satisfies η<a and gprime>0:")
    print("  -> weak small-gap windows are local barrier / reopening cases.")
    print("If η<a ever predicts reopen but actual gprime<=0:")
    print("  -> the finite-model sufficient condition was stated too strongly or implemented incorrectly.")
    print("If many windows reopen even without η<a:")
    print("  -> η<a is sufficient, not necessary.")
```

```python
# kernel_basic_diagnostics.py
# ============================================================
# Basic diagnostics for the de Bruijn-Newman / Riemann Xi kernel
#
# Tests:
#   1. Kernel positivity on u >= 0
#   2. Tail decay of Phi_lambda(u) = exp(lambda*u^2) Phi(u)
#   3. Log-curvature Q''_lambda(u)
#   4. Even moments M_k(lambda)
#   5. Small Hankel minors of the moment sequence
#
# This is diagnostic only. It does not prove real-rootedness.
# ============================================================

import mpmath as mp

mp.mp.dps = 80

# ------------------------------------------------------------
# Configuration
# ------------------------------------------------------------

LAMBDA_VALUES = [
    mp.mpf("0"),
    mp.mpf("1e-6"),
    mp.mpf("1e-5"),
    mp.mpf("1e-4"),
    mp.mpf("1e-3"),
    mp.mpf("1e-2"),
    mp.mpf("5e-2"),
    mp.mpf("1e-1"),
]

U_MIN = mp.mpf("0")
U_MAX = mp.mpf("6")
N_GRID = 241

# Moment settings
MAX_MOMENT_K = 12
HANKEL_MAX_R = 5

# Kernel series cutoff.
# The n-sum decays extremely fast for u >= 0 because of exp(-pi*n^2*exp(4u)).
# Near u=0, n up to ~20-40 is already usually excessive at this precision.
N_SUM = 80

# Numerical derivative step for log-curvature
DERIV_STEP = mp.mpf("1e-4")


# ------------------------------------------------------------
# Kernel convention
# ------------------------------------------------------------
#
# Use the standard-looking Newman/Riemann kernel form:
#
#   Phi(u) = sum_{n>=1} (2*pi^2*n^4*e^(9u) - 3*pi*n^2*e^(5u))
#                     * exp(-pi*n^2*e^(4u))
#
# extended evenly Phi(-u)=Phi(u).
#
# Different papers use scaled variants. This script is for internal
# structural diagnostics only; keep this convention fixed throughout.

def phi_pos_u(u: mp.mpf, n_sum: int = N_SUM) -> mp.mpf:
    """Phi(u) for u >= 0 under the fixed convention."""
    u = mp.mpf(u)
    total = mp.mpf("0")
    e4u = mp.e ** (4 * u)
    e5u = mp.e ** (5 * u)
    e9u = mp.e ** (9 * u)

    for n in range(1, n_sum + 1):
        n2 = mp.mpf(n) ** 2
        n4 = n2 ** 2
        term = (2 * mp.pi**2 * n4 * e9u - 3 * mp.pi * n2 * e5u)
        term *= mp.e ** (-mp.pi * n2 * e4u)
        total += term

    return total


def phi(u: mp.mpf, n_sum: int = N_SUM) -> mp.mpf:
    """Even extension."""
    return phi_pos_u(abs(mp.mpf(u)), n_sum=n_sum)


def phi_lambda(u: mp.mpf, lam: mp.mpf, n_sum: int = N_SUM) -> mp.mpf:
    u = mp.mpf(u)
    lam = mp.mpf(lam)
    return mp.e ** (lam * u * u) * phi(u, n_sum=n_sum)


def log_phi_lambda(u: mp.mpf, lam: mp.mpf) -> mp.mpf:
    val = phi_lambda(u, lam)
    if val <= 0:
        raise ValueError(f"Phi_lambda nonpositive at u={u}, lambda={lam}: {val}")
    return mp.log(val)


# ------------------------------------------------------------
# Numerical derivatives
# ------------------------------------------------------------

def second_derivative_log_phi_lambda(u: mp.mpf, lam: mp.mpf, h: mp.mpf = DERIV_STEP) -> mp.mpf:
    """
    Central finite-difference approximation to Q''_lambda(u).

    Avoid crossing u=0 using the even kernel by evaluating log_phi_lambda directly.
    """
    u = mp.mpf(u)
    h = mp.mpf(h)
    return (
        log_phi_lambda(u + h, lam)
        - 2 * log_phi_lambda(u, lam)
        + log_phi_lambda(u - h, lam)
    ) / (h * h)


# ------------------------------------------------------------
# Grid construction
# ------------------------------------------------------------

def make_grid(u_min=U_MIN, u_max=U_MAX, n_grid=N_GRID):
    if n_grid < 2:
        return [mp.mpf(u_min)]
    step = (mp.mpf(u_max) - mp.mpf(u_min)) / (n_grid - 1)
    return [mp.mpf(u_min) + i * step for i in range(n_grid)]


# ------------------------------------------------------------
# Moment calculation
# ------------------------------------------------------------

def moment_even(k: int, lam: mp.mpf, cutoff: mp.mpf = mp.mpf("8")) -> mp.mpf:
    """
    M_k(lambda) = integral_{-inf}^{inf} u^(2k) Phi_lambda(u) du
                = 2 integral_0^inf u^(2k) Phi_lambda(u) du.

    The kernel decays super-exponentially for large u, so finite cutoff
    with mp.quad is usually stable. We integrate over subintervals to
    help mpmath.
    """
    k = int(k)
    lam = mp.mpf(lam)

    def integrand(u):
        return (u ** (2 * k)) * phi_lambda(u, lam)

    # Piecewise intervals improve stability.
    intervals = [mp.mpf("0"), mp.mpf("0.25"), mp.mpf("0.5"), mp.mpf("1"),
                 mp.mpf("2"), mp.mpf("4"), cutoff]

    val = mp.quad(integrand, intervals)
    return 2 * val


def hankel_det(moments, r: int) -> mp.mpf:
    """
    det [M_{i+j}]_{i,j=0}^{r}
    Requires moments up to index 2r.
    """
    mat = mp.matrix([[moments[i + j] for j in range(r + 1)] for i in range(r + 1)])
    return mp.det(mat)


# ------------------------------------------------------------
# Diagnostics
# ------------------------------------------------------------

def kernel_grid_diagnostics(lam: mp.mpf):
    grid = make_grid()
    phi_vals = []
    phlam_vals = []
    q2_vals = []

    positivity_failures = []

    for u in grid:
        p = phi(u)
        pl = phi_lambda(u, lam)

        phi_vals.append(p)
        phlam_vals.append(pl)

        if p <= 0 or pl <= 0:
            positivity_failures.append((u, p, pl))
            q2_vals.append(mp.nan)
        else:
            try:
                q2_vals.append(second_derivative_log_phi_lambda(u, lam))
            except Exception:
                q2_vals.append(mp.nan)

    finite_q2 = [x for x in q2_vals if x == x]

    min_phi = min(phi_vals)
    min_philam = min(phlam_vals)
    max_q2 = max(finite_q2) if finite_q2 else mp.nan
    min_q2 = min(finite_q2) if finite_q2 else mp.nan
    q2_positive_count = sum(1 for x in finite_q2 if x > 0)

    return {
        "lambda": lam,
        "min_phi": min_phi,
        "min_phi_lambda": min_philam,
        "positivity_failures": positivity_failures,
        "max_Q2": max_q2,
        "min_Q2": min_q2,
        "Q2_positive_count": q2_positive_count,
        "Q2_total": len(finite_q2),
        "tail_phi_lambda_at_Umax": phlam_vals[-1],
        "grid": grid,
        "q2_vals": q2_vals,
    }


def moment_hankel_diagnostics(lam: mp.mpf):
    moments = []
    for k in range(MAX_MOMENT_K + 1):
        moments.append(moment_even(k, lam))

    hankels = []
    for r in range(1, HANKEL_MAX_R + 1):
        if 2 * r <= MAX_MOMENT_K:
            det = hankel_det(moments, r)
            hankels.append((r, det))
        else:
            hankels.append((r, None))

    return {
        "lambda": lam,
        "moments": moments,
        "hankels": hankels,
    }


def fmt(x, digits=10):
    if x is None:
        return "None"
    if x != x:
        return "nan"
    try:
        return mp.nstr(x, digits)
    except Exception:
        return str(x)


# ------------------------------------------------------------
# Main report
# ------------------------------------------------------------

def run_all():
    print("kernel_basic_diagnostics.py")
    print("=" * 100)
    print("Kernel convention:")
    print("Phi(u)=sum_n [2*pi^2*n^4*exp(9u)-3*pi*n^2*exp(5u)] exp(-pi*n^2 exp(4u))")
    print("even extension Phi(-u)=Phi(u)")
    print()
    print(f"mp.dps={mp.mp.dps}, N_SUM={N_SUM}, grid=[{U_MIN},{U_MAX}], N_GRID={N_GRID}")
    print(f"MAX_MOMENT_K={MAX_MOMENT_K}, HANKEL_MAX_R={HANKEL_MAX_R}")
    print()

    print("=" * 100)
    print("PART 1 — positivity, decay, log-curvature")
    print("-" * 100)
    header = (
        f"{'lambda':>12} "
        f"{'min Phi':>18} "
        f"{'min Phi_lam':>18} "
        f"{'Phi_lam(Umax)':>18} "
        f"{'max Q2':>18} "
        f"{'min Q2':>18} "
        f"{'Q2>0':>10}"
    )
    print(header)
    print("-" * len(header))

    grid_results = []
    for lam in LAMBDA_VALUES:
        res = kernel_grid_diagnostics(lam)
        grid_results.append(res)
        q2_pos = f"{res['Q2_positive_count']}/{res['Q2_total']}"
        print(
            f"{fmt(lam, 6):>12} "
            f"{fmt(res['min_phi'], 8):>18} "
            f"{fmt(res['min_phi_lambda'], 8):>18} "
            f"{fmt(res['tail_phi_lambda_at_Umax'], 8):>18} "
            f"{fmt(res['max_Q2'], 8):>18} "
            f"{fmt(res['min_Q2'], 8):>18} "
            f"{q2_pos:>10}"
        )

    print()
    print("Interpretation:")
    print("  - positivity failures would damage basic kernel routes.")
    print("  - Q2>0 means ordinary log-concavity fails at sampled points.")
    print("  - positive lambda adds +2lambda to Q2, so log-concavity should not improve.")
    print()

    print("=" * 100)
    print("PART 2 — moments and Hankel minors")
    print("-" * 100)

    for lam in LAMBDA_VALUES:
        print()
        print(f"lambda = {fmt(lam, 8)}")
        print("-" * 80)
        mh = moment_hankel_diagnostics(lam)

        print("Moments M_k = ∫ u^(2k) Phi_lambda(u) du over R")
        for k, m in enumerate(mh["moments"]):
            print(f"  M_{k:02d} = {fmt(m, 16)}")

        print("Hankel determinants det[M_{i+j}]_{i,j=0}^r")
        for r, det in mh["hankels"]:
            if det is None:
                print(f"  r={r}: not computed")
            else:
                sign = "positive" if det > 0 else ("negative" if det < 0 else "zero")
                print(f"  r={r}: det={fmt(det, 16)}  sign={sign}")

    print()
    print("=" * 100)
    print("SUMMARY GUIDE")
    print("-" * 100)
    print("If Q2 is positive anywhere, naive log-concavity is not available.")
    print("If low-order Hankel determinants are negative, simple moment-total-positivity routes fail.")
    print("If all these diagnostics pass, that is support only, not proof.")
    print("Next script after this: kernel_jensen_diagnostics.py")


if __name__ == "__main__":
    run_all()
```

```python
# kernel_jensen_diagnostics.py
# ============================================================
# Jensen polynomial diagnostics for the de Bruijn-Newman /
# Riemann Xi kernel.
#
# Uses the same kernel convention as kernel_basic_diagnostics.py.
#
# We compute moments:
#
#   M_k(lambda) = ∫_{-∞}^{∞} u^(2k) Phi_lambda(u) du
#
# For
#
#   H_lambda(t) = ∫ Phi_lambda(u) e^{itu} du
#
# we have
#
#   H_lambda(t) = Σ (-1)^k M_k t^(2k)/(2k)!
#
# Define the positive-coefficient auxiliary function
#
#   B_lambda(y) = H_lambda(i sqrt(y))
#               = Σ M_k y^k/(2k)!
#
# Write
#
#   B_lambda(y) = Σ A_k y^k/k!,
#
# where
#
#   A_k = M_k * k! / (2k)!.
#
# If H_lambda has only real zeros ±rho_j, then
#
#   B_lambda(y) = C Π_j (1 + y/rho_j^2),
#
# so B_lambda has only real nonpositive zeros.
#
# Therefore Jensen polynomials built from A_k should have only real
# nonpositive roots.
#
# This is diagnostic only. Passing finite tests is not proof.
# ============================================================

import mpmath as mp
from math import comb

mp.mp.dps = 80

# ------------------------------------------------------------
# Configuration
# ------------------------------------------------------------

LAMBDA_VALUES = [
    mp.mpf("0"),
    mp.mpf("1e-6"),
    mp.mpf("1e-5"),
    mp.mpf("1e-4"),
    mp.mpf("1e-3"),
    mp.mpf("1e-2"),
    mp.mpf("5e-2"),
    mp.mpf("1e-1"),
]

DEGREES = [2, 3, 4, 5, 6, 8, 10]
N_MAX = 20

# Need moments up to N_MAX + max(DEGREES)
MAX_MOMENT_K = N_MAX + max(DEGREES)

# Moment integration cutoff
INTEGRATION_CUTOFF = mp.mpf("8")

# Kernel n-sum cutoff
N_SUM = 80

# Root classification tolerance
ROOT_IMAG_TOL = mp.mpf("1e-30")
ROOT_POS_TOL = mp.mpf("1e-25")

# If polyroots struggles, increase maxsteps
POLYROOT_MAXSTEPS = 200
POLYROOT_ERROR = False


# ------------------------------------------------------------
# Kernel convention
# ------------------------------------------------------------

def phi_pos_u(u: mp.mpf, n_sum: int = N_SUM) -> mp.mpf:
    """Phi(u) for u >= 0 under fixed convention."""
    u = mp.mpf(u)
    total = mp.mpf("0")
    e4u = mp.e ** (4 * u)
    e5u = mp.e ** (5 * u)
    e9u = mp.e ** (9 * u)

    for n in range(1, n_sum + 1):
        n2 = mp.mpf(n) ** 2
        n4 = n2 ** 2
        term = (2 * mp.pi**2 * n4 * e9u - 3 * mp.pi * n2 * e5u)
        term *= mp.e ** (-mp.pi * n2 * e4u)
        total += term

    return total


def phi(u: mp.mpf, n_sum: int = N_SUM) -> mp.mpf:
    """Even extension."""
    return phi_pos_u(abs(mp.mpf(u)), n_sum=n_sum)


def phi_lambda(u: mp.mpf, lam: mp.mpf, n_sum: int = N_SUM) -> mp.mpf:
    u = mp.mpf(u)
    lam = mp.mpf(lam)
    return mp.e ** (lam * u * u) * phi(u, n_sum=n_sum)


# ------------------------------------------------------------
# Moments and coefficient sequence
# ------------------------------------------------------------

def moment_even(k: int, lam: mp.mpf, cutoff: mp.mpf = INTEGRATION_CUTOFF) -> mp.mpf:
    """
    M_k(lambda) = ∫_{-∞}^{∞} u^(2k) Phi_lambda(u) du
                = 2∫_0^∞ u^(2k) Phi_lambda(u) du.
    """
    k = int(k)
    lam = mp.mpf(lam)

    def integrand(u):
        return (u ** (2 * k)) * phi_lambda(u, lam)

    intervals = [
        mp.mpf("0"),
        mp.mpf("0.25"),
        mp.mpf("0.5"),
        mp.mpf("1"),
        mp.mpf("2"),
        mp.mpf("4"),
        cutoff,
    ]

    return 2 * mp.quad(integrand, intervals)


def compute_moments(lam: mp.mpf, max_k: int = MAX_MOMENT_K):
    return [moment_even(k, lam) for k in range(max_k + 1)]


def compute_A_sequence(moments):
    """
    A_k = M_k * k! / (2k)! so that

        B_lambda(y) = Σ A_k y^k/k!
    """
    A = []
    for k, M in enumerate(moments):
        Ak = M * mp.factorial(k) / mp.factorial(2 * k)
        A.append(Ak)
    return A


# ------------------------------------------------------------
# Jensen polynomial construction
# ------------------------------------------------------------

def jensen_coeffs(A, n: int, d: int):
    """
    J_{d,n}(X) = Σ_{k=0}^{d} binom(d,k) A_{n+k} X^k.

    Return coefficients in ascending order.
    """
    coeffs = []
    for k in range(d + 1):
        coeffs.append(mp.mpf(comb(d, k)) * A[n + k])
    return coeffs


def scale_coeffs(coeffs):
    max_abs = max(abs(c) for c in coeffs)
    if max_abs == 0:
        return coeffs
    return [c / max_abs for c in coeffs]


def roots_of_poly_ascending(coeffs_asc):
    """
    mpmath.polyroots expects descending coefficients.
    """
    coeffs_scaled = scale_coeffs(coeffs_asc)
    coeffs_desc = list(reversed(coeffs_scaled))

    roots = mp.polyroots(
        coeffs_desc,
        maxsteps=POLYROOT_MAXSTEPS,
        error=POLYROOT_ERROR,
    )
    return roots


def eval_poly_ascending(coeffs, x):
    total = mp.mpf("0")
    power = mp.mpf("1")
    for c in coeffs:
        total += c * power
        power *= x
    return total


def classify_roots(roots):
    """
    For B_lambda Jensen polynomials, expected roots:
        real and <= 0.
    """
    max_imag = max(abs(mp.im(r)) for r in roots) if roots else mp.mpf("0")
    max_real = max(mp.re(r) for r in roots) if roots else mp.mpf("-inf")

    nonreal_count = sum(1 for r in roots if abs(mp.im(r)) > ROOT_IMAG_TOL)
    positive_real_count = sum(
        1
        for r in roots
        if abs(mp.im(r)) <= ROOT_IMAG_TOL and mp.re(r) > ROOT_POS_TOL
    )

    all_real_nonpositive = (nonreal_count == 0 and positive_real_count == 0)

    return {
        "max_imag": max_imag,
        "max_real": max_real,
        "nonreal_count": nonreal_count,
        "positive_real_count": positive_real_count,
        "all_real_nonpositive": all_real_nonpositive,
    }


def root_residuals(coeffs_asc, roots):
    vals = []
    denom = sum(abs(c) for c in coeffs_asc)
    if denom == 0:
        denom = mp.mpf("1")
    for r in roots:
        vals.append(abs(eval_poly_ascending(coeffs_asc, r)) / denom)
    return vals


# ------------------------------------------------------------
# Turan / log-concavity checks for A sequence
# ------------------------------------------------------------

def turan_checks(A, max_n: int):
    """
    For positive sequence A_n associated with real negative zeros,
    low-level Turan/log-concavity diagnostics include:

        A_n^2 - A_{n-1} A_{n+1} >= 0

    for n>=1.
    """
    out = []
    for n in range(1, max_n):
        val = A[n] ** 2 - A[n - 1] * A[n + 1]
        out.append((n, val))
    return out


# ------------------------------------------------------------
# Main diagnostics
# ------------------------------------------------------------

def fmt(x, digits=10):
    try:
        return mp.nstr(x, digits)
    except Exception:
        return str(x)


def run_for_lambda(lam: mp.mpf):
    moments = compute_moments(lam)
    A = compute_A_sequence(moments)

    # Turan checks
    turans = turan_checks(A, N_MAX + max(DEGREES) - 1)
    turan_failures = [(n, v) for n, v in turans if v < 0]
    min_turan = min(v for _, v in turans)

    records = []

    for d in DEGREES:
        for n in range(0, N_MAX + 1):
            if n + d >= len(A):
                continue

            coeffs = jensen_coeffs(A, n=n, d=d)

            try:
                roots = roots_of_poly_ascending(coeffs)
                cls = classify_roots(roots)
                residuals = root_residuals(coeffs, roots)
                max_resid = max(residuals) if residuals else mp.mpf("0")

                record = {
                    "lambda": lam,
                    "degree": d,
                    "n": n,
                    "ok": cls["all_real_nonpositive"],
                    "nonreal_count": cls["nonreal_count"],
                    "positive_real_count": cls["positive_real_count"],
                    "max_imag": cls["max_imag"],
                    "max_real": cls["max_real"],
                    "max_resid": max_resid,
                    "roots": roots,
                }

            except Exception as e:
                record = {
                    "lambda": lam,
                    "degree": d,
                    "n": n,
                    "ok": False,
                    "nonreal_count": None,
                    "positive_real_count": None,
                    "max_imag": mp.inf,
                    "max_real": mp.inf,
                    "max_resid": mp.inf,
                    "roots": [],
                    "error": str(e),
                }

            records.append(record)

    failures = [r for r in records if not r["ok"]]

    worst_imag = max(records, key=lambda r: r["max_imag"])
    worst_real = max(records, key=lambda r: r["max_real"])
    worst_resid = max(records, key=lambda r: r["max_resid"])

    return {
        "lambda": lam,
        "moments": moments,
        "A": A,
        "min_turan": min_turan,
        "turan_failures": turan_failures,
        "records": records,
        "failures": failures,
        "worst_imag": worst_imag,
        "worst_real": worst_real,
        "worst_resid": worst_resid,
    }


def print_record_short(r):
    print(
        f"lambda={fmt(r['lambda'], 8):>10} "
        f"d={r['degree']:>2} "
        f"n={r['n']:>3} "
        f"ok={str(r['ok']):>5} "
        f"nonreal={str(r['nonreal_count']):>4} "
        f"posreal={str(r['positive_real_count']):>4} "
        f"max_imag={fmt(r['max_imag'], 8):>14} "
        f"max_real={fmt(r['max_real'], 8):>14} "
        f"max_resid={fmt(r['max_resid'], 8):>14}"
    )


def run_all():
    print("kernel_jensen_diagnostics.py")
    print("=" * 110)
    print("Kernel convention matches kernel_basic_diagnostics.py")
    print("Auxiliary function: B_lambda(y)=H_lambda(i*sqrt(y))")
    print("Coefficient sequence: A_k = M_k*k!/(2k)!")
    print("Expected Jensen roots for real-zero H_lambda: real and nonpositive")
    print()
    print(f"mp.dps={mp.mp.dps}, N_SUM={N_SUM}, cutoff={INTEGRATION_CUTOFF}")
    print(f"degrees={DEGREES}, N_MAX={N_MAX}, MAX_MOMENT_K={MAX_MOMENT_K}")
    print()

    all_results = []

    print("=" * 110)
    print("SUMMARY BY LAMBDA")
    print("-" * 110)
    header = (
        f"{'lambda':>12} "
        f"{'tests':>8} "
        f"{'fail':>8} "
        f"{'min Turan':>18} "
        f"{'Turan fail':>12} "
        f"{'worst imag':>18} "
        f"{'worst real':>18} "
        f"{'worst resid':>18}"
    )
    print(header)
    print("-" * len(header))

    for lam in LAMBDA_VALUES:
        res = run_for_lambda(lam)
        all_results.append(res)

        records = res["records"]
        failures = res["failures"]
        tur_fail = len(res["turan_failures"])

        print(
            f"{fmt(lam, 8):>12} "
            f"{len(records):>8} "
            f"{len(failures):>8} "
            f"{fmt(res['min_turan'], 8):>18} "
            f"{tur_fail:>12} "
            f"{fmt(res['worst_imag']['max_imag'], 8):>18} "
            f"{fmt(res['worst_real']['max_real'], 8):>18} "
            f"{fmt(res['worst_resid']['max_resid'], 8):>18}"
        )

    print()
    print("=" * 110)
    print("FAILURE DETAILS")
    print("-" * 110)

    any_fail = False
    for res in all_results:
        failures = res["failures"]
        if failures:
            any_fail = True
            print()
            print(f"lambda={fmt(res['lambda'], 8)} failures: {len(failures)}")
            for r in failures[:20]:
                print_record_short(r)
                if "error" in r:
                    print(f"  error: {r['error']}")
            if len(failures) > 20:
                print(f"  ... {len(failures)-20} more failures not printed")

    if not any_fail:
        print("No Jensen root failures under configured tolerance.")

    print()
    print("=" * 110)
    print("WORST CASES")
    print("-" * 110)
    for res in all_results:
        print()
        print(f"lambda={fmt(res['lambda'], 8)}")
        print("Worst imaginary part:")
        print_record_short(res["worst_imag"])
        print("Worst real part:")
        print_record_short(res["worst_real"])
        print("Worst residual:")
        print_record_short(res["worst_resid"])

    print()
    print("=" * 110)
    print("INTERPRETATION GUIDE")
    print("-" * 110)
    print("PASS:")
    print("  All tested Jensen polynomials have roots real and nonpositive.")
    print("  This supports, but does not prove, Laguerre-Pólya behaviour.")
    print()
    print("FAIL:")
    print("  Any robust nonreal or positive root for lambda>0 would damage")
    print("  the kernel route for that lambda, subject to precision verification.")
    print()
    print("Next script if this passes:")
    print("  kernel_total_positivity_minors.py")


if __name__ == "__main__":
    run_all()
```

```python
# kernel_jensen_diagnostics_v2.py
# ============================================================
# More stable Jensen polynomial diagnostics.
#
# Fixes v1 issue:
#   mpmath.polyroots often fails to converge on badly scaled positive
#   coefficient polynomials.
#
# v2:
#   - keeps high-precision moment calculation
#   - rescales Jensen coefficients
#   - uses numpy.roots as the primary finite-degree diagnostic backend
#   - separates numerical root failures from mathematical failures
#   - checks residuals directly
#
# Diagnostic only. Passing finite tests is not proof.
# ============================================================

import mpmath as mp
import numpy as np
from math import comb

mp.mp.dps = 80

# ------------------------------------------------------------
# Configuration
# ------------------------------------------------------------

LAMBDA_VALUES = [
    mp.mpf("0"),
    mp.mpf("1e-6"),
    mp.mpf("1e-5"),
    mp.mpf("1e-4"),
    mp.mpf("1e-3"),
    mp.mpf("1e-2"),
    mp.mpf("5e-2"),
    mp.mpf("1e-1"),
]

DEGREES = [2, 3, 4, 5, 6, 8, 10]
N_MAX = 20
MAX_MOMENT_K = N_MAX + max(DEGREES)

N_SUM = 80
INTEGRATION_CUTOFF = mp.mpf("8")

IMAG_TOL = 1e-7
POS_TOL = 1e-8
RESID_TOL = 1e-6


# ------------------------------------------------------------
# Kernel
# ------------------------------------------------------------

def phi_pos_u(u, n_sum=N_SUM):
    u = mp.mpf(u)
    total = mp.mpf("0")
    e4u = mp.e ** (4 * u)
    e5u = mp.e ** (5 * u)
    e9u = mp.e ** (9 * u)

    for n in range(1, n_sum + 1):
        n2 = mp.mpf(n) ** 2
        n4 = n2 ** 2
        term = (2 * mp.pi**2 * n4 * e9u - 3 * mp.pi * n2 * e5u)
        term *= mp.e ** (-mp.pi * n2 * e4u)
        total += term

    return total


def phi(u, n_sum=N_SUM):
    return phi_pos_u(abs(mp.mpf(u)), n_sum=n_sum)


def phi_lambda(u, lam, n_sum=N_SUM):
    u = mp.mpf(u)
    lam = mp.mpf(lam)
    return mp.e ** (lam * u * u) * phi(u, n_sum=n_sum)


# ------------------------------------------------------------
# Moments
# ------------------------------------------------------------

def moment_even(k, lam, cutoff=INTEGRATION_CUTOFF):
    k = int(k)
    lam = mp.mpf(lam)

    def integrand(u):
        return (u ** (2 * k)) * phi_lambda(u, lam)

    intervals = [
        mp.mpf("0"),
        mp.mpf("0.25"),
        mp.mpf("0.5"),
        mp.mpf("1"),
        mp.mpf("2"),
        mp.mpf("4"),
        cutoff,
    ]

    return 2 * mp.quad(integrand, intervals)


def compute_moments(lam):
    return [moment_even(k, lam) for k in range(MAX_MOMENT_K + 1)]


def compute_A_sequence(moments):
    # Bλ(y)=Σ A_k y^k/k!, with A_k=M_k k!/(2k)!
    A = []
    for k, M in enumerate(moments):
        A.append(M * mp.factorial(k) / mp.factorial(2 * k))
    return A


# ------------------------------------------------------------
# Jensen polynomials
# ------------------------------------------------------------

def jensen_coeffs_ascending(A, n, d):
    return [mp.mpf(comb(d, k)) * A[n + k] for k in range(d + 1)]


def normalise_coeffs_float(coeffs_asc):
    """
    Convert to descending float coefficients for numpy.roots.

    Uses max-absolute coefficient scaling.
    """
    coeffs = np.array([float(c) for c in coeffs_asc], dtype=np.float64)

    scale = np.max(np.abs(coeffs))
    if scale == 0 or not np.isfinite(scale):
        raise ValueError("bad coefficient scale")

    coeffs = coeffs / scale

    # numpy.roots expects descending order.
    return coeffs[::-1]


def eval_poly_asc_mp(coeffs_asc, z):
    z = mp.mpc(z)
    total = mp.mpc("0")
    power = mp.mpc("1")
    for c in coeffs_asc:
        total += c * power
        power *= z
    return total


def residual_mp(coeffs_asc, root):
    denom = mp.fsum([abs(c) for c in coeffs_asc])
    if denom == 0:
        denom = mp.mpf("1")
    return abs(eval_poly_asc_mp(coeffs_asc, root)) / denom


def analyse_roots(coeffs_asc):
    coeffs_desc = normalise_coeffs_float(coeffs_asc)
    roots = np.roots(coeffs_desc)

    max_imag = float(np.max(np.abs(np.imag(roots)))) if len(roots) else 0.0
    max_real = float(np.max(np.real(roots))) if len(roots) else -np.inf

    nonreal = int(np.sum(np.abs(np.imag(roots)) > IMAG_TOL))
    positive_real = int(
        np.sum((np.abs(np.imag(roots)) <= IMAG_TOL) & (np.real(roots) > POS_TOL))
    )

    residuals = []
    for r in roots:
        rr = mp.mpc(float(np.real(r)), float(np.imag(r)))
        residuals.append(residual_mp(coeffs_asc, rr))

    max_resid = max(residuals) if residuals else mp.mpf("0")

    ok = (
        nonreal == 0
        and positive_real == 0
        and max_resid < RESID_TOL
    )

    return {
        "ok": ok,
        "roots": roots,
        "nonreal": nonreal,
        "positive_real": positive_real,
        "max_imag": max_imag,
        "max_real": max_real,
        "max_resid": max_resid,
    }


# ------------------------------------------------------------
# Turán checks
# ------------------------------------------------------------

def turan_checks(A):
    out = []
    for n in range(1, len(A) - 1):
        val = A[n] ** 2 - A[n - 1] * A[n + 1]
        out.append((n, val))
    return out


# ------------------------------------------------------------
# Main
# ------------------------------------------------------------

def fmt(x, digits=10):
    try:
        return mp.nstr(x, digits)
    except Exception:
        return str(x)


def run_for_lambda(lam):
    moments = compute_moments(lam)
    A = compute_A_sequence(moments)

    turans = turan_checks(A)
    turan_failures = [(n, v) for n, v in turans if v < 0]
    min_turan = min(v for _, v in turans)

    records = []
    root_errors = []

    for d in DEGREES:
        for n in range(N_MAX + 1):
            coeffs = jensen_coeffs_ascending(A, n, d)

            try:
                ana = analyse_roots(coeffs)
                records.append({
                    "lambda": lam,
                    "degree": d,
                    "n": n,
                    **ana,
                })
            except Exception as e:
                root_errors.append({
                    "lambda": lam,
                    "degree": d,
                    "n": n,
                    "error": str(e),
                })

    math_failures = [r for r in records if not r["ok"]]

    return {
        "lambda": lam,
        "records": records,
        "root_errors": root_errors,
        "math_failures": math_failures,
        "min_turan": min_turan,
        "turan_failures": turan_failures,
    }


def print_record(r):
    print(
        f"λ={fmt(r['lambda'],8):>10} "
        f"d={r['degree']:>2} "
        f"n={r['n']:>3} "
        f"ok={str(r['ok']):>5} "
        f"nonreal={r['nonreal']:>3} "
        f"posreal={r['positive_real']:>3} "
        f"max_imag={r['max_imag']:.3e} "
        f"max_real={r['max_real']:.3e} "
        f"max_resid={fmt(r['max_resid'],8)}"
    )


def run_all():
    print("kernel_jensen_diagnostics_v2.py")
    print("=" * 110)
    print("Root backend: numpy.roots on scaled coefficients")
    print("Mathematical failure is separate from root backend failure")
    print()
    print(f"mp.dps={mp.mp.dps}, degrees={DEGREES}, N_MAX={N_MAX}")
    print(f"IMAG_TOL={IMAG_TOL}, POS_TOL={POS_TOL}, RESID_TOL={RESID_TOL}")
    print()

    results = []

    print("=" * 110)
    print("SUMMARY BY LAMBDA")
    print("-" * 110)

    header = (
        f"{'lambda':>12} "
        f"{'tests':>8} "
        f"{'math_fail':>10} "
        f"{'root_err':>9} "
        f"{'min Turan':>18} "
        f"{'Turan fail':>12} "
        f"{'worst imag':>12} "
        f"{'worst real':>12} "
        f"{'worst resid':>14}"
    )
    print(header)
    print("-" * len(header))

    for lam in LAMBDA_VALUES:
        res = run_for_lambda(lam)
        results.append(res)

        records = res["records"]
        worst_imag = max([r["max_imag"] for r in records], default=float("nan"))
        worst_real = max([r["max_real"] for r in records], default=float("nan"))
        worst_resid = max([r["max_resid"] for r in records], default=mp.mpf("nan"))

        print(
            f"{fmt(lam,8):>12} "
            f"{len(records):>8} "
            f"{len(res['math_failures']):>10} "
            f"{len(res['root_errors']):>9} "
            f"{fmt(res['min_turan'],8):>18} "
            f"{len(res['turan_failures']):>12} "
            f"{worst_imag:>12.3e} "
            f"{worst_real:>12.3e} "
            f"{fmt(worst_resid,8):>14}"
        )

    print()
    print("=" * 110)
    print("MATHEMATICAL FAILURE DETAILS")
    print("-" * 110)

    any_math_fail = False
    for res in results:
        fails = res["math_failures"]
        if fails:
            any_math_fail = True
            print()
            print(f"λ={fmt(res['lambda'],8)} math failures: {len(fails)}")
            for r in fails[:30]:
                print_record(r)
            if len(fails) > 30:
                print(f"  ... {len(fails)-30} more not printed")

    if not any_math_fail:
        print("No mathematical Jensen failures under configured tolerances.")

    print()
    print("=" * 110)
    print("ROOT BACKEND ERRORS")
    print("-" * 110)

    any_root_err = False
    for res in results:
        errs = res["root_errors"]
        if errs:
            any_root_err = True
            print()
            print(f"λ={fmt(res['lambda'],8)} root backend errors: {len(errs)}")
            for e in errs[:20]:
                print(
                    f"λ={fmt(e['lambda'],8):>10} "
                    f"d={e['degree']:>2} "
                    f"n={e['n']:>3} "
                    f"error={e['error']}"
                )
            if len(errs) > 20:
                print(f"  ... {len(errs)-20} more not printed")

    if not any_root_err:
        print("No root backend errors.")

    print()
    print("=" * 110)
    print("INTERPRETATION")
    print("-" * 110)
    print("If math_fail=0 and root_err=0, finite Jensen tests passed.")
    print("If failures have large residuals, treat as numerical conditioning first.")
    print("If robust nonreal/positive roots remain with small residuals, the kernel route is damaged.")
    print("Passing finite tests is not proof.")


if __name__ == "__main__":
    run_all()
```

```python
# kernel_total_positivity_minors.py
# ============================================================
# Sampled total-positivity / PF-style minor diagnostics for the
# de Bruijn-Newman / Riemann Xi kernel.
#
# Uses same kernel convention as previous scripts:
#
#   Phi(u)=Σ [2π²n⁴e^(9u)-3πn²e^(5u)] exp(-πn²e^(4u))
#   Phi_lambda(u)=exp(lambda*u^2) Phi(u), even extension.
#
# Tests:
#   1. Translation-kernel minors det[Phi_lambda(x_i-y_j)]
#   2. Moment Hankel minors at higher order
#   3. Coefficient Toeplitz-style PF minors for A_k
#
# Important:
#   These are diagnostic only.
#   Low-order sampled total positivity is not proof.
#   A robust negative minor can rule out a simple TP/PF route.
# ============================================================

import itertools
import mpmath as mp
from math import comb

mp.mp.dps = 80

# ------------------------------------------------------------
# Configuration
# ------------------------------------------------------------

LAMBDA_VALUES = [
    mp.mpf("0"),
    mp.mpf("1e-6"),
    mp.mpf("1e-5"),
    mp.mpf("1e-4"),
    mp.mpf("1e-3"),
    mp.mpf("1e-2"),
    mp.mpf("5e-2"),
    mp.mpf("1e-1"),
]

N_SUM = 80

# Translation-kernel sampled grids.
# Keep modest: determinant tests get expensive quickly.
X_GRID = [mp.mpf(v) for v in ["0.0", "0.15", "0.35", "0.60", "0.90", "1.25", "1.65"]]
Y_GRID = [mp.mpf(v) for v in ["0.0", "0.10", "0.30", "0.55", "0.85", "1.20", "1.60"]]

TP_ORDERS = [2, 3, 4, 5]

# Moment/Hankel settings
MAX_MOMENT_K = 18
HANKEL_MAX_R = 8
INTEGRATION_CUTOFF = mp.mpf("8")

# Toeplitz/PF-style settings
A_MAX_K = 30
TOEPLITZ_ORDERS = [2, 3, 4, 5]
TOEPLITZ_MAX_START = 16

# Classification thresholds
NEG_TOL = mp.mpf("1e-50")
NEAR_ZERO_TOL = mp.mpf("1e-40")


# ------------------------------------------------------------
# Kernel
# ------------------------------------------------------------

def phi_pos_u(u, n_sum=N_SUM):
    u = mp.mpf(u)
    total = mp.mpf("0")
    e4u = mp.e ** (4 * u)
    e5u = mp.e ** (5 * u)
    e9u = mp.e ** (9 * u)

    for n in range(1, n_sum + 1):
        n2 = mp.mpf(n) ** 2
        n4 = n2 ** 2
        term = (2 * mp.pi**2 * n4 * e9u - 3 * mp.pi * n2 * e5u)
        term *= mp.e ** (-mp.pi * n2 * e4u)
        total += term

    return total


def phi(u, n_sum=N_SUM):
    return phi_pos_u(abs(mp.mpf(u)), n_sum=n_sum)


def phi_lambda(u, lam, n_sum=N_SUM):
    u = mp.mpf(u)
    lam = mp.mpf(lam)
    return mp.e ** (lam * u * u) * phi(u, n_sum=n_sum)


# ------------------------------------------------------------
# Moments / A-sequence
# ------------------------------------------------------------

def moment_even(k, lam, cutoff=INTEGRATION_CUTOFF):
    k = int(k)
    lam = mp.mpf(lam)

    def integrand(u):
        return (u ** (2 * k)) * phi_lambda(u, lam)

    intervals = [
        mp.mpf("0"),
        mp.mpf("0.25"),
        mp.mpf("0.5"),
        mp.mpf("1"),
        mp.mpf("2"),
        mp.mpf("4"),
        cutoff,
    ]

    return 2 * mp.quad(integrand, intervals)


def compute_moments(lam, max_k):
    return [moment_even(k, lam) for k in range(max_k + 1)]


def compute_A_sequence(moments):
    # Bλ(y)=Σ A_k y^k/k!, A_k=M_k*k!/(2k)!
    A = []
    for k, M in enumerate(moments):
        A.append(M * mp.factorial(k) / mp.factorial(2 * k))
    return A


# ------------------------------------------------------------
# Determinants
# ------------------------------------------------------------

def det_matrix(mat):
    return mp.det(mp.matrix(mat))


def classify_det(det):
    if det < -NEG_TOL:
        return "negative"
    if abs(det) <= NEAR_ZERO_TOL:
        return "near_zero"
    if det < 0:
        return "tiny_negative"
    return "positive"


# ------------------------------------------------------------
# Test 1: sampled translation-kernel minors
# ------------------------------------------------------------

def translation_minor(lam, xs, ys):
    mat = []
    for x in xs:
        row = []
        for y in ys:
            row.append(phi_lambda(x - y, lam))
        mat.append(row)
    return det_matrix(mat)


def translation_tp_scan(lam):
    records = []

    for r in TP_ORDERS:
        for xs in itertools.combinations(X_GRID, r):
            for ys in itertools.combinations(Y_GRID, r):
                det = translation_minor(lam, xs, ys)
                records.append({
                    "order": r,
                    "xs": xs,
                    "ys": ys,
                    "det": det,
                    "class": classify_det(det),
                })

    return records


# ------------------------------------------------------------
# Test 2: Hankel moment minors
# ------------------------------------------------------------

def hankel_det(moments, r):
    mat = [[moments[i + j] for j in range(r + 1)] for i in range(r + 1)]
    return det_matrix(mat)


def hankel_scan(lam):
    moments = compute_moments(lam, MAX_MOMENT_K)
    records = []

    for r in range(1, HANKEL_MAX_R + 1):
        if 2 * r <= MAX_MOMENT_K:
            det = hankel_det(moments, r)
            records.append({
                "order": r + 1,
                "r": r,
                "det": det,
                "class": classify_det(det),
            })

    return records, moments


# ------------------------------------------------------------
# Test 3: Toeplitz/PF-style minors for A_k
# ------------------------------------------------------------
#
# This is only a rough diagnostic. PF∞ is normally formulated for a
# two-sided sequence with Toeplitz matrix [a_{j-i}]. Here we use:
#
#   a_k = A_k for k>=0, a_k=0 for k<0
#
# and test contiguous minors.
#
# Negative robust minors would damage a simple PF route.
# Positive sampled minors do not prove PF∞.

def toeplitz_value(A, idx):
    if idx < 0 or idx >= len(A):
        return mp.mpf("0")
    return A[idx]


def toeplitz_minor(A, row_indices, col_indices):
    mat = []
    for i in row_indices:
        row = []
        for j in col_indices:
            row.append(toeplitz_value(A, j - i))
        mat.append(row)
    return det_matrix(mat)


def toeplitz_pf_scan(lam):
    moments = compute_moments(lam, A_MAX_K)
    A = compute_A_sequence(moments)

    records = []

    for r in TOEPLITZ_ORDERS:
        # Sample contiguous row/col index blocks.
        # This is not exhaustive but catches simple PF violations.
        for row_start in range(0, TOEPLITZ_MAX_START + 1):
            for col_start in range(0, TOEPLITZ_MAX_START + 1):
                rows = tuple(range(row_start, row_start + r))
                cols = tuple(range(col_start, col_start + r))
                det = toeplitz_minor(A, rows, cols)
                records.append({
                    "order": r,
                    "rows": rows,
                    "cols": cols,
                    "det": det,
                    "class": classify_det(det),
                })

    return records, A


# ------------------------------------------------------------
# Summaries
# ------------------------------------------------------------

def fmt(x, digits=10):
    try:
        return mp.nstr(x, digits)
    except Exception:
        return str(x)


def summarise_records(records):
    counts = {
        "positive": 0,
        "near_zero": 0,
        "tiny_negative": 0,
        "negative": 0,
    }

    for r in records:
        counts[r["class"]] += 1

    min_rec = min(records, key=lambda r: r["det"]) if records else None
    max_rec = max(records, key=lambda r: r["det"]) if records else None

    return counts, min_rec, max_rec


def print_min_record(label, rec):
    if rec is None:
        print(f"{label}: none")
        return

    print(f"{label}:")
    print(f"  class={rec['class']} det={fmt(rec['det'], 16)}")

    if "xs" in rec:
        print(f"  order={rec['order']} xs={[fmt(x,5) for x in rec['xs']]} ys={[fmt(y,5) for y in rec['ys']]}")
    elif "rows" in rec:
        print(f"  order={rec['order']} rows={rec['rows']} cols={rec['cols']}")
    else:
        print(f"  order={rec.get('order')} r={rec.get('r')}")


# ------------------------------------------------------------
# Main
# ------------------------------------------------------------

def run_all():
    print("kernel_total_positivity_minors.py")
    print("=" * 120)
    print("Diagnostic sampled TP/PF minor tests")
    print("Kernel convention matches previous scripts")
    print()
    print(f"mp.dps={mp.mp.dps}, N_SUM={N_SUM}")
    print(f"translation grid sizes: X={len(X_GRID)}, Y={len(Y_GRID)}, orders={TP_ORDERS}")
    print(f"Hankel: MAX_MOMENT_K={MAX_MOMENT_K}, HANKEL_MAX_R={HANKEL_MAX_R}")
    print(f"Toeplitz: A_MAX_K={A_MAX_K}, orders={TOEPLITZ_ORDERS}, max_start={TOEPLITZ_MAX_START}")
    print(f"NEG_TOL={fmt(NEG_TOL)}, NEAR_ZERO_TOL={fmt(NEAR_ZERO_TOL)}")
    print()

    all_results = []

    print("=" * 120)
    print("SUMMARY BY LAMBDA")
    print("-" * 120)
    header = (
        f"{'lambda':>12} "
        f"{'trans N':>8} {'trans neg':>10} {'trans tiny':>11} {'trans near0':>12} "
        f"{'hankel N':>8} {'hankel neg':>10} "
        f"{'toepl N':>8} {'toepl neg':>10} {'toepl tiny':>11} {'toepl near0':>12}"
    )
    print(header)
    print("-" * len(header))

    for lam in LAMBDA_VALUES:
        trans = translation_tp_scan(lam)
        hankel, moments = hankel_scan(lam)
        toepl, A = toeplitz_pf_scan(lam)

        trans_counts, trans_min, trans_max = summarise_records(trans)
        hank_counts, hank_min, hank_max = summarise_records(hankel)
        toep_counts, toep_min, toep_max = summarise_records(toepl)

        all_results.append({
            "lambda": lam,
            "trans": trans,
            "hankel": hankel,
            "toepl": toepl,
            "trans_counts": trans_counts,
            "hank_counts": hank_counts,
            "toep_counts": toep_counts,
            "trans_min": trans_min,
            "hank_min": hank_min,
            "toep_min": toep_min,
        })

        print(
            f"{fmt(lam,8):>12} "
            f"{len(trans):>8} {trans_counts['negative']:>10} {trans_counts['tiny_negative']:>11} {trans_counts['near_zero']:>12} "
            f"{len(hankel):>8} {hank_counts['negative']:>10} "
            f"{len(toepl):>8} {toep_counts['negative']:>10} {toep_counts['tiny_negative']:>11} {toep_counts['near_zero']:>12}"
        )

    print()
    print("=" * 120)
    print("MINIMUM DETERMINANT DETAILS")
    print("-" * 120)

    for res in all_results:
        print()
        print(f"lambda={fmt(res['lambda'],8)}")
        print_min_record("translation min", res["trans_min"])
        print_min_record("hankel min", res["hank_min"])
        print_min_record("toeplitz min", res["toep_min"])

    print()
    print("=" * 120)
    print("ROBUST NEGATIVE DETAILS")
    print("-" * 120)

    any_robust_neg = False

    for res in all_results:
        lam = res["lambda"]

        trans_neg = [r for r in res["trans"] if r["class"] == "negative"]
        hank_neg = [r for r in res["hankel"] if r["class"] == "negative"]
        toepl_neg = [r for r in res["toepl"] if r["class"] == "negative"]

        if trans_neg or hank_neg or toepl_neg:
            any_robust_neg = True
            print()
            print(f"lambda={fmt(lam,8)}")
            print(f"  translation robust negatives: {len(trans_neg)}")
            for r in trans_neg[:10]:
                print_min_record("  translation negative", r)
            print(f"  hankel robust negatives: {len(hank_neg)}")
            for r in hank_neg[:10]:
                print_min_record("  hankel negative", r)
            print(f"  toeplitz robust negatives: {len(toepl_neg)}")
            for r in toepl_neg[:10]:
                print_min_record("  toeplitz negative", r)

    if not any_robust_neg:
        print("No robust negative minors under configured thresholds.")

    print()
    print("=" * 120)
    print("INTERPRETATION GUIDE")
    print("-" * 120)
    print("translation minors:")
    print("  Negative det[Phi_lambda(x_i-y_j)] rules out simple translation-kernel total positivity.")
    print()
    print("hankel minors:")
    print("  Negative moment Hankel minors rule out simple moment total positivity.")
    print()
    print("toeplitz minors:")
    print("  Negative sampled Toeplitz minors damage a simple PF∞ coefficient route.")
    print()
    print("near_zero/tiny_negative:")
    print("  Treat cautiously; rerun at higher precision and altered grids before interpreting.")
    print()
    print("Passing sampled minors is support only, not proof.")


if __name__ == "__main__":
    run_all()
```

```python
# kernel_jensen_extended_scan.py
# ============================================================
# Extended Jensen polynomial scan for the de Bruijn-Newman kernel.
#
# Kernel:
#   Phi(u)=sum_n [2*pi^2*n^4*exp(9u)-3*pi*n^2*exp(5u)]
#          * exp(-pi*n^2*exp(4u))
#
# Deformation:
#   Phi_lambda(u)=exp(lambda*u^2)*Phi(u)
#
# Moments:
#   M_k(lambda)=int_{-inf}^{inf} u^(2k) Phi_lambda(u) du
#
# Auxiliary coefficients:
#   B_lambda(y)=H_lambda(i*sqrt(y))=sum_k A_k y^k/k!
#   A_k = M_k*k!/(2k)!
#
# Jensen polynomial:
#   J_{d,n}(X)=sum_{k=0}^d binom(d,k) A_{n+k} X^k
#
# Expected test condition:
#   roots real and nonpositive
# ============================================================

import mpmath as mp
import numpy as np
from math import comb

mp.mp.dps = 90

LAMBDA_VALUES = [
    mp.mpf("0"),
    mp.mpf("1e-6"),
    mp.mpf("1e-4"),
    mp.mpf("1e-2"),
    mp.mpf("1e-1"),
]

DEGREES = list(range(2, 21))
N_MAX = 80
MAX_MOMENT_K = N_MAX + max(DEGREES)

N_SUM = 100
INTEGRATION_CUTOFF = mp.mpf("8")

IMAG_TOL = 1e-7
POS_TOL = 1e-8
RESID_TOL = 1e-6


def phi_pos_u(u, n_sum=N_SUM):
    u = mp.mpf(u)
    total = mp.mpf("0")

    e4u = mp.e ** (4 * u)
    e5u = mp.e ** (5 * u)
    e9u = mp.e ** (9 * u)

    for n in range(1, n_sum + 1):
        n2 = mp.mpf(n) ** 2
        n4 = n2 ** 2
        term = (2 * mp.pi**2 * n4 * e9u - 3 * mp.pi * n2 * e5u)
        term *= mp.e ** (-mp.pi * n2 * e4u)
        total += term

    return total


def phi(u, n_sum=N_SUM):
    return phi_pos_u(abs(mp.mpf(u)), n_sum=n_sum)


def phi_lambda(u, lam, n_sum=N_SUM):
    u = mp.mpf(u)
    lam = mp.mpf(lam)
    return mp.e ** (lam * u * u) * phi(u, n_sum=n_sum)


def moment_even(k, lam, cutoff=INTEGRATION_CUTOFF):
    k = int(k)
    lam = mp.mpf(lam)

    def integrand(u):
        return (u ** (2 * k)) * phi_lambda(u, lam)

    intervals = [
        mp.mpf("0"),
        mp.mpf("0.2"),
        mp.mpf("0.4"),
        mp.mpf("0.7"),
        mp.mpf("1.0"),
        mp.mpf("1.5"),
        mp.mpf("2.0"),
        mp.mpf("3.0"),
        mp.mpf("4.5"),
        cutoff,
    ]

    return 2 * mp.quad(integrand, intervals)


def compute_moments(lam, max_k):
    return [moment_even(k, lam) for k in range(max_k + 1)]


def compute_A_sequence(moments):
    A = []
    for k, M in enumerate(moments):
        A.append(M * mp.factorial(k) / mp.factorial(2 * k))
    return A


def jensen_coeffs_ascending(A, n, d):
    return [mp.mpf(comb(d, k)) * A[n + k] for k in range(d + 1)]


def normalise_coeffs_float(coeffs_asc):
    coeffs = np.array([float(c) for c in coeffs_asc], dtype=np.float64)
    scale = np.max(np.abs(coeffs))
    if scale == 0 or not np.isfinite(scale):
        raise ValueError("bad coefficient scale")
    coeffs = coeffs / scale
    return coeffs[::-1]


def eval_poly_asc_mp(coeffs_asc, z):
    z = mp.mpc(z)
    total = mp.mpc("0")
    power = mp.mpc("1")

    for c in coeffs_asc:
        total += c * power
        power *= z

    return total


def residual_mp(coeffs_asc, root):
    denom = mp.fsum([abs(c) for c in coeffs_asc])
    if denom == 0:
        denom = mp.mpf("1")

    return abs(eval_poly_asc_mp(coeffs_asc, root)) / denom


def analyse_roots(coeffs_asc):
    coeffs_desc = normalise_coeffs_float(coeffs_asc)
    roots = np.roots(coeffs_desc)

    max_imag = float(np.max(np.abs(np.imag(roots)))) if len(roots) else 0.0
    max_real = float(np.max(np.real(roots))) if len(roots) else -np.inf

    nonreal = int(np.sum(np.abs(np.imag(roots)) > IMAG_TOL))
    positive_real = int(
        np.sum((np.abs(np.imag(roots)) <= IMAG_TOL) & (np.real(roots) > POS_TOL))
    )

    residuals = []
    for r in roots:
        rr = mp.mpc(float(np.real(r)), float(np.imag(r)))
        residuals.append(residual_mp(coeffs_asc, rr))

    max_resid = max(residuals) if residuals else mp.mpf("0")

    ok = (
        nonreal == 0
        and positive_real == 0
        and max_resid < RESID_TOL
    )

    return {
        "ok": ok,
        "roots": roots,
        "nonreal": nonreal,
        "positive_real": positive_real,
        "max_imag": max_imag,
        "max_real": max_real,
        "max_resid": max_resid,
    }


def turan_summary(A):
    vals = []
    for n in range(1, len(A) - 1):
        vals.append(A[n] ** 2 - A[n - 1] * A[n + 1])

    fail = sum(1 for v in vals if v < 0)
    return min(vals), fail


def fmt(x, digits=10):
    try:
        return mp.nstr(x, digits)
    except Exception:
        return str(x)


def run_for_lambda(lam):
    moments = compute_moments(lam, MAX_MOMENT_K)
    A = compute_A_sequence(moments)

    min_turan, turan_fail = turan_summary(A)

    records = []
    failures = []
    errors = []

    for d in DEGREES:
        for n in range(N_MAX + 1):
            coeffs = jensen_coeffs_ascending(A, n, d)

            try:
                ana = analyse_roots(coeffs)
                rec = {
                    "lambda": lam,
                    "degree": d,
                    "n": n,
                    **ana,
                }
                records.append(rec)

                if not ana["ok"]:
                    failures.append(rec)

            except Exception as e:
                errors.append({
                    "lambda": lam,
                    "degree": d,
                    "n": n,
                    "error": str(e),
                })

    return {
        "lambda": lam,
        "records": records,
        "failures": failures,
        "errors": errors,
        "min_turan": min_turan,
        "turan_fail": turan_fail,
    }


def print_failure(r):
    print(
        f"lambda={fmt(r['lambda'], 8):>10} "
        f"d={r['degree']:>2} "
        f"n={r['n']:>3} "
        f"nonreal={r['nonreal']:>3} "
        f"posreal={r['positive_real']:>3} "
        f"max_imag={r['max_imag']:.3e} "
        f"max_real={r['max_real']:.3e} "
        f"max_resid={fmt(r['max_resid'], 8)}"
    )


def run_all():
    print("kernel_jensen_extended_scan.py")
    print("=" * 120)
    print(f"mp.dps={mp.mp.dps}")
    print(f"N_SUM={N_SUM}")
    print(f"lambda values={[fmt(x, 8) for x in LAMBDA_VALUES]}")
    print(f"degrees={DEGREES[0]}..{DEGREES[-1]}")
    print(f"N_MAX={N_MAX}")
    print(f"MAX_MOMENT_K={MAX_MOMENT_K}")
    print()

    print("=" * 120)
    print("SUMMARY BY LAMBDA")
    print("-" * 120)

    header = (
        f"{'lambda':>12} "
        f"{'tests':>8} "
        f"{'fail':>8} "
        f"{'errors':>8} "
        f"{'min Turan':>18} "
        f"{'Turan fail':>12} "
        f"{'worst imag':>12} "
        f"{'worst real':>12} "
        f"{'worst resid':>14}"
    )
    print(header)
    print("-" * len(header))

    results = []

    for lam in LAMBDA_VALUES:
        res = run_for_lambda(lam)
        results.append(res)

        records = res["records"]
        worst_imag = max([r["max_imag"] for r in records], default=float("nan"))
        worst_real = max([r["max_real"] for r in records], default=float("nan"))
        worst_resid = max([r["max_resid"] for r in records], default=mp.mpf("nan"))

        print(
            f"{fmt(lam,8):>12} "
            f"{len(records):>8} "
            f"{len(res['failures']):>8} "
            f"{len(res['errors']):>8} "
            f"{fmt(res['min_turan'],8):>18} "
            f"{res['turan_fail']:>12} "
            f"{worst_imag:>12.3e} "
            f"{worst_real:>12.3e} "
            f"{fmt(worst_resid,8):>14}"
        )

    print()
    print("=" * 120)
    print("FAILURE DETAILS")
    print("-" * 120)

    total_fail = sum(len(r["failures"]) for r in results)
    if total_fail == 0:
        print("No failures.")
    else:
        for res in results:
            if not res["failures"]:
                continue
            print()
            print(f"lambda={fmt(res['lambda'],8)} failures={len(res['failures'])}")
            for r in res["failures"][:50]:
                print_failure(r)
            if len(res["failures"]) > 50:
                print(f"... {len(res['failures']) - 50} more")

    print()
    print("=" * 120)
    print("ERROR DETAILS")
    print("-" * 120)

    total_errors = sum(len(r["errors"]) for r in results)
    if total_errors == 0:
        print("No errors.")
    else:
        for res in results:
            if not res["errors"]:
                continue
            print()
            print(f"lambda={fmt(res['lambda'],8)} errors={len(res['errors'])}")
            for e in res["errors"][:50]:
                print(
                    f"lambda={fmt(e['lambda'],8):>10} "
                    f"d={e['degree']:>2} "
                    f"n={e['n']:>3} "
                    f"error={e['error']}"
                )
            if len(res["errors"]) > 50:
                print(f"... {len(res['errors']) - 50} more")


if __name__ == "__main__":
    run_all()
```

```python
# kernel_jensen_failure_isolation.py
# ============================================================
# Isolate Jensen failures from kernel_jensen_extended_scan.py.
#
# Runs lambda=0 only.
# Finds failing (degree,n) cases.
# Prints strongest failures by:
#   1. nonreal count
#   2. max imaginary part
#   3. residual
#
# Keeps the same kernel and coefficient convention.
# ============================================================

import mpmath as mp
import numpy as np
from math import comb

mp.mp.dps = 100

LAM = mp.mpf("0")

DEGREES = list(range(2, 21))
N_MAX = 80
MAX_MOMENT_K = N_MAX + max(DEGREES)

N_SUM = 100
INTEGRATION_CUTOFF = mp.mpf("8")

IMAG_TOL = 1e-7
POS_TOL = 1e-8
RESID_TOL = 1e-6


def phi_pos_u(u, n_sum=N_SUM):
    u = mp.mpf(u)
    total = mp.mpf("0")

    e4u = mp.e ** (4 * u)
    e5u = mp.e ** (5 * u)
    e9u = mp.e ** (9 * u)

    for n in range(1, n_sum + 1):
        n2 = mp.mpf(n) ** 2
        n4 = n2 ** 2
        term = (2 * mp.pi**2 * n4 * e9u - 3 * mp.pi * n2 * e5u)
        term *= mp.e ** (-mp.pi * n2 * e4u)
        total += term

    return total


def phi(u, n_sum=N_SUM):
    return phi_pos_u(abs(mp.mpf(u)), n_sum=n_sum)


def phi_lambda(u, lam, n_sum=N_SUM):
    u = mp.mpf(u)
    lam = mp.mpf(lam)
    return mp.e ** (lam * u * u) * phi(u, n_sum=n_sum)


def moment_even(k, lam, cutoff=INTEGRATION_CUTOFF):
    k = int(k)
    lam = mp.mpf(lam)

    def integrand(u):
        return (u ** (2 * k)) * phi_lambda(u, lam)

    intervals = [
        mp.mpf("0"),
        mp.mpf("0.2"),
        mp.mpf("0.4"),
        mp.mpf("0.7"),
        mp.mpf("1.0"),
        mp.mpf("1.5"),
        mp.mpf("2.0"),
        mp.mpf("3.0"),
        mp.mpf("4.5"),
        cutoff,
    ]

    return 2 * mp.quad(integrand, intervals)


def compute_moments(lam, max_k):
    return [moment_even(k, lam) for k in range(max_k + 1)]


def compute_A_sequence(moments):
    A = []
    for k, M in enumerate(moments):
        A.append(M * mp.factorial(k) / mp.factorial(2 * k))
    return A


def jensen_coeffs_ascending(A, n, d):
    return [mp.mpf(comb(d, k)) * A[n + k] for k in range(d + 1)]


def coeffs_for_numpy(coeffs_asc):
    vals = np.array([float(c) for c in coeffs_asc], dtype=np.float64)
    scale = np.max(np.abs(vals))
    vals = vals / scale
    return vals[::-1]


def eval_poly_asc_mp(coeffs_asc, z):
    z = mp.mpc(z)
    total = mp.mpc("0")
    power = mp.mpc("1")

    for c in coeffs_asc:
        total += c * power
        power *= z

    return total


def residual_mp(coeffs_asc, root):
    denom = mp.fsum([abs(c) for c in coeffs_asc])
    if denom == 0:
        denom = mp.mpf("1")
    return abs(eval_poly_asc_mp(coeffs_asc, root)) / denom


def analyse(coeffs_asc):
    roots = np.roots(coeffs_for_numpy(coeffs_asc))

    max_imag = float(np.max(np.abs(np.imag(roots)))) if len(roots) else 0.0
    max_real = float(np.max(np.real(roots))) if len(roots) else -np.inf

    nonreal = int(np.sum(np.abs(np.imag(roots)) > IMAG_TOL))
    posreal = int(
        np.sum((np.abs(np.imag(roots)) <= IMAG_TOL) & (np.real(roots) > POS_TOL))
    )

    residuals = []
    for r in roots:
        rr = mp.mpc(float(np.real(r)), float(np.imag(r)))
        residuals.append(residual_mp(coeffs_asc, rr))

    max_resid = max(residuals) if residuals else mp.mpf("0")

    ok = nonreal == 0 and posreal == 0 and max_resid < RESID_TOL

    return {
        "ok": ok,
        "roots": roots,
        "nonreal": nonreal,
        "posreal": posreal,
        "max_imag": max_imag,
        "max_real": max_real,
        "max_resid": max_resid,
    }


def fmt(x, digits=10):
    try:
        return mp.nstr(x, digits)
    except Exception:
        return str(x)


def print_record(r):
    print(
        f"d={r['d']:>2} n={r['n']:>3} "
        f"nonreal={r['nonreal']:>3} "
        f"posreal={r['posreal']:>3} "
        f"max_imag={r['max_imag']:.6e} "
        f"max_real={r['max_real']:.6e} "
        f"max_resid={fmt(r['max_resid'], 8)}"
    )


def main():
    print("kernel_jensen_failure_isolation.py")
    print("=" * 100)
    print(f"lambda={fmt(LAM)}")
    print(f"degrees={DEGREES[0]}..{DEGREES[-1]}")
    print(f"N_MAX={N_MAX}")
    print(f"MAX_MOMENT_K={MAX_MOMENT_K}")
    print()

    print("Computing moments...")
    moments = compute_moments(LAM, MAX_MOMENT_K)
    A = compute_A_sequence(moments)

    records = []
    failures = []

    for d in DEGREES:
        for n in range(N_MAX + 1):
            coeffs = jensen_coeffs_ascending(A, n, d)
            ana = analyse(coeffs)

            rec = {
                "d": d,
                "n": n,
                **ana,
            }

            records.append(rec)

            if not ana["ok"]:
                failures.append(rec)

    print()
    print("SUMMARY")
    print("-" * 100)
    print(f"tests={len(records)}")
    print(f"failures={len(failures)}")

    if records:
        print(f"worst_imag={max(r['max_imag'] for r in records):.6e}")
        print(f"worst_real={max(r['max_real'] for r in records):.6e}")
        print(f"worst_resid={fmt(max(r['max_resid'] for r in records), 8)}")

    print()
    print("FAILURES SORTED BY MAX_IMAG")
    print("-" * 100)
    for r in sorted(failures, key=lambda x: x["max_imag"], reverse=True)[:40]:
        print_record(r)

    print()
    print("FAILURES SORTED BY RESIDUAL")
    print("-" * 100)
    for r in sorted(failures, key=lambda x: x["max_resid"], reverse=True)[:40]:
        print_record(r)

    print()
    print("FAILURE COUNTS BY DEGREE")
    print("-" * 100)
    for d in DEGREES:
        c = sum(1 for r in failures if r["d"] == d)
        print(f"d={d:>2}: {c}")

    print()
    print("FIRST FAILURE BY DEGREE")
    print("-" * 100)
    for d in DEGREES:
        fs = [r for r in failures if r["d"] == d]
        if fs:
            r = min(fs, key=lambda x: x["n"])
            print_record(r)


if __name__ == "__main__":
    main()
```

```python
# kernel_jensen_single_case_verify.py
# ============================================================
# Verify one Jensen polynomial case at high precision.
#
# Target:
#   lambda = 0
#   degree d = 16
#   index n = 54
#
# Computes:
#   moments M_k
#   A_k = M_k*k!/(2k)!
#   Jensen coefficients
#   roots using mpmath.polyroots
#   roots using numpy.roots
#   residuals for both
# ============================================================

import mpmath as mp
import numpy as np
from math import comb

mp.mp.dps = 140

LAM = mp.mpf("0")
D = 16
N = 54

MAX_MOMENT_K = N + D

N_SUM = 140
INTEGRATION_CUTOFF = mp.mpf("9")

IMAG_TOL = mp.mpf("1e-40")
POS_TOL = mp.mpf("1e-40")


def phi_pos_u(u, n_sum=N_SUM):
    u = mp.mpf(u)
    total = mp.mpf("0")

    e4u = mp.e ** (4 * u)
    e5u = mp.e ** (5 * u)
    e9u = mp.e ** (9 * u)

    for n in range(1, n_sum + 1):
        n2 = mp.mpf(n) ** 2
        n4 = n2 ** 2
        term = (2 * mp.pi**2 * n4 * e9u - 3 * mp.pi * n2 * e5u)
        term *= mp.e ** (-mp.pi * n2 * e4u)
        total += term

    return total


def phi(u, n_sum=N_SUM):
    return phi_pos_u(abs(mp.mpf(u)), n_sum=n_sum)


def phi_lambda(u, lam, n_sum=N_SUM):
    u = mp.mpf(u)
    lam = mp.mpf(lam)
    return mp.e ** (lam * u * u) * phi(u, n_sum=n_sum)


def moment_even(k, lam, cutoff=INTEGRATION_CUTOFF):
    k = int(k)
    lam = mp.mpf(lam)

    def integrand(u):
        return (u ** (2 * k)) * phi_lambda(u, lam)

    intervals = [
        mp.mpf("0"),
        mp.mpf("0.15"),
        mp.mpf("0.3"),
        mp.mpf("0.5"),
        mp.mpf("0.75"),
        mp.mpf("1.0"),
        mp.mpf("1.4"),
        mp.mpf("2.0"),
        mp.mpf("3.0"),
        mp.mpf("4.5"),
        mp.mpf("6.5"),
        cutoff,
    ]

    return 2 * mp.quad(integrand, intervals)


def compute_moments(lam, max_k):
    return [moment_even(k, lam) for k in range(max_k + 1)]


def compute_A_sequence(moments):
    return [
        M * mp.factorial(k) / mp.factorial(2 * k)
        for k, M in enumerate(moments)
    ]


def jensen_coeffs_ascending(A, n, d):
    return [mp.mpf(comb(d, k)) * A[n + k] for k in range(d + 1)]


def eval_poly_asc(coeffs, z):
    z = mp.mpc(z)
    total = mp.mpc("0")
    power = mp.mpc("1")

    for c in coeffs:
        total += c * power
        power *= z

    return total


def residual(coeffs, z):
    denom = mp.fsum([abs(c) for c in coeffs])
    if denom == 0:
        denom = mp.mpf("1")
    return abs(eval_poly_asc(coeffs, z)) / denom


def scaled_coeffs_for_mpmath(coeffs_asc):
    scale = max(abs(c) for c in coeffs_asc)
    return [c / scale for c in coeffs_asc]


def numpy_roots(coeffs_asc):
    vals = np.array([float(c) for c in coeffs_asc], dtype=np.float64)
    vals = vals / np.max(np.abs(vals))
    return np.roots(vals[::-1])


def fmt(z, digits=30):
    return mp.nstr(z, digits)


def root_summary(label, roots, coeffs):
    print()
    print(label)
    print("-" * 100)

    nonreal = 0
    posreal = 0
    max_imag = mp.mpf("0")
    max_real = mp.mpf("-inf")
    max_resid = mp.mpf("0")

    converted = []
    for r in roots:
        rr = mp.mpc(r)
        converted.append(rr)

        max_imag = max(max_imag, abs(mp.im(rr)))
        max_real = max(max_real, mp.re(rr))
        max_resid = max(max_resid, residual(coeffs, rr))

        if abs(mp.im(rr)) > IMAG_TOL:
            nonreal += 1
        if abs(mp.im(rr)) <= IMAG_TOL and mp.re(rr) > POS_TOL:
            posreal += 1

    print(f"count={len(converted)}")
    print(f"nonreal={nonreal}")
    print(f"posreal={posreal}")
    print(f"max_imag={fmt(max_imag)}")
    print(f"max_real={fmt(max_real)}")
    print(f"max_resid={fmt(max_resid)}")

    print()
    print("roots:")
    for r in sorted(converted, key=lambda z: (float(mp.re(z)), float(mp.im(z)))):
        print(f"  {fmt(mp.re(r), 25)}  {fmt(mp.im(r), 25)}j  resid={fmt(residual(coeffs, r), 8)}")


def main():
    print("kernel_jensen_single_case_verify.py")
    print("=" * 100)
    print(f"mp.dps={mp.mp.dps}")
    print(f"lambda={LAM}")
    print(f"d={D}")
    print(f"n={N}")
    print(f"MAX_MOMENT_K={MAX_MOMENT_K}")
    print(f"N_SUM={N_SUM}")
    print(f"INTEGRATION_CUTOFF={INTEGRATION_CUTOFF}")
    print()

    print("Computing moments...")
    moments = compute_moments(LAM, MAX_MOMENT_K)
    A = compute_A_sequence(moments)
    coeffs = jensen_coeffs_ascending(A, N, D)

    print()
    print("Jensen coefficients ascending:")
    print("-" * 100)
    for i, c in enumerate(coeffs):
        print(f"c[{i}] = {fmt(c, 40)}")

    coeffs_scaled_asc = scaled_coeffs_for_mpmath(coeffs)
    coeffs_scaled_desc = list(reversed(coeffs_scaled_asc))

    print()
    print("Computing mpmath roots...")
    roots_mp = mp.polyroots(
        coeffs_scaled_desc,
        maxsteps=1000,
        error=False,
        cleanup=True,
    )

    roots_np = numpy_roots(coeffs)

    root_summary("mpmath.polyroots", roots_mp, coeffs)
    root_summary("numpy.roots", roots_np, coeffs)


if __name__ == "__main__":
    main()
```

```python
# kernel_jensen_sturm_single_case.py
# ============================================================
# Sturm real-root count for one Jensen polynomial case.
#
# Target:
#   lambda = 0
#   degree d = 16
#   index n = 54
#
# Jensen polynomial:
#   P(x)=sum_{k=0}^d c_k x^k
#
# Expected:
#   P roots real and negative.
#
# Define:
#   Q(y)=P(-y)=sum_{k=0}^d c_k (-1)^k y^k
#
# Expected:
#   Q has d positive real roots.
#
# This script counts positive real roots of Q using a Sturm chain.
# ============================================================

import mpmath as mp
from math import comb

mp.mp.dps = 160

LAM = mp.mpf("0")
D = 16
N = 54

MAX_MOMENT_K = N + D
N_SUM = 160
INTEGRATION_CUTOFF = mp.mpf("9")


def phi_pos_u(u, n_sum=N_SUM):
    u = mp.mpf(u)
    total = mp.mpf("0")

    e4u = mp.e ** (4 * u)
    e5u = mp.e ** (5 * u)
    e9u = mp.e ** (9 * u)

    for n in range(1, n_sum + 1):
        n2 = mp.mpf(n) ** 2
        n4 = n2 ** 2
        term = (2 * mp.pi**2 * n4 * e9u - 3 * mp.pi * n2 * e5u)
        term *= mp.e ** (-mp.pi * n2 * e4u)
        total += term

    return total


def phi(u, n_sum=N_SUM):
    return phi_pos_u(abs(mp.mpf(u)), n_sum=n_sum)


def phi_lambda(u, lam, n_sum=N_SUM):
    u = mp.mpf(u)
    lam = mp.mpf(lam)
    return mp.e ** (lam * u * u) * phi(u, n_sum=n_sum)


def moment_even(k, lam, cutoff=INTEGRATION_CUTOFF):
    k = int(k)
    lam = mp.mpf(lam)

    def integrand(u):
        return (u ** (2 * k)) * phi_lambda(u, lam)

    intervals = [
        mp.mpf("0"),
        mp.mpf("0.15"),
        mp.mpf("0.3"),
        mp.mpf("0.5"),
        mp.mpf("0.75"),
        mp.mpf("1.0"),
        mp.mpf("1.4"),
        mp.mpf("2.0"),
        mp.mpf("3.0"),
        mp.mpf("4.5"),
        mp.mpf("6.5"),
        cutoff,
    ]

    return 2 * mp.quad(integrand, intervals)


def compute_moments(lam, max_k):
    return [moment_even(k, lam) for k in range(max_k + 1)]


def compute_A_sequence(moments):
    return [
        M * mp.factorial(k) / mp.factorial(2 * k)
        for k, M in enumerate(moments)
    ]


def jensen_coeffs_p_ascending(A, n, d):
    return [mp.mpf(comb(d, k)) * A[n + k] for k in range(d + 1)]


def q_coeffs_desc_from_p_asc(p_asc):
    # Q(y)=P(-y)
    q_asc = [p_asc[k] * ((-1) ** k) for k in range(len(p_asc))]
    return list(reversed(q_asc))


def trim(poly_desc):
    poly = list(poly_desc)
    while len(poly) > 1 and abs(poly[0]) == 0:
        poly.pop(0)
    return poly


def poly_derivative_desc(poly):
    poly = trim(poly)
    n = len(poly) - 1
    if n <= 0:
        return [mp.mpf("0")]
    return [poly[i] * (n - i) for i in range(n)]


def poly_eval_desc(poly, x):
    total = mp.mpf("0")
    for c in poly:
        total = total * x + c
    return total


def poly_divrem_desc(a, b):
    a = trim(a)
    b = trim(b)

    if len(b) == 1 and b[0] == 0:
        raise ZeroDivisionError

    q = [mp.mpf("0")] * max(1, (len(a) - len(b) + 1))
    r = a[:]

    while len(r) >= len(b) and not (len(r) == 1 and r[0] == 0):
        coeff = r[0] / b[0]
        deg_diff = len(r) - len(b)
        q_index = len(q) - deg_diff - 1
        q[q_index] = coeff

        subtract = [coeff * bj for bj in b] + [mp.mpf("0")] * deg_diff
        r = [ri - si for ri, si in zip(r, subtract)]

        r = trim_small(r)

    return trim(q), trim(r)


def trim_small(poly):
    poly = list(poly)
    if not poly:
        return [mp.mpf("0")]

    maxc = max(abs(c) for c in poly)
    if maxc == 0:
        return [mp.mpf("0")]

    tol = mp.mpf("1e-110") * maxc

    while len(poly) > 1 and abs(poly[0]) < tol:
        poly.pop(0)

    for i in range(len(poly)):
        if abs(poly[i]) < tol:
            poly[i] = mp.mpf("0")

    return trim(poly)


def sturm_chain(poly_desc):
    p0 = trim_small(poly_desc)
    p1 = trim_small(poly_derivative_desc(p0))

    chain = [p0, p1]

    while True:
        _, rem = poly_divrem_desc(chain[-2], chain[-1])
        rem = [-c for c in rem]
        rem = trim_small(rem)

        if len(rem) == 1 and rem[0] == 0:
            break

        chain.append(rem)

        if len(rem) == 1:
            break

    return chain


def sign_at(poly, x):
    val = poly_eval_desc(poly, x)
    if val > 0:
        return 1
    if val < 0:
        return -1
    return 0


def sign_at_infinity(poly):
    # positive infinity
    lead = trim(poly)[0]
    return 1 if lead > 0 else -1 if lead < 0 else 0


def variations(signs):
    filtered = [s for s in signs if s != 0]
    count = 0
    for a, b in zip(filtered[:-1], filtered[1:]):
        if a * b < 0:
            count += 1
    return count


def positive_root_count(chain):
    signs_zero = [sign_at(p, mp.mpf("0")) for p in chain]
    signs_inf = [sign_at_infinity(p) for p in chain]

    v_zero = variations(signs_zero)
    v_inf = variations(signs_inf)

    return v_zero - v_inf, v_zero, v_inf, signs_zero, signs_inf


def fmt(x, digits=30):
    return mp.nstr(x, digits)


def main():
    print("kernel_jensen_sturm_single_case.py")
    print("=" * 100)
    print(f"mp.dps={mp.mp.dps}")
    print(f"lambda={LAM}")
    print(f"d={D}")
    print(f"n={N}")
    print(f"N_SUM={N_SUM}")
    print(f"INTEGRATION_CUTOFF={INTEGRATION_CUTOFF}")
    print()

    print("Computing moments...")
    moments = compute_moments(LAM, MAX_MOMENT_K)
    A = compute_A_sequence(moments)
    p_asc = jensen_coeffs_p_ascending(A, N, D)
    q_desc = q_coeffs_desc_from_p_asc(p_asc)

    # Scale Q coefficients
    scale = max(abs(c) for c in q_desc)
    q_desc = [c / scale for c in q_desc]

    print("Building Sturm chain...")
    chain = sturm_chain(q_desc)

    count, v0, vinf, s0, sinf = positive_root_count(chain)

    print()
    print("RESULT")
    print("-" * 100)
    print(f"degree={D}")
    print(f"sturm_chain_length={len(chain)}")
    print(f"positive_roots={count}")
    print(f"expected_positive_roots={D}")
    print(f"V(0)={v0}")
    print(f"V(inf)={vinf}")
    print(f"passes_all_real_negative_P_roots={count == D}")

    print()
    print("Sturm signs at 0:")
    print(s0)

    print()
    print("Sturm signs at +inf:")
    print(sinf)

    print()
    print("Q coefficients descending, scaled:")
    for i, c in enumerate(q_desc):
        print(f"q[{i}] = {fmt(c, 40)}")


if __name__ == "__main__":
    main()
```

```python
# kernel_jensen_sturm_audit.py
# ============================================================
# Sturm audit of Jensen polynomial cases.
#
# Target:
#   lambda = 0
#   degrees = 16..20
#   n = 0..80
#
# Definitions:
#   P(x)=sum_{k=0}^d binom(d,k) A_{n+k} x^k
#   Q(y)=P(-y)
#
# Test:
#   P has d real negative roots iff Q has d positive real roots.
# ============================================================

import mpmath as mp
import sympy as sp
from math import comb

mp.mp.dps = 100

LAM = mp.mpf("0")
DEGREES = list(range(16, 21))
N_MAX = 80
MAX_MOMENT_K = N_MAX + max(DEGREES)

N_SUM = 120
INTEGRATION_CUTOFF = mp.mpf("8.5")
SYMPY_PREC = 80

y = sp.Symbol("y")


def phi_pos_u(u, n_sum=N_SUM):
    u = mp.mpf(u)
    total = mp.mpf("0")

    e4u = mp.e ** (4 * u)
    e5u = mp.e ** (5 * u)
    e9u = mp.e ** (9 * u)

    for n in range(1, n_sum + 1):
        n2 = mp.mpf(n) ** 2
        n4 = n2 ** 2
        term = (2 * mp.pi**2 * n4 * e9u - 3 * mp.pi * n2 * e5u)
        term *= mp.e ** (-mp.pi * n2 * e4u)
        total += term

    return total


def phi(u, n_sum=N_SUM):
    return phi_pos_u(abs(mp.mpf(u)), n_sum=n_sum)


def phi_lambda(u, lam, n_sum=N_SUM):
    u = mp.mpf(u)
    lam = mp.mpf(lam)
    return mp.e ** (lam * u * u) * phi(u, n_sum=n_sum)


def moment_even(k, lam, cutoff=INTEGRATION_CUTOFF):
    k = int(k)
    lam = mp.mpf(lam)

    def integrand(u):
        return (u ** (2 * k)) * phi_lambda(u, lam)

    intervals = [
        mp.mpf("0"),
        mp.mpf("0.2"),
        mp.mpf("0.4"),
        mp.mpf("0.7"),
        mp.mpf("1.0"),
        mp.mpf("1.5"),
        mp.mpf("2.0"),
        mp.mpf("3.0"),
        mp.mpf("4.5"),
        mp.mpf("6.5"),
        cutoff,
    ]

    return 2 * mp.quad(integrand, intervals)


def compute_moments(lam, max_k):
    out = []
    for k in range(max_k + 1):
        out.append(moment_even(k, lam))
        if k % 10 == 0:
            print(f"moment {k}/{max_k}")
    return out


def compute_A_sequence(moments):
    return [
        M * mp.factorial(k) / mp.factorial(2 * k)
        for k, M in enumerate(moments)
    ]


def sympy_float(x):
    return sp.Float(mp.nstr(x, 90), SYMPY_PREC)


def build_Q(A, n, d):
    expr = 0
    for k in range(d + 1):
        c = mp.mpf(comb(d, k)) * A[n + k]
        expr += sympy_float(c) * (-y) ** k
    return sp.Poly(expr, y)


def count_positive_roots(Q):
    return sp.polys.polytools.count_roots(Q.as_expr(), 0, sp.oo)


def main():
    print("kernel_jensen_sturm_audit.py")
    print("=" * 100)
    print(f"lambda={LAM}")
    print(f"degrees={DEGREES[0]}..{DEGREES[-1]}")
    print(f"N_MAX={N_MAX}")
    print(f"MAX_MOMENT_K={MAX_MOMENT_K}")
    print(f"mp.dps={mp.mp.dps}")
    print(f"SYMPY_PREC={SYMPY_PREC}")
    print()

    print("Computing moments...")
    moments = compute_moments(LAM, MAX_MOMENT_K)
    A = compute_A_sequence(moments)

    total = 0
    failures = []

    print()
    print("AUDIT")
    print("-" * 100)
    print(f"{'d':>3} {'n':>4} {'positive_roots':>16} {'expected':>10} {'pass':>8}")

    for d in DEGREES:
        for n in range(N_MAX + 1):
            Q = build_Q(A, n, d)

            count = count_positive_roots(Q)
            count_int = int(count)
            ok = count_int == d

            total += 1

            if not ok:
                failures.append((d, n, count_int))

            print(f"{d:>3} {n:>4} {count_int:>16} {d:>10} {str(ok):>8}")

    print()
    print("SUMMARY")
    print("-" * 100)
    print(f"total={total}")
    print(f"failures={len(failures)}")

    if failures:
        print()
        print("FAILURES")
        for d, n, count_int in failures:
            print(f"d={d} n={n} positive_roots={count_int} expected={d}")


if __name__ == "__main__":
    main()
```

```python
# kernel_jensen_sturm_positive_lambda.py
# ============================================================
# Sturm audit of Jensen polynomial cases for positive lambda.
#
# Lambda values:
#   1e-4, 1e-2, 1e-1
#
# Degrees:
#   16..20
#
# n:
#   0..80
#
# Definitions:
#   P(x)=sum_{k=0}^d binom(d,k) A_{n+k} x^k
#   Q(y)=P(-y)
#
# Test:
#   P has d real negative roots iff Q has d positive real roots.
# ============================================================

import mpmath as mp
import sympy as sp
from math import comb

mp.mp.dps = 100

LAMBDA_VALUES = [
    mp.mpf("1e-4"),
    mp.mpf("1e-2"),
    mp.mpf("1e-1"),
]

DEGREES = list(range(16, 21))
N_MAX = 80
MAX_MOMENT_K = N_MAX + max(DEGREES)

N_SUM = 120
INTEGRATION_CUTOFF = mp.mpf("8.5")
SYMPY_PREC = 80

y = sp.Symbol("y")


def phi_pos_u(u, n_sum=N_SUM):
    u = mp.mpf(u)
    total = mp.mpf("0")

    e4u = mp.e ** (4 * u)
    e5u = mp.e ** (5 * u)
    e9u = mp.e ** (9 * u)

    for n in range(1, n_sum + 1):
        n2 = mp.mpf(n) ** 2
        n4 = n2 ** 2
        term = (2 * mp.pi**2 * n4 * e9u - 3 * mp.pi * n2 * e5u)
        term *= mp.e ** (-mp.pi * n2 * e4u)
        total += term

    return total


def phi(u, n_sum=N_SUM):
    return phi_pos_u(abs(mp.mpf(u)), n_sum=n_sum)


def phi_lambda(u, lam, n_sum=N_SUM):
    u = mp.mpf(u)
    lam = mp.mpf(lam)
    return mp.e ** (lam * u * u) * phi(u, n_sum=n_sum)


def moment_even(k, lam, cutoff=INTEGRATION_CUTOFF):
    k = int(k)
    lam = mp.mpf(lam)

    def integrand(u):
        return (u ** (2 * k)) * phi_lambda(u, lam)

    intervals = [
        mp.mpf("0"),
        mp.mpf("0.2"),
        mp.mpf("0.4"),
        mp.mpf("0.7"),
        mp.mpf("1.0"),
        mp.mpf("1.5"),
        mp.mpf("2.0"),
        mp.mpf("3.0"),
        mp.mpf("4.5"),
        mp.mpf("6.5"),
        cutoff,
    ]

    return 2 * mp.quad(integrand, intervals)


def compute_moments(lam, max_k):
    out = []
    for k in range(max_k + 1):
        out.append(moment_even(k, lam))
        if k % 10 == 0:
            print(f"lambda={mp.nstr(lam, 8)} moment {k}/{max_k}")
    return out


def compute_A_sequence(moments):
    return [
        M * mp.factorial(k) / mp.factorial(2 * k)
        for k, M in enumerate(moments)
    ]


def sympy_float(x):
    return sp.Float(mp.nstr(x, 90), SYMPY_PREC)


def build_Q(A, n, d):
    expr = 0
    for k in range(d + 1):
        c = mp.mpf(comb(d, k)) * A[n + k]
        expr += sympy_float(c) * (-y) ** k
    return sp.Poly(expr, y)


def count_positive_roots(Q):
    return sp.polys.polytools.count_roots(Q.as_expr(), 0, sp.oo)


def run_lambda(lam):
    print()
    print("=" * 100)
    print(f"lambda={mp.nstr(lam, 12)}")
    print("-" * 100)

    print("Computing moments...")
    moments = compute_moments(lam, MAX_MOMENT_K)
    A = compute_A_sequence(moments)

    total = 0
    failures = []

    print()
    print("AUDIT")
    print("-" * 100)
    print(f"{'d':>3} {'n':>4} {'positive_roots':>16} {'expected':>10} {'pass':>8}")

    for d in DEGREES:
        for n in range(N_MAX + 1):
            Q = build_Q(A, n, d)

            count = count_positive_roots(Q)
            count_int = int(count)
            ok = count_int == d

            total += 1

            if not ok:
                failures.append((d, n, count_int))

            print(f"{d:>3} {n:>4} {count_int:>16} {d:>10} {str(ok):>8}")

    print()
    print("LAMBDA SUMMARY")
    print("-" * 100)
    print(f"lambda={mp.nstr(lam, 12)}")
    print(f"total={total}")
    print(f"failures={len(failures)}")

    if failures:
        print()
        print("FAILURES")
        for d, n, count_int in failures:
            print(f"d={d} n={n} positive_roots={count_int} expected={d}")

    return total, failures


def main():
    print("kernel_jensen_sturm_positive_lambda.py")
    print("=" * 100)
    print(f"lambda values={[mp.nstr(v, 8) for v in LAMBDA_VALUES]}")
    print(f"degrees={DEGREES[0]}..{DEGREES[-1]}")
    print(f"N_MAX={N_MAX}")
    print(f"MAX_MOMENT_K={MAX_MOMENT_K}")
    print(f"mp.dps={mp.mp.dps}")
    print(f"SYMPY_PREC={SYMPY_PREC}")

    grand_total = 0
    grand_failures = []

    for lam in LAMBDA_VALUES:
        total, failures = run_lambda(lam)
        grand_total += total
        for item in failures:
            grand_failures.append((lam, *item))

    print()
    print("=" * 100)
    print("GRAND SUMMARY")
    print("-" * 100)
    print(f"total={grand_total}")
    print(f"failures={len(grand_failures)}")

    if grand_failures:
        print()
        print("ALL FAILURES")
        for lam, d, n, count_int in grand_failures:
            print(
                f"lambda={mp.nstr(lam, 12)} "
                f"d={d} n={n} positive_roots={count_int} expected={d}"
            )


if __name__ == "__main__":
    main()
```
